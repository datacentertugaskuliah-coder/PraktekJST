```python
# ============================================================
# JUDUL PROGRAM
# IMPLEMENTASI CONFUSION MATRIX UNTUK PREDIKSI PURCHASE INTENTION
# PADA DATASET ONLINE SHOPPERS PURCHASING INTENTION
# DENGAN ANALISIS DIAGNOSTIK, REKOMENDASI PERBAIKAN,
# DAN PERBAIKAN OTOMATIS BERBASIS HASIL EVALUASI
# ============================================================
#
# DESKRIPSI SINGKAT
# Program ini dirancang untuk membangun model klasifikasi yang logis,
# defensif terhadap error, dan konsisten secara metodologis.
# Fokus utamanya adalah:
# 1. memuat dataset dari beberapa sumber,
# 2. memvalidasi struktur data,
# 3. membangun model klasifikasi,
# 4. mengevaluasi model menggunakan confusion matrix,
# 5. mendeteksi potensi masalah analitik dan logika kritikal,
# 6. menghasilkan 3 rekomendasi perbaikan,
# 7. melakukan perbaikan otomatis berdasarkan rekomendasi tersebut,
# 8. menyimpan model,
# 9. melakukan pengujian deployment sederhana.
#
# CATATAN AKADEMIK
# - Program ini tidak hanya menilai accuracy, tetapi juga precision,
#   recall, F1-score, ROC-AUC, PR-AUC, specificity, dan false positive rate.
# - Program ini dirancang untuk mengurangi dampak imbalance,
#   underfitting, dan overfitting secara logis.
# - Program ini menggunakan alur train-validation-test agar evaluasi
#   lebih jujur dan tidak tercampur dengan proses tuning.
# ============================================================

!pip -q install ucimlrepo joblib

import warnings
warnings.filterwarnings("ignore")

import os
import json
import joblib
import traceback
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    balanced_accuracy_score,
    roc_curve,
    precision_recall_curve,
    auc
)

# ------------------------------------------------------------
# BAGIAN 1. FUNGSI UTILITAS DASAR
# ------------------------------------------------------------

def print_section(title):
    print("\n" + "="*90)
    print(title)
    print("="*90)

def safe_float(x):
    try:
        return float(x)
    except:
        return x

def standardize_target(df, target_col="Revenue"):
    """
    Menstandarkan target menjadi 0 dan 1.
    Fungsi ini dibuat defensif untuk mengurangi error akibat variasi format.
    """
    if target_col not in df.columns:
        raise ValueError(f"Kolom target '{target_col}' tidak ditemukan pada dataset.")

    y = df[target_col]

    # Kasus boolean
    if y.dtype == bool:
        df[target_col] = y.astype(int)
        return df

    # Kasus numerik 0/1
    if pd.api.types.is_numeric_dtype(y):
        uniq = sorted(pd.Series(y).dropna().unique().tolist())
        if set(uniq).issubset({0, 1}):
            df[target_col] = y.astype(int)
            return df

    # Kasus string TRUE/FALSE atau 0/1 dalam bentuk teks
    y_str = y.astype(str).str.strip().str.upper()
    mapped = y_str.map({"TRUE": 1, "FALSE": 0, "1": 1, "0": 0})
    if mapped.isna().sum() == 0:
        df[target_col] = mapped.astype(int)
        return df

    raise ValueError("Kolom Revenue tidak dapat distandarkan menjadi 0/1.")

def validate_dataframe(df):
    """
    Validasi dasar terhadap DataFrame untuk mencegah kegagalan logika kritikal.
    """
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Objek data harus berupa pandas DataFrame.")

    if df.empty:
        raise ValueError("Dataset kosong dan tidak dapat diproses.")

    df.columns = [str(c).strip() for c in df.columns]

    if "Revenue" not in df.columns:
        raise ValueError("Kolom target 'Revenue' wajib tersedia.")

    # Menghapus duplikasi untuk mencegah bias ringan
    n_before = len(df)
    df = df.drop_duplicates().reset_index(drop=True)
    n_after = len(df)

    print(f"Jumlah baris sebelum hapus duplikasi : {n_before}")
    print(f"Jumlah baris sesudah hapus duplikasi : {n_after}")

    return df

def load_dataset(choice="uci", local_path=None):
    """
    Memuat dataset dari:
    - upload manual,
    - file lokal,
    - UCI repository.
    """
    if choice == "upload":
        try:
            from google.colab import files
        except Exception:
            raise EnvironmentError("Mode upload hanya dapat digunakan di Google Colab.")

        print("Silakan unggah file CSV dataset...")
        uploaded = files.upload()
        if len(uploaded) == 0:
            raise ValueError("Tidak ada file yang diunggah.")
        filename = list(uploaded.keys())[0]
        df = pd.read_csv(filename)
        source_info = f"Dataset dimuat dari upload manual: {filename}"

    elif choice == "local":
        if local_path is None:
            raise ValueError("Parameter local_path wajib diisi untuk choice='local'.")
        if not os.path.exists(local_path):
            raise FileNotFoundError(f"File tidak ditemukan pada path: {local_path}")
        df = pd.read_csv(local_path)
        source_info = f"Dataset dimuat dari file lokal: {local_path}"

    elif choice == "uci":
        dataset = fetch_ucirepo(id=468)
        X = dataset.data.features.copy()
        y = dataset.data.targets.copy()
        if isinstance(y, pd.DataFrame):
            y = y.iloc[:, 0]
        df = X.copy()
        df["Revenue"] = y
        source_info = "Dataset dimuat langsung dari UCI (id=468)."

    else:
        raise ValueError("Pilihan sumber data tidak valid. Gunakan 'upload', 'local', atau 'uci'.")

    df = validate_dataframe(df)
    df = standardize_target(df, target_col="Revenue")

    print(source_info)
    print(f"Ukuran dataset akhir: {df.shape}")
    return df

def get_metrics(y_true, y_pred, y_proba):
    """
    Menghitung metrik evaluasi utama klasifikasi biner.
    """
    cm = confusion_matrix(y_true, y_pred)
    if cm.shape != (2, 2):
        raise ValueError("Confusion matrix bukan 2x2. Target harus biner.")

    tn, fp, fn, tp = cm.ravel()
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0.0
    fpr = fp / (fp + tn) if (fp + tn) != 0 else 0.0

    precision_curve, recall_curve, _ = precision_recall_curve(y_true, y_proba)
    pr_auc = auc(recall_curve, precision_curve)

    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Balanced_Accuracy": balanced_accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1": f1_score(y_true, y_pred, zero_division=0),
        "ROC_AUC": roc_auc_score(y_true, y_proba),
        "PR_AUC": pr_auc,
        "Specificity": specificity,
        "False_Positive_Rate": fpr,
        "TN": int(tn), "FP": int(fp), "FN": int(fn), "TP": int(tp)
    }

def choose_best_threshold(y_true, y_proba, beta=1.0):
    """
    Menentukan threshold terbaik secara aman.
    Jika thresholds kosong, fallback ke 0.5.
    """
    precision_arr, recall_arr, thresholds = precision_recall_curve(y_true, y_proba)

    if len(thresholds) == 0:
        return 0.5, 0.0

    scores = []
    for i in range(len(thresholds)):
        p = precision_arr[i]
        r = recall_arr[i]
        denom = (beta**2 * p) + r
        if denom == 0:
            score = 0.0
        else:
            score = (1 + beta**2) * (p * r) / denom
        scores.append(score)

    best_idx = int(np.argmax(scores))
    return float(thresholds[best_idx]), float(scores[best_idx])

def evaluate_pipeline(model, X_train, y_train, X_valid, y_valid, X_test, y_test, threshold=0.5):
    """
    Melatih model, memprediksi, dan mengevaluasi train-validation-test.
    """
    model.fit(X_train, y_train)

    train_proba = model.predict_proba(X_train)[:, 1]
    valid_proba = model.predict_proba(X_valid)[:, 1]
    test_proba  = model.predict_proba(X_test)[:, 1]

    train_pred = (train_proba >= threshold).astype(int)
    valid_pred = (valid_proba >= threshold).astype(int)
    test_pred  = (test_proba >= threshold).astype(int)

    train_metrics = get_metrics(y_train, train_pred, train_proba)
    valid_metrics = get_metrics(y_valid, valid_pred, valid_proba)
    test_metrics  = get_metrics(y_test, test_pred, test_proba)

    return {
        "model": model,
        "threshold": threshold,
        "train_metrics": train_metrics,
        "valid_metrics": valid_metrics,
        "test_metrics": test_metrics,
        "train_pred": train_pred,
        "valid_pred": valid_pred,
        "test_pred": test_pred,
        "train_proba": train_proba,
        "valid_proba": valid_proba,
        "test_proba": test_proba
    }

def plot_conf_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots(figsize=(5, 4))
    ConfusionMatrixDisplay(cm, display_labels=["Tidak Beli", "Beli"]).plot(ax=ax, colorbar=False)
    ax.set_title(title)
    plt.show()

def simple_predict(input_df, artifact):
    """
    Prediksi sederhana untuk deployment awal.
    """
    if not isinstance(input_df, pd.DataFrame):
        raise TypeError("Input prediksi harus berupa pandas DataFrame.")

    required_cols = artifact["feature_columns"]
    missing_cols = [c for c in required_cols if c not in input_df.columns]
    if len(missing_cols) > 0:
        raise ValueError(f"Kolom input belum lengkap: {missing_cols}")

    input_df = input_df[required_cols].copy()

    model = artifact["model"]
    threshold = artifact["threshold"]

    proba = model.predict_proba(input_df)[:, 1]
    pred = (proba >= threshold).astype(int)

    out = input_df.copy()
    out["Probabilitas_Beli"] = np.round(proba, 4)
    out["Prediksi_Revenue"] = pred
    out["Label_Prediksi"] = out["Prediksi_Revenue"].map({0: "Tidak Beli", 1: "Beli"})
    return out

# ------------------------------------------------------------
# BAGIAN 2. PEMILIHAN SUMBER DATASET
# ------------------------------------------------------------

print_section("PEMILIHAN SUMBER DATASET")
print("1 = Upload file CSV secara manual")
print("2 = Membaca file CSV dari path lokal")
print("3 = Mengambil dataset langsung dari UCI")

pilihan = input("Masukkan pilihan (1/2/3): ").strip()

try:
    if pilihan == "1":
        df = load_dataset(choice="upload")
    elif pilihan == "2":
        path_file = input("Masukkan path file CSV: ").strip()
        df = load_dataset(choice="local", local_path=path_file)
    elif pilihan == "3":
        df = load_dataset(choice="uci")
    else:
        raise ValueError("Pilihan tidak valid. Gunakan 1, 2, atau 3.")
except Exception as e:
    print("Terjadi error saat memuat dataset:", e)
    raise

print_section("PRATINJAU DATA")
display(df.head())
display(df.isna().sum().to_frame("Jumlah Missing"))

print("\nDistribusi target:")
display(df["Revenue"].value_counts().rename_axis("Revenue").to_frame("Jumlah"))
display(df["Revenue"].value_counts(normalize=True).rename_axis("Revenue").to_frame("Proporsi"))

# ------------------------------------------------------------
# BAGIAN 3. VISUALISASI EKSPLORATIF
# ------------------------------------------------------------

print_section("VISUALISASI EKSPLORATIF")

plt.figure(figsize=(6,4))
target_counts = df["Revenue"].value_counts().sort_index()
plt.bar(["Tidak Beli (0)", "Beli (1)"], target_counts.values)
plt.title("Distribusi Kelas Target Revenue")
plt.ylabel("Jumlah Sesi")
plt.show()

numeric_all = df.select_dtypes(include=np.number).columns.tolist()
if "Revenue" in numeric_all:
    numeric_all.remove("Revenue")

if len(numeric_all) > 0:
    corr_target = df[numeric_all + ["Revenue"]].corr(numeric_only=True)["Revenue"].drop("Revenue").sort_values()
    plt.figure(figsize=(8,6))
    plt.barh(corr_target.index, corr_target.values)
    plt.title("Korelasi Fitur Numerik terhadap Revenue")
    plt.xlabel("Koefisien Korelasi")
    plt.show()

# ------------------------------------------------------------
# BAGIAN 4. PEMISAHAN FITUR DAN TARGET
# ------------------------------------------------------------

print_section("PEMISAHAN FITUR DAN TARGET")

X = df.drop(columns=["Revenue"]).copy()
y = df["Revenue"].copy()

if y.nunique() != 2:
    raise ValueError("Target harus terdiri dari dua kelas untuk klasifikasi biner.")

categorical_cols = X.select_dtypes(include=["object", "bool", "category"]).columns.tolist()
numeric_cols = X.select_dtypes(include=np.number).columns.tolist()

if len(numeric_cols) == 0 and len(categorical_cols) == 0:
    raise ValueError("Tidak ada fitur yang dapat diproses.")

print("Kolom numerik     :", numeric_cols)
print("Kolom kategorikal :", categorical_cols)

# ------------------------------------------------------------
# BAGIAN 5. PREPROCESSING DAN SPLIT
# ------------------------------------------------------------

print_section("PREPROCESSING DAN PEMBAGIAN DATA")

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_cols),
    ("cat", categorical_transformer, categorical_cols)
])

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full, test_size=0.20, random_state=42, stratify=y_train_full
)

print("Train      :", X_train.shape, y_train.shape)
print("Validation :", X_valid.shape, y_valid.shape)
print("Test       :", X_test.shape, y_test.shape)

# ------------------------------------------------------------
# BAGIAN 6. MODEL AWAL
# ------------------------------------------------------------

print_section("PEMBANGUNAN MODEL AWAL")

# Model awal dipilih sengaja sederhana dan stabil.
baseline_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(
        class_weight="balanced",
        max_iter=5000,
        random_state=42
    ))
])

baseline_result = evaluate_pipeline(
    baseline_model, X_train, y_train, X_valid, y_valid, X_test, y_test, threshold=0.5
)

baseline_table = pd.DataFrame([
    baseline_result["train_metrics"],
    baseline_result["valid_metrics"],
    baseline_result["test_metrics"]
], index=["Train", "Validation", "Test"])

display(baseline_table.round(4))
plot_conf_matrix(y_test, baseline_result["test_pred"], "Confusion Matrix - Model Awal (Test)")

# ------------------------------------------------------------
# BAGIAN 7. DIAGNOSIS ANALITIK DAN 3 REKOMENDASI
# ------------------------------------------------------------

print_section("DIAGNOSIS ANALITIK DAN REKOMENDASI")

def diagnose_and_recommend(result):
    train_m = result["train_metrics"]
    valid_m = result["valid_metrics"]

    issues = []
    recommendations = []

    # Diagnosis 1: dampak imbalance / recall kelas positif rendah
    if valid_m["Recall"] < 0.70:
        issues.append("imbalance_recall_problem")
        recommendations.append(
            "Rekomendasi 1: lakukan optimasi threshold keputusan pada validation set "
            "agar recall kelas pembelian meningkat dan model tidak terlalu konservatif."
        )
    else:
        recommendations.append(
            "Rekomendasi 1: pertahankan evaluasi berbasis confusion matrix dan recall, "
            "karena kelas positif pada data klasifikasi tetap perlu dijaga."
        )

    # Diagnosis 2: overfitting
    if (train_m["F1"] - valid_m["F1"]) > 0.08:
        issues.append("overfitting_problem")
        recommendations.append(
            "Rekomendasi 2: sederhanakan model atau tambah regularisasi untuk menurunkan "
            "gap antara performa train dan validation."
        )
    # Diagnosis 3: underfitting
    elif train_m["F1"] < 0.60 and valid_m["F1"] < 0.60:
        issues.append("underfitting_problem")
        recommendations.append(
            "Rekomendasi 2: tingkatkan kapasitas model secara terkontrol melalui "
            "hyperparameter tuning agar model tidak terlalu sederhana."
        )
    else:
        recommendations.append(
            "Rekomendasi 2: lakukan tuning hyperparameter terstruktur agar model tidak "
            "berhenti pada konfigurasi default."
        )

    # Diagnosis 4: precision rendah / false positive tinggi
    if valid_m["Precision"] < 0.40 or valid_m["False_Positive_Rate"] > 0.25:
        issues.append("precision_problem")
        recommendations.append(
            "Rekomendasi 3: kendalikan false positive melalui penyesuaian threshold "
            "atau penguatan regularisasi agar intervensi marketing tidak terlalu salah sasaran."
        )
    else:
        recommendations.append(
            "Rekomendasi 3: validasi akhir perlu menjaga keseimbangan precision dan recall "
            "agar model tetap relevan secara operasional."
        )

    return issues, recommendations[:3]

issues, recommendations = diagnose_and_recommend(baseline_result)

print("Masalah yang terdeteksi:", issues if issues else ["tidak ada masalah dominan"])
print("\nTIGA REKOMENDASI PERBAIKAN")
for i, rec in enumerate(recommendations, 1):
    print(f"{i}. {rec}")

# ------------------------------------------------------------
# BAGIAN 8. PERBAIKAN OTOMATIS BERDASARKAN REKOMENDASI
# ------------------------------------------------------------

print_section("PERBAIKAN OTOMATIS BERDASARKAN REKOMENDASI")

# Logika auto-fix:
# - Jika indikasi overfitting: prioritaskan Logistic Regression dengan tuning regularisasi.
# - Jika indikasi underfitting: bandingkan Logistic Regression dan Random Forest terkontrol.
# - Jika masalah recall: optimasi threshold berbasis F2 agar recall lebih diprioritaskan.
# - Jika masalah precision/FPR: threshold disesuaikan lebih hati-hati.
#
# Program ini menerapkan perbaikan otomatis tanpa menebak-nebak hasil,
# tetapi tetap menjaga validitas train-validation-test.

candidate_results = {}

# Kandidat 1: Logistic Regression tuning regularisasi
lr_pipe = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(
        class_weight="balanced",
        max_iter=5000,
        random_state=42
    ))
])

lr_grid = {
    "model__C": [0.05, 0.1, 0.5, 1.0, 2.0, 5.0],
    "model__solver": ["liblinear", "lbfgs"]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

lr_search = GridSearchCV(
    estimator=lr_pipe,
    param_grid=lr_grid,
    scoring="f1",
    cv=cv,
    n_jobs=-1,
    error_score="raise"
)
lr_search.fit(X_train, y_train)
lr_best = lr_search.best_estimator_

# Threshold otomatis untuk LR
lr_valid_proba = lr_best.predict_proba(X_valid)[:, 1]
if "imbalance_recall_problem" in issues:
    lr_threshold, lr_thr_score = choose_best_threshold(y_valid, lr_valid_proba, beta=2.0)
else:
    lr_threshold, lr_thr_score = choose_best_threshold(y_valid, lr_valid_proba, beta=1.0)

lr_result = evaluate_pipeline(lr_best, X_train, y_train, X_valid, y_valid, X_test, y_test, threshold=lr_threshold)
candidate_results["LogisticRegression_AutoFix"] = {
    "search_best_params": lr_search.best_params_,
    "threshold_score": lr_thr_score,
    "result": lr_result
}

# Kandidat 2: Random Forest terkontrol untuk mengatasi underfitting bila perlu
rf_pipe = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

rf_grid = {
    "model__n_estimators": [200, 300],
    "model__max_depth": [6, 8, 10],
    "model__min_samples_leaf": [3, 5, 8]
}

rf_search = GridSearchCV(
    estimator=rf_pipe,
    param_grid=rf_grid,
    scoring="f1",
    cv=cv,
    n_jobs=-1,
    error_score="raise"
)
rf_search.fit(X_train, y_train)
rf_best = rf_search.best_estimator_

rf_valid_proba = rf_best.predict_proba(X_valid)[:, 1]
if "imbalance_recall_problem" in issues:
    rf_threshold, rf_thr_score = choose_best_threshold(y_valid, rf_valid_proba, beta=2.0)
else:
    rf_threshold, rf_thr_score = choose_best_threshold(y_valid, rf_valid_proba, beta=1.0)

rf_result = evaluate_pipeline(rf_best, X_train, y_train, X_valid, y_valid, X_test, y_test, threshold=rf_threshold)
candidate_results["RandomForest_AutoFix"] = {
    "search_best_params": rf_search.best_params_,
    "threshold_score": rf_thr_score,
    "result": rf_result
}

# Pemilihan model akhir secara logis:
# - Prioritas 1: Validation F1
# - Prioritas 2: Validation Recall jika recall menjadi masalah
# - Prioritas 3: Validation Precision bila false positive menjadi masalah
def select_final_model(candidates, issues):
    rows = []
    for name, obj in candidates.items():
        r = obj["result"]
        vm = r["valid_metrics"]
        rows.append({
            "Model": name,
            "Validation_F1": vm["F1"],
            "Validation_Recall": vm["Recall"],
            "Validation_Precision": vm["Precision"],
            "Validation_FPR": vm["False_Positive_Rate"],
            "Threshold": r["threshold"]
        })
    df_sel = pd.DataFrame(rows)

    if "imbalance_recall_problem" in issues:
        df_sel = df_sel.sort_values(
            by=["Validation_F1", "Validation_Recall", "Validation_Precision"],
            ascending=[False, False, False]
        )
    elif "precision_problem" in issues:
        df_sel = df_sel.sort_values(
            by=["Validation_F1", "Validation_Precision", "Validation_FPR"],
            ascending=[False, False, True]
        )
    else:
        df_sel = df_sel.sort_values(
            by=["Validation_F1", "Validation_Precision", "Validation_Recall"],
            ascending=[False, False, False]
        )

    return df_sel.reset_index(drop=True)

selection_table = select_final_model(candidate_results, issues)
display(selection_table.round(4))

final_name = selection_table.iloc[0]["Model"]
final_object = candidate_results[final_name]
final_result = final_object["result"]
final_best_params = final_object["search_best_params"]

print(f"Model akhir terpilih: {final_name}")
print(f"Best parameter      : {final_best_params}")
print(f"Threshold akhir     : {final_result['threshold']:.4f}")

# ------------------------------------------------------------
# BAGIAN 9. EVALUASI AKHIR SETELAH AUTO-FIX
# ------------------------------------------------------------

print_section("EVALUASI AKHIR SETELAH PERBAIKAN OTOMATIS")

final_table = pd.DataFrame([
    final_result["train_metrics"],
    final_result["valid_metrics"],
    final_result["test_metrics"]
], index=["Train", "Validation", "Test"])

display(final_table.round(4))
plot_conf_matrix(y_test, final_result["test_pred"], f"Confusion Matrix - {final_name} (Test)")

print("\nClassification Report - Test Set")
print(classification_report(y_test, final_result["test_pred"], target_names=["Tidak Beli", "Beli"], zero_division=0))

# Perbandingan sebelum dan sesudah
comparison = pd.DataFrame({
    "Sebelum_AutoFix": pd.Series(baseline_result["test_metrics"]),
    "Sesudah_AutoFix": pd.Series(final_result["test_metrics"])
})
comparison["Delta"] = comparison["Sesudah_AutoFix"] - comparison["Sebelum_AutoFix"]

selected_metrics = [
    "Accuracy", "Balanced_Accuracy", "Precision", "Recall", "F1",
    "ROC_AUC", "PR_AUC", "Specificity", "False_Positive_Rate",
    "TN", "FP", "FN", "TP"
]
comparison = comparison.loc[selected_metrics]

print_section("PERBANDINGAN HASIL TEST SET SEBELUM DAN SESUDAH AUTO-FIX")
display(comparison.round(4))

# ------------------------------------------------------------
# BAGIAN 10. VISUALISASI ROC DAN PR CURVE
# ------------------------------------------------------------

print_section("VISUALISASI ROC CURVE DAN PRECISION-RECALL CURVE")

# ROC Curve
fpr_vals, tpr_vals, _ = roc_curve(y_test, final_result["test_proba"])
roc_auc = roc_auc_score(y_test, final_result["test_proba"])

plt.figure(figsize=(6,4))
plt.plot(fpr_vals, tpr_vals, label=f"ROC AUC = {roc_auc:.4f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Model Akhir")
plt.legend()
plt.show()

# PR Curve
precision_vals, recall_vals, _ = precision_recall_curve(y_test, final_result["test_proba"])
pr_auc = auc(recall_vals, precision_vals)

plt.figure(figsize=(6,4))
plt.plot(recall_vals, precision_vals, label=f"PR AUC = {pr_auc:.4f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - Model Akhir")
plt.legend()
plt.show()

# ------------------------------------------------------------
# BAGIAN 11. PENYIMPANAN MODEL
# ------------------------------------------------------------

print_section("PENYIMPANAN MODEL")

MODEL_PATH = "/content/model_online_shoppers_autofix.joblib"
META_PATH = "/content/model_online_shoppers_autofix_metadata.json"

artifact = {
    "model": final_result["model"],
    "threshold": final_result["threshold"],
    "feature_columns": X.columns.tolist(),
    "numeric_cols": numeric_cols,
    "categorical_cols": categorical_cols,
    "best_params": final_best_params,
    "selected_model_name": final_name
}

joblib.dump(artifact, MODEL_PATH)

metadata = {
    "selected_model_name": final_name,
    "threshold": safe_float(final_result["threshold"]),
    "best_params": final_best_params,
    "recommendations": recommendations,
    "issues_detected": issues,
    "test_metrics": {k: safe_float(v) for k, v in final_result["test_metrics"].items()}
}

with open(META_PATH, "w", encoding="utf-8") as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)

print(f"Model tersimpan di   : {MODEL_PATH}")
print(f"Metadata tersimpan di: {META_PATH}")

# ------------------------------------------------------------
# BAGIAN 12. UJI DEPLOYMENT SEDERHANA
# ------------------------------------------------------------

print_section("UJI DEPLOYMENT SEDERHANA")

loaded_artifact = joblib.load(MODEL_PATH)

# Lima data uji sederhana diambil dari test set agar struktur kolom pasti valid.
n_cases = 5 if len(X_test) >= 5 else len(X_test)
sample_cases = X_test.head(n_cases).copy()
actual_cases = y_test.head(n_cases).reset_index(drop=True)

deployment_result = simple_predict(sample_cases, loaded_artifact).reset_index(drop=True)
deployment_result["Aktual_Revenue"] = actual_cases
deployment_result["Aktual_Label"] = deployment_result["Aktual_Revenue"].map({0: "Tidak Beli", 1: "Beli"})
deployment_result["Kesesuaian"] = np.where(
    deployment_result["Prediksi_Revenue"] == deployment_result["Aktual_Revenue"],
    "Benar", "Salah"
)

display(deployment_result)

success_rate = (deployment_result["Kesesuaian"] == "Benar").mean()
print(f"Tingkat keberhasilan pada {n_cases} data uji sederhana: {success_rate:.2%}")

# ------------------------------------------------------------
# BAGIAN 13. INTERPRETASI OTOMATIS HASIL
# ------------------------------------------------------------

print_section("INTERPRETASI OTOMATIS HASIL")

before_test = baseline_result["test_metrics"]
after_test = final_result["test_metrics"]

if after_test["Recall"] > before_test["Recall"]:
    print(f"1. Recall meningkat dari {before_test['Recall']:.4f} menjadi {after_test['Recall']:.4f}, "
          "yang berarti model hasil perbaikan lebih baik dalam menangkap sesi yang benar-benar berakhir pembelian.")
else:
    print(f"1. Recall tidak meningkat secara berarti ({before_test['Recall']:.4f} -> {after_test['Recall']:.4f}).")

if after_test["Precision"] > before_test["Precision"]:
    print(f"2. Precision meningkat dari {before_test['Precision']:.4f} menjadi {after_test['Precision']:.4f}, "
          "yang berarti prediksi pembelian menjadi lebih tepat.")
else:
    print(f"2. Precision tidak meningkat secara berarti ({before_test['Precision']:.4f} -> {after_test['Precision']:.4f}).")

if after_test["F1"] > before_test["F1"]:
    print(f"3. F1-score meningkat dari {before_test['F1']:.4f} menjadi {after_test['F1']:.4f}, "
          "yang menunjukkan keseimbangan precision dan recall menjadi lebih baik.")
else:
    print(f"3. F1-score tidak meningkat secara berarti ({before_test['F1']:.4f} -> {after_test['F1']:.4f}).")

if after_test["False_Positive_Rate"] < before_test["False_Positive_Rate"]:
    print(f"4. False Positive Rate turun dari {before_test['False_Positive_Rate']:.4f} menjadi {after_test['False_Positive_Rate']:.4f}, "
          "sehingga risiko intervensi yang salah sasaran menjadi lebih rendah.")
else:
    print(f"4. False Positive Rate tidak turun ({before_test['False_Positive_Rate']:.4f} -> {after_test['False_Positive_Rate']:.4f}).")


print("\nPROGRAM SELESAI DIJALANKAN.")
```
