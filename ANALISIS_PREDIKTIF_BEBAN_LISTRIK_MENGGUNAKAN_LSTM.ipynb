```python
# =============================================================================
# JUDUL PROGRAM: ANALISIS PREDIKTIF BEBAN LISTRIK MENGGUNAKAN LSTM
# DESKRIPSI: Studi kasus prediksi beban listrik menggunakan dataset PGCB Hourly
#            Generation Dataset (Bangladesh) dengan pendekatan Deep Learning.
# TUJUAN: Membangun model LSTM untuk memprediksi beban listrik masa depan
#         dengan meminimalkan overfitting, underfitting, dan bias data.
#
# =============================================================================

# =============================================================================
# BAGIAN 1: INSTALASI DAN IMPORT LIBRARY
# =============================================================================
# Instruksi: Pastikan lingkungan Google Colab memiliki akses internet.
# Library yang diperlukan diinstal secara otomatis sebelum eksekusi.
!pip install pandas numpy matplotlib seaborn scikit-learn tensorflow requests -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras import regularizers # Import regularizers

# =============================================================================
# BAGIAN 2: KONFIGURASI VISUALISASI DAN LINGKUNGAN
# =============================================================================
# Konfigurasi ini diterapkan untuk memastikan konsistensi visualisasi
# dan keterbacaan grafik dalam laporan akademik.
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'sans-serif'

# =============================================================================
# BAGIAN 3: FUNGSI MUAT DATA (ADAPTIF DAN INTEGRITAS)
# =============================================================================
def load_dataset_academic():
    """
    Fungsi ini dirancang untuk memuat dataset dengan strategi adaptif.
    Mengingat URL yang diberikan adalah halaman informasi dataset,
    kode ini mencoba mengakses file langsung atau menggunakan data simulasi
    jika akses eksternal gagal demi menjaga kelangsungan analisis.
    """
    print("=" * 60)
    print("LANGKAH 1: INISIALISASI DAN MUAT DATA")
    print("=" * 60)

    # URL Referensi (Halaman Dataset UCI)
    dataset_page_url = 'https://archive.ics.uci.edu/dataset/1175/pgcb+hourly+generation+dataset+(bangladesh)'

    # URL Alternatif untuk File CSV Langsung
    csv_urls = [
        'https://archive.ics.uci.edu/ml/machine-learning-databases/00515/PGCB_Hourly_Generation_Dataset.csv',
        'https://archive.ics.uci.edu/ml/machine-learning-databases/00515/PGCB_Hourly_Generation_Dataset.zip'
    ]

    try:
        print("🔍 Mencoba mengakses file data langsung...")
        for url in csv_urls:
            try:
                response = requests.get(url, timeout=10)
                if response.status_code == 200:
                    print(f"✅ Berhasil mengakses: {url}")
                    if url.endswith('.csv'):
                        df = pd.read_csv(url)
                        return df
                    elif url.endswith('.zip'):
                        import zipfile
                        with zipfile.ZipFile('dataset.zip', 'w') as zip_file:
                            zip_file.writestr('dataset.zip', response.content)
                        with zipfile.ZipFile('dataset.zip', 'r') as zip_ref:
                            zip_ref.extractall('dataset_folder')
                        df = pd.read_csv('dataset_folder/PGCB_Hourly_Generation_Dataset.csv')
                        return df
            except Exception as e:
                print(f"⚠️ Gagal dari URL: {e}")

        # Fallback: Data Simulasi (Hanya jika akses gagal)
        print("⚠️ Akses data eksternal gagal. Menggunakan data simulasi untuk demonstrasi logika.")
        print("💡 Catatan: Untuk hasil nyata, silakan upload file CSV ke Google Colab.")

        # Struktur Data Simulasi (Merepresentasikan pola PGCB)
        dates = pd.date_range(start='2020-01-01', end='2023-12-31', freq='h')
        np.random.seed(42)
        hourly = np.arange(len(dates))
        daily_pattern = 100 + 50 * np.sin(2 * np.pi * hourly / 24)
        weekly_pattern = 20 * np.sin(2 * np.pi * hourly / (24 * 7))
        noise = np.random.normal(0, 15, len(dates))
        generation = daily_pattern + weekly_pattern + noise + 500

        df = pd.DataFrame({
            'Date': dates,
            'Time': dates.time,
            'Generation_MW': generation
        })
        return df

    except Exception as e:
        print(f"❌ Error kritis: {e}")
        raise

# Eksekusi Fungsi Muat Data
df = load_dataset_academic()

# =============================================================================
# BAGIAN 4: ANALISIS DATA AWAL (EXPLORATORY DATA ANALYSIS)
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 2: ANALISIS DATA AWAL (EDA)")
print("=" * 60)

# 4.1 Informasi Struktur Data
print(f"📊 Dimensi Data: {df.shape}")
print(f"📊 Kolom: {df.columns.tolist()}")
print(f"📊 Statistik Deskriptif:\n{df.describe()}")

# 4.2 Visualisasi Tren Harian
plt.figure(figsize=(14, 6))
df['Generation_MW'].plot(title='Tren Beban Listrik Harian', linewidth=1)
plt.xlabel('Waktu')
plt.ylabel('Beban (MW)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 4.3 Visualisasi Pola Mingguan (Heatmap)
# Analisis ini membantu mengidentifikasi pola musiman harian dan mingguan
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['Hour'] = df['Date'].dt.hour
plt.figure(figsize=(14, 6))
pivot = df.pivot_table(values='Generation_MW', index='Hour', columns='DayOfWeek', aggfunc='mean')
sns.heatmap(pivot, cmap='YlOrRd', annot=False, fmt='.0f')
plt.title('Pola Beban Listrik Mingguan')
plt.xlabel('Hari (0=Senin, 6=Minggu)')
plt.ylabel('Jam')
plt.tight_layout()
plt.show()

# 4.4 Analisis Distribusi (Mencegah Imbalance/Skewness)
plt.figure(figsize=(10, 5))
sns.histplot(df['Generation_MW'], bins=50, kde=True)
plt.title('Distribusi Beban Listrik')
plt.xlabel('Beban (MW)')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()

# =============================================================================
# BAGIAN 5: PRAPROSESAN DATA
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 3: PRAPROSESAN DATA")
print("=" * 60)

# 5.1 Penanganan Missing Values
# Interpolasi linear digunakan untuk menjaga kontinuitas data deret waktu
print("🔍 Mengecek Missing Values...")
missing = df.isnull().sum()
if missing.sum() > 0:
    df = df.interpolate(method='linear')
    print("✅ Missing values diimputasi dengan interpolasi linear")
else:
    print("✅ Tidak ada missing values")

# 5.2 Feature Engineering
# Menambahkan fitur waktu untuk membantu model memahami pola musiman
print("\n🔧 Membuat Fitur Tambahan...")
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['Hour'] = df['Date'].dt.hour
df['IsWeekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
df['Month'] = df['Date'].dt.month

# 5.3 Scaling Data
# MinMaxScaler digunakan untuk menormalkan data ke rentang [0, 1]
# Hal ini krusial untuk stabilitas konvergensi pada model LSTM
print("\n📏 Scaling Data...")
scaler = MinMaxScaler(feature_range=(0, 1))
df['Generation_Scaled'] = scaler.fit_transform(df[['Generation_MW']])

# 5.4 Pembagian Data (Time-Series Split)
# Data dibagi berdasarkan urutan waktu untuk mencegah data leakage
print("\n✂️ Membagi Data (Time-Based Split)...")
train_size = int(len(df) * 0.8)
train_data = df.iloc[:train_size]
test_data = df.iloc[train_size:]

print(f"📊 Data Training: {len(train_data)} sampel")
print(f"📊 Data Testing: {len(test_data)} sampel")

# =============================================================================
# BAGIAN 6: PREPARASI DATA UNTUK LSTM
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 4: PREPARASI DATA UNTUK LSTM")
print("=" * 60)

def create_sequences(data, look_back=24):
    """
    Fungsi untuk mengubah data deret waktu menjadi format supervised learning.
    look_back: Jumlah langkah waktu sebelumnya yang digunakan untuk memprediksi langkah berikutnya.
    """
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i-look_back:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

look_back = 24  # Menggunakan 24 jam sebelumnya untuk prediksi 1 jam ke depan
X_train, y_train = create_sequences(train_data['Generation_Scaled'].values.reshape(-1, 1), look_back)
X_test, y_test = create_sequences(test_data['Generation_Scaled'].values.reshape(-1, 1), look_back)

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

print(f"📊 Shape X_train: {X_train.shape}")
print(f"📊 Shape y_train: {y_train.shape}")

# =============================================================================
# BAGIAN 7: MEMBANGUN MODEL LSTM (MITIGASI OVERFITTING & UNDERFITTING)
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 5: MEMBANGUN MODEL LSTM")
print("=" * 60)

# 7.1 Arsitektur Model
# Menggunakan Dropout untuk mencegah overfitting dan LSTM untuk menangkap ketergantungan jangka panjang
model = Sequential([
    LSTM(50, activation='relu', input_shape=(look_back, 1), return_sequences=True, kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.2),  # Regularisasi untuk mencegah overfitting
    LSTM(50, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.2),  # Regularisasi tambahan
    Dense(25, activation='relu'),
    Dense(1)
])

# 7.2 Kompilasi Model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.summary()

# 7.3 Callback untuk Optimasi Pelatihan
# EarlyStopping: Menghentikan pelatihan jika validasi loss tidak membaik
# ReduceLROnPlateau: Menurunkan learning rate jika loss stagnan
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001)

# 7.4 Training Model
print("\n🚀 Melatih Model...")
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# =============================================================================
# BAGIAN 8: EVALUASI MODEL
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 6: EVALUASI MODEL")
print("=" * 60)

# 8.1 Prediksi
y_pred_scaled = model.predict(X_test)

# 8.2 Inverse Scaling
# Mengembalikan skala prediksi ke nilai asli (MW)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_actual = scaler.inverse_transform(y_pred_scaled)

# 8.3 Metrik Evaluasi
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))
mae = mean_absolute_error(y_test_actual, y_pred_actual)

print(f"\n📊 METRIK EVALUASI:")
print(f"   RMSE: {rmse:.2f} MW")
print(f"   MAE: {mae:.2f} MW")

# 8.4 Tabel Perbandingan
evaluation_table = pd.DataFrame({
    'Metrik': ['RMSE', 'MAE'],
    'Nilai': [rmse, mae],
    'Satuan': ['MW', 'MW']
})
print("\n📋 Tabel Evaluasi Model:")
print(evaluation_table.to_string(index=False))

# =============================================================================
# BAGIAN 9: VISUALISASI HASIL
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 7: VISUALISASI HASIL")
print("=" * 60)

# 9.1 Grafik Prediksi vs Aktual
plt.figure(figsize=(14, 6))
plt.plot(y_test_actual[:100], label='Aktual', linewidth=2)
plt.plot(y_pred_actual[:100], label='Prediksi', linewidth=2, linestyle='--')
plt.title('Perbandingan Prediksi vs Aktual (100 Jam Terakhir)')
plt.xlabel('Waktu (Jam)')
plt.ylabel('Beban (MW)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 9.2 Grafik Error
error = y_test_actual - y_pred_actual
plt.figure(figsize=(14, 6))
plt.scatter(range(len(error)), error, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Distribusi Error Prediksi')
plt.xlabel('Waktu (Jam)')
plt.ylabel('Error (MW)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 9.3 Grafik Training History
plt.figure(figsize=(14, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# =============================================================================
# BAGIAN 10: DEPLOYMENT PENGUJIAN SEDERHANA (REVISI - FIXED)
# =============================================================================
print("\n" + "=" * 60)
print("LANGKAH 8: DEPLOYMENT PENGUJIAN SEDERHANA")
print("=" * 60)

# Fungsi untuk melakukan prediksi pada data baru
def predict_future_load(model, scaler, last_sequence, look_back=24, steps=24):
    """
    Fungsi untuk memprediksi beban listrik untuk beberapa langkah ke depan.
    last_sequence: Data historis terakhir yang digunakan sebagai input.
    steps: Jumlah langkah waktu yang ingin diprediksi.
    """
    try:
        predictions = []
        current_sequence = last_sequence.copy()

        for _ in range(steps):
            # Reshape untuk input model
            input_data = current_sequence.reshape(1, look_back, 1)

            # Prediksi
            pred = model.predict(input_data, verbose=0)
            predictions.append(pred[0][0])

            # Update sequence untuk langkah berikutnya
            current_sequence = np.append(current_sequence[1:], pred[0][0])

        # Inverse scaling
        predictions = np.array(predictions).reshape(-1, 1)
        predictions = scaler.inverse_transform(predictions)

        return predictions.flatten()

    except Exception as e:
        print(f"❌ Error pada prediksi: {e}")
        return None

# Skenario Pengujian 1: Prediksi 24 Jam ke Depan (Hari Kerja)
print("\n🧪 Skenario 1: Prediksi 24 Jam ke Depan (Hari Kerja)")
try:
    last_sequence_train = train_data['Generation_Scaled'].values[-look_back:].reshape(-1, 1)
    future_pred_1 = predict_future_load(model, scaler, last_sequence_train, look_back=24, steps=24)
    if future_pred_1 is not None:
        print(f"✅ Prediksi 24 jam ke depan (dari data training):\n{future_pred_1}")
    else:
        print("❌ Gagal memprediksi beban masa depan.")
except Exception as e:
    print(f"❌ Terjadi kesalahan pada Skenario 1: {e}")
```
