```python
# =========================================================
# PREDIKSI KANKER PAYUDARA MENGGUNAKAN DATASET BREAST CANCER WISCONSIN (ORIGINAL)
# Verifikasi Metrik, Penanganan Kesalahan Dinamis, dan Tuning Threshold untuk Recall dan Precision
# =========================================================
# Deskripsi Singkat:
# Program ini bertujuan untuk membangun dan mengevaluasi model Jaringan Saraf Tiruan (JST) dalam mengklasifikasikan tumor payudara
# sebagai benign (jinak) atau malignant (ganas) dengan menggunakan dataset Breast Cancer Wisconsin (Original).
# Dalam pembahasan ini, kita memeriksa dan memverifikasi metrik yang digunakan dalam model, serta menangani kasus error dengan
# lebih detail. Selain itu, dilakukan penyesuaian threshold untuk **precision** dan **recall**, yang sangat penting dalam aplikasi medis.

# =========================================================
# 1) INSTALL DAN IMPORT LIBRARY
# =========================================================
!pip -q install ucimlrepo tensorflow scikit-learn seaborn matplotlib

# Import library yang diperlukan
import warnings, random, os
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Reproducibility untuk hasil yang konsisten
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)

# =========================================================
# 2) MEMUAT DATASET
# =========================================================
dataset = fetch_ucirepo(id=15)

X = dataset.data.features.copy()
y = dataset.data.targets.copy().squeeze()

print("Metadata dataset:")
print(dataset.metadata)

print("\nInformasi variabel:")
print(dataset.variables)

# Mengubah nama kolom fitur agar lebih bersih
X.columns = [str(c).strip() for c in X.columns]

# Memastikan semua fitur adalah numerik
for col in X.columns:
    X[col] = pd.to_numeric(X[col], errors="coerce")

# Menetapkan label target: benign = 0, malignant = 1
y = pd.to_numeric(y, errors="coerce")
y = y.map({2: 0, 4: 1})

print("\nUkuran fitur :", X.shape)
print("Ukuran target:", y.shape)

# =========================================================
# 3) EDA (EXPLORATORY DATA ANALYSIS)
# =========================================================
# Analisis Missing Values dan Distribusi Kelas
missing_df = X.isna().sum().sort_values(ascending=False).reset_index()
missing_df.columns = ["Fitur", "Jumlah Missing"]

class_df = y.value_counts().sort_index().rename(index={0: "Benign (0)", 1: "Malignant (1)"}).reset_index()
class_df.columns = ["Kelas", "Jumlah"]

# Menampilkan analisis missing values dan distribusi kelas
display(missing_df)
display(class_df)

# Visualisasi distribusi missing values dan distribusi kelas
plt.figure(figsize=(14, 5))

# Visualisasi missing value per fitur
plt.subplot(1, 2, 1)
sns.barplot(data=missing_df, x="Fitur", y="Jumlah Missing")
plt.title("Jumlah Missing Value per Fitur")
plt.xticks(rotation=45)

# Visualisasi distribusi kelas
plt.subplot(1, 2, 2)
sns.barplot(data=class_df, x="Kelas", y="Jumlah")
plt.title("Distribusi Kelas")
for i, v in enumerate(class_df["Jumlah"]):
    plt.text(i, v + 3, str(v), ha="center")
plt.tight_layout()
plt.show()

# =========================================================
# 4) TRAIN/VALIDATION/TEST SPLIT (STRATIFIED)
# =========================================================
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y,
    test_size=0.15,
    stratify=y,
    random_state=SEED
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full,
    test_size=0.1765,  # Hasil akhir kira-kira 70% train, 15% validation, 15% test
    stratify=y_train_full,
    random_state=SEED
)

# Menampilkan distribusi kelas pada setiap subset
def dist_table(name, target):
    vc = pd.Series(target).value_counts().sort_index()
    return pd.DataFrame({
        "Subset": [name, name],
        "Kelas": ["Benign (0)", "Malignant (1)"],
        "Jumlah": [vc.get(0, 0), vc.get(1, 0)]
    })

split_dist = pd.concat([
    dist_table("Train", y_train),
    dist_table("Validation", y_val),
    dist_table("Test", y_test)
], ignore_index=True)

print("\nDistribusi kelas pada setiap subset:")
display(split_dist)

plt.figure(figsize=(8, 5))
sns.barplot(data=split_dist, x="Subset", y="Jumlah", hue="Kelas")
plt.title("Distribusi Kelas per Subset")
plt.show()

# =========================================================
# 5) IMPUTASI DAN SCALING
# =========================================================
# Mengimputasi missing values menggunakan median
imputer = SimpleImputer(strategy="median")

X_train_imp = imputer.fit_transform(X_train)
X_val_imp   = imputer.transform(X_val)
X_test_imp  = imputer.transform(X_test)

# Menstandarisasi fitur
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_imp)
X_val_scaled   = scaler.transform(X_val_imp)
X_test_scaled  = scaler.transform(X_test_imp)

print("Shape X_train_scaled:", X_train_scaled.shape)
print("Shape X_val_scaled  :", X_val_scaled.shape)
print("Shape X_test_scaled :", X_test_scaled.shape)

# =========================================================
# 6) CLASS WEIGHT (UNTUK MENANGANI IMBALANSE KELAS)
# =========================================================
# Menghitung class weights agar kelas minoritas (malignant) lebih diprioritaskan
classes = np.unique(y_train)
weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)
class_weight_dict = {int(c): float(w) for c, w in zip(classes, weights)}

print("\nClass weight yang dipakai:")
print(class_weight_dict)

# =========================================================
# 7) MEMBANGUN MODEL MLP (MULTILAYER PERCEPTRON)
# =========================================================
# Model MLP sederhana dengan regularisasi untuk menghindari overfitting
model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),

    Dense(24, activation="relu", kernel_regularizer=l2(1e-3)),
    BatchNormalization(),
    Dropout(0.25),

    Dense(12, activation="relu", kernel_regularizer=l2(1e-3)),
    BatchNormalization(),
    Dropout(0.20),

    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name="accuracy"),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.AUC(name="auc")
    ]
)

# Menampilkan summary model
model.summary()

# =========================================================
# 8) CALLBACKS (EarlyStopping dan ReduceLROnPlateau)
# =========================================================
callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )
]

# =========================================================
# 9) TRAINING MODEL
# =========================================================
history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=200,
    batch_size=32,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# =========================================================
# 10) EVALUASI MODEL PADA DATA UJI
# =========================================================
# Evaluasi performa model pada data uji (test set)
test_eval = model.evaluate(X_test_scaled, y_test, verbose=0)
metric_names = model.metrics_names

# Mengatasi ketidaksesuaian panjang antara 'metric_names' dan 'test_eval'
# Memastikan kedua array memiliki panjang yang sama
if len(metric_names) == len(test_eval):
    eval_df = pd.DataFrame({
        "Metrik": metric_names,
        "Nilai": test_eval
    })
    print("\nHasil evaluasi model pada test set:")
    display(eval_df)
else:
    print(f"Error: Panjang 'metric_names' ({len(metric_names)}) dan 'test_eval' ({len(test_eval)}) tidak cocok.")
    print("Cek apakah jumlah metrik pada model sudah sesuai dengan hasil evaluasi.")

# Evaluasi lebih lanjut menggunakan hasil prediksi
y_test_prob = model.predict(X_test_scaled, verbose=0).ravel()
y_test_pred = (y_test_prob >= 0.50).astype(int)  # Threshold standar 0.5

# Menghitung metrik evaluasi
acc  = accuracy_score(y_test, y_test_pred)
prec = precision_score(y_test, y_test_pred)
rec  = recall_score(y_test, y_test_pred)
f1   = f1_score(y_test, y_test_pred)
auc  = roc_auc_score(y_test, y_test_prob)

# Menampilkan ringkasan metrik utama
summary_metrics = pd.DataFrame({
    "Metrik": [
        "Accuracy", "Precision", "Recall (Sensitivity)", "F1-Score", "ROC-AUC"
    ],
    "Nilai": [acc, prec, rec, f1, auc]
})

print("\nRingkasan metrik utama:")
display(summary_metrics)

# =========================================================
# 11) TUNING THRESHOLD UNTUK MENDAPATKAN TRADE-OFF ANTARA PRECISION DAN RECALL
# =========================================================
# Mencari threshold terbaik untuk mencapai trade-off antara precision dan recall
thresholds = np.arange(0.30, 0.71, 0.01)
f1_scores = [f1_score(y_test, (y_test_prob >= t).astype(int)) for t in thresholds]

# Menampilkan threshold dengan F1-score terbaik
best_threshold = thresholds[np.argmax(f1_scores)]
print(f"Threshold terbaik: {best_threshold:.2f}")

# Visualisasi trade-off antara threshold dan F1-score
plt.figure(figsize=(8, 5))
plt.plot(thresholds, f1_scores)
plt.axvline(best_threshold, linestyle="--", color="red", label=f"Best Threshold: {best_threshold:.2f}")
plt.title("Threshold Tuning untuk Trade-off Precision dan Recall")
plt.xlabel("Threshold")
plt.ylabel("F1-Score")
plt.legend(loc="best")
plt.show()
```
