```python
# ===============================================
# PROGRAM: Inisialisasi Bobot Awal untuk Prediksi Kanker Payudara
# DESKRIPSI: Program ini memuat dataset Wisconsin
# (Original) dari *UCI Machine Learning Repository*,
# melakukan *preprocessing*, dan menginisialisasi
# *matriks bobot* dan *bias* awal secara manual untuk
# jaringan syaraf sederhana.
# ===============================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ---------- 1. Pemuatan Dataset (UCI) ----------
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
kolom = [
    "id", "clump_thickness", "uniformity_cell_size", "uniformity_cell_shape",
    "marginal_adhesion", "single_epithelial_size", "bare_nuclei",
    "bland_chromatin", "normal_nucleoli", "mitoses", "class"
]
df = pd.read_csv(url, names=kolom)

# ---------- 2. Pembersihan Data ----------
# Hapus baris dengan nilai '?' dalam kolom 'bare_nuclei'
df = df[df["bare_nuclei"] != "?"]
df["bare_nuclei"] = df["bare_nuclei"].astype(int)

# ---------- 3. Menentukan Fitur (X) & Target (y) ----------
X = df.iloc[:, 1:-1].values  # 9 fitur numerik
y = df["class"].replace({2:0,4:1}).values  # 0 = benign, 1 = malignant

# ---------- 4. Standarisasi Fitur ----------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------- 5. Membagi Data Train/Test ----------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# ---------- 6. Inisialisasi Bobot & Bias Awal ----------
np.random.seed(42)  # untuk reprodusibilitas
input_dim = X_train.shape[1]   # jumlah fitur input = 9
hidden_dim = 6                 # neuron lapisan tersembunyi
output_dim = 1                 # output tunggal (klasifikasi)

# Bobot menghimpun pengaruh input ke neuron tersembunyi
W1 = np.random.randn(input_dim, hidden_dim) * 0.01
b1 = np.zeros((1, hidden_dim))

# Bobot menghimpun pengaruh lapisan tersembunyi ke output
W2 = np.random.randn(hidden_dim, output_dim) * 0.01
b2 = np.zeros((1, output_dim))

# ---------- 7. Tampilkan Bobot &
print("Bobot awal lapisan tersembunyi (W1):")
print(W1)
print("\nBias awal lapisan tersembunyi (b1):")
print(b1)
print("\nBobot awal lapisan output (W2):")
print(W2)
print("\nBias awal lapisan output (b2):")
print(b2)
```
