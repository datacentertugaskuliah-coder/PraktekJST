```python
# ==========================================================
# JUDUL PROGRAM
# Klasifikasi Tingkat Aktivitas Review pada Dataset
# Amazon Product and Google Locations Reviews
# dengan Strategi Pemuatan Data Bertingkat (Fallback Loading)
# ==========================================================
#
# DESKRIPSI SINGKAT
# Program ini memuat dataset resmi UCI "Amazon Product and Google Locations Reviews"
# secara robust. Karena pada beberapa runtime fungsi fetch_ucirepo(id=1276)
# dapat gagal walaupun halaman dataset tersedia di UCI, maka digunakan
# strategi pemuatan bertingkat:
#
# 1. Mencoba memuat dataset melalui ucimlrepo
# 2. Jika gagal, mengunduh file ZIP resmi dari UCI
# 3. Membaca file CSV di dalam ZIP secara manual
#
# Pendekatan ini menjaga integritas analisis karena:
# - sumber data tetap berasal dari UCI
# - error akses library tidak mengganggu alur penelitian
# - struktur data divalidasi sebelum diproses lebih lanjut
# ==========================================================

# ==========================================================
# 1. INSTALASI DAN IMPORT PUSTAKA
# ==========================================================
!pip -q install ucimlrepo joblib

import warnings
warnings.filterwarnings("ignore")

import io
import os
import zipfile
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score, f1_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)

import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("="*80)
print("PROGRAM ANALISIS KLASIFIKASI TINGKAT AKTIVITAS REVIEW")
print("="*80)

# ==========================================================
# 2. FUNGSI PEMUATAN DATASET YANG ROBUST
# ==========================================================
def load_dataset_uci_1276():
    """
    Memuat dataset UCI id=1276 dengan dua tahap:
    1) ucimlrepo
    2) fallback ke unduhan ZIP resmi UCI jika tahap 1 gagal
    """
    sumber_data = None

    # Tahap 1: mencoba jalur impor melalui ucimlrepo
    try:
        print("\n[2A] Mencoba memuat dataset melalui ucimlrepo...")
        dataset = fetch_ucirepo(id=1276)
        X_raw = dataset.data.features.copy()
        y_raw = dataset.data.targets.copy()

        df = X_raw.copy()
        # Antisipasi bentuk target bisa series atau dataframe
        if isinstance(y_raw, pd.DataFrame):
            df["y"] = y_raw.iloc[:, 0]
        else:
            df["y"] = y_raw

        sumber_data = "ucimlrepo"
        print("Pemuatan berhasil melalui ucimlrepo.")
        return df, sumber_data

    except Exception as e:
        print("Pemuatan melalui ucimlrepo gagal.")
        print("Alasan error:", str(e))

    # Tahap 2: fallback ke file ZIP resmi dari UCI
    try:
        print("\n[2B] Mencoba fallback dengan mengunduh file ZIP resmi dari UCI...")
        zip_url = "https://archive.ics.uci.edu/static/public/1276/amazon%2Bproduct%2Band%2Bgoogle%2Blocations%2Breviews.zip"
        response = requests.get(zip_url, timeout=120)
        response.raise_for_status()

        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            # Menampilkan isi ZIP agar transparan
            print("Isi file ZIP:", z.namelist())

            # Mencari file CSV utama
            csv_candidates = [name for name in z.namelist() if name.lower().endswith(".csv")]
            if len(csv_candidates) == 0:
                raise FileNotFoundError("Tidak ditemukan file CSV di dalam ZIP.")

            # Mengambil kandidat pertama; pada halaman UCI file utamanya adalah y_amazon-google-large.csv
            csv_name = csv_candidates[0]
            print("File CSV yang digunakan:", csv_name)

            with z.open(csv_name) as f:
                df = pd.read_csv(f)

        sumber_data = "zip_resmi_uci"
        print("Pemuatan berhasil melalui file ZIP resmi UCI.")
        return df, sumber_data

    except Exception as e:
        raise RuntimeError(
            "Dataset gagal dimuat baik melalui ucimlrepo maupun melalui ZIP resmi UCI. "
            f"Detail error terakhir: {str(e)}"
        )

# ==========================================================
# 3. MEMUAT DATASET
# ==========================================================
print("\n[3] Memuat dataset resmi dari UCI dengan strategi fallback...")
df, sumber_data = load_dataset_uci_1276()

print("\nSumber pemuatan data:", sumber_data)
print("Ukuran data awal:", df.shape)
print("\nLima baris pertama dataset:")
display(df.head())

print("\nNama kolom dataset:")
print(df.columns.tolist())

print("\nInformasi struktur data:")
print(df.info())

print("\nJumlah missing value per kolom:")
display(df.isnull().sum())

# ==========================================================
# 4. VALIDASI STRUKTUR DATA
# ==========================================================
print("\n[4] Melakukan validasi struktur kolom...")
kolom_wajib = ["ds", "unique_id", "y"]
for kolom in kolom_wajib:
    if kolom not in df.columns:
        raise ValueError(f"Kolom wajib '{kolom}' tidak ditemukan pada dataset hasil pemuatan.")
print("Validasi berhasil. Dataset memiliki kolom ds, unique_id, dan y.")

# ==========================================================
# 5. PRA-PEMROSESAN DATA
# ==========================================================
print("\n[5] Melakukan pra-pemrosesan data...")
df["ds"] = pd.to_datetime(df["ds"], errors="coerce")
df["y"] = pd.to_numeric(df["y"], errors="coerce")
df = df.dropna(subset=["ds", "unique_id", "y"]).copy()
df = df.sort_values(["unique_id", "ds"]).reset_index(drop=True)

print("Ukuran data setelah cleaning:", df.shape)

print("\nStatistik deskriptif target numerik (y):")
display(df["y"].describe())

# ==========================================================
# 6. VISUALISASI AWAL DATA
# ==========================================================
print("\n[6] Menampilkan visualisasi awal data...")

df_agg = df.groupby("ds", as_index=False)["y"].sum()

plt.figure(figsize=(14, 5))
plt.plot(df_agg["ds"], df_agg["y"])
plt.title("Tren Agregat Jumlah Review per Waktu")
plt.xlabel("Waktu")
plt.ylabel("Jumlah Review per Jam")
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df["y"], bins=50)
plt.title("Distribusi Jumlah Review per Jam")
plt.xlabel("Jumlah Review")
plt.ylabel("Frekuensi")
plt.grid(True, alpha=0.3)
plt.show()

# ==========================================================
# 7. FEATURE ENGINEERING
# ==========================================================
print("\n[7] Melakukan feature engineering...")

df["hour"] = df["ds"].dt.hour
df["dayofweek"] = df["ds"].dt.dayofweek
df["day"] = df["ds"].dt.day
df["month"] = df["ds"].dt.month
df["quarter"] = df["ds"].dt.quarter
df["is_weekend"] = df["dayofweek"].isin([5, 6]).astype(int)

df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

df["lag_1"] = df.groupby("unique_id")["y"].shift(1)
df["lag_2"] = df.groupby("unique_id")["y"].shift(2)
df["lag_3"] = df.groupby("unique_id")["y"].shift(3)
df["lag_24"] = df.groupby("unique_id")["y"].shift(24)

df["rolling_mean_3"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=3).mean().reset_index(level=0, drop=True)
)
df["rolling_mean_24"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=24).mean().reset_index(level=0, drop=True)
)
df["rolling_std_24"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=24).std().reset_index(level=0, drop=True)
)

uid_encoder = LabelEncoder()
df["unique_id_encoded"] = uid_encoder.fit_transform(df["unique_id"].astype(str))

# ==========================================================
# 8. PEMBENTUKAN LABEL KLASIFIKASI
# ==========================================================
print("\n[8] Membentuk label klasifikasi aktivitas review...")

q1 = df["y"].quantile(0.33)
q2 = df["y"].quantile(0.66)

def label_aktivitas(nilai):
    if nilai <= q1:
        return "rendah"
    elif nilai <= q2:
        return "sedang"
    else:
        return "tinggi"

df["aktivitas_review"] = df["y"].apply(label_aktivitas)

print(f"Batas kuantil: q1 = {q1:.4f}, q2 = {q2:.4f}")
display(df["aktivitas_review"].value_counts())

plt.figure(figsize=(7, 4))
df["aktivitas_review"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribusi Kelas Aktivitas Review")
plt.xlabel("Kelas")
plt.ylabel("Jumlah Data")
plt.grid(axis="y", alpha=0.3)
plt.show()

# ==========================================================
# 9. MENYIAPKAN DATA MODEL
# ==========================================================
print("\n[9] Menyiapkan data untuk pemodelan...")
df_model = df.dropna().copy()

fitur = [
    "unique_id_encoded",
    "hour", "dayofweek", "day", "month", "quarter", "is_weekend",
    "hour_sin", "hour_cos", "dow_sin", "dow_cos",
    "lag_1", "lag_2", "lag_3", "lag_24",
    "rolling_mean_3", "rolling_mean_24", "rolling_std_24"
]

X = df_model[fitur].copy()
y = df_model["aktivitas_review"].copy()

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# ==========================================================
# 10. SPLIT DATA BERBASIS WAKTU
# ==========================================================
print("\n[10] Membagi data secara kronologis...")
n = len(df_model)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

X_train = X.iloc[:train_end]
y_train = y_encoded[:train_end]
X_val = X.iloc[train_end:val_end]
y_val = y_encoded[train_end:val_end]
X_test = X.iloc[val_end:]
y_test = y_encoded[val_end:]

display(pd.DataFrame({
    "Subset": ["Train", "Validation", "Test"],
    "Jumlah Data": [len(X_train), len(X_val), len(X_test)],
    "Proporsi": [len(X_train)/n, len(X_val)/n, len(X_test)/n]
}))

# ==========================================================
# 11. STANDARDISASI DAN PENANGANAN IMBALANCE
# ==========================================================
print("\n[11] Melakukan standardisasi fitur dan menghitung class weight...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

classes = np.unique(y_train)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
class_weights = {int(c): float(w) for c, w in zip(classes, weights)}
print("Class weights:", class_weights)

# ==========================================================
# 12. MODEL BASELINE
# ==========================================================
print("\n[12] Melatih model baseline Logistic Regression...")
baseline = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=SEED)
baseline.fit(X_train_scaled, y_train)
y_pred_baseline = baseline.predict(X_test_scaled)

baseline_acc = accuracy_score(y_test, y_pred_baseline)
baseline_f1_macro = f1_score(y_test, y_pred_baseline, average="macro")
baseline_f1_weighted = f1_score(y_test, y_pred_baseline, average="weighted")

print("Accuracy baseline   :", round(baseline_acc, 4))
print("F1 macro baseline   :", round(baseline_f1_macro, 4))
print("F1 weighted baseline:", round(baseline_f1_weighted, 4))

# ==========================================================
# 13. MODEL MLP
# ==========================================================
print("\n[13] Melatih model MLP...")

def build_mlp(input_dim, n_classes):
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation="relu", kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        Dropout(0.30),

        Dense(64, activation="relu", kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        Dropout(0.25),

        Dense(32, activation="relu", kernel_regularizer=l2(1e-4)),
        Dropout(0.20),

        Dense(n_classes, activation="softmax")
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

mlp = build_mlp(X_train_scaled.shape[1], len(label_encoder.classes_))

early_stopping = EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6, verbose=1)

history = mlp.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=60,
    batch_size=256,
    class_weight=class_weights,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

history_df = pd.DataFrame(history.history)

plt.figure(figsize=(10, 4))
plt.plot(history_df["loss"], label="Train Loss")
plt.plot(history_df["val_loss"], label="Validation Loss")
plt.title("Kurva Loss Selama Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history_df["accuracy"], label="Train Accuracy")
plt.plot(history_df["val_accuracy"], label="Validation Accuracy")
plt.title("Kurva Accuracy Selama Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ==========================================================
# 14. EVALUASI MODEL
# ==========================================================
print("\n[14] Mengevaluasi model...")
y_prob_mlp = mlp.predict(X_test_scaled, verbose=0)
y_pred_mlp = np.argmax(y_prob_mlp, axis=1)

mlp_acc = accuracy_score(y_test, y_pred_mlp)
mlp_f1_macro = f1_score(y_test, y_pred_mlp, average="macro")
mlp_f1_weighted = f1_score(y_test, y_pred_mlp, average="weighted")

print("Accuracy MLP   :", round(mlp_acc, 4))
print("F1 macro MLP   :", round(mlp_f1_macro, 4))
print("F1 weighted MLP:", round(mlp_f1_weighted, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_mlp, target_names=label_encoder.classes_))

cm = confusion_matrix(y_test, y_pred_mlp)
fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_).plot(
    ax=ax, cmap="Blues", values_format="d"
)
plt.title("Confusion Matrix - Model MLP")
plt.show()

display(pd.DataFrame({
    "Model": ["Logistic Regression (Baseline)", "MLP"],
    "Accuracy": [baseline_acc, mlp_acc],
    "F1 Macro": [baseline_f1_macro, mlp_f1_macro],
    "F1 Weighted": [baseline_f1_weighted, mlp_f1_weighted]
}).style.format({
    "Accuracy": "{:.4f}",
    "F1 Macro": "{:.4f}",
    "F1 Weighted": "{:.4f}"
}))

# ==========================================================
# 15. INTERPRETASI SINGKAT
# ==========================================================
print("\n[15] Interpretasi singkat")
print("="*80)
print(f"Sumber data yang berhasil digunakan: {sumber_data}")
print("Jika sumber_data = 'zip_resmi_uci', berarti fallback berhasil mengatasi")
print("kegagalan impor melalui ucimlrepo tanpa mengubah sumber dataset.")
if mlp_f1_macro > baseline_f1_macro:
    print("MLP menunjukkan kinerja yang lebih baik daripada baseline pada F1 Macro.")
else:
    print("MLP belum melampaui baseline secara jelas pada F1 Macro.")
print("Kesimpulan harus tetap dibatasi pada klasifikasi tingkat aktivitas review,")
print("bukan isi atau sentimen teks review.")
print("="*80)
```

```python
# ==========================================================
# JUDUL PROGRAM
# Klasifikasi Tingkat Aktivitas Review pada Dataset
# Amazon Product and Google Locations Reviews
# dengan Strategi Pemuatan Data Bertingkat (Fallback Loading)
# ==========================================================
#
# DESKRIPSI SINGKAT
# Program ini memuat dataset resmi UCI "Amazon Product and Google Locations Reviews"
# secara robust. Karena pada beberapa runtime fungsi fetch_ucirepo(id=1276)
# dapat gagal walaupun halaman dataset tersedia di UCI, maka digunakan
# strategi pemuatan bertingkat:
#
# 1. Mencoba memuat dataset melalui ucimlrepo
# 2. Jika gagal, mengunduh file ZIP resmi dari UCI
# 3. Membaca file CSV di dalam ZIP secara manual
#
# Pendekatan ini menjaga integritas analisis karena:
# - sumber data tetap berasal dari UCI
# - error akses library tidak mengganggu alur penelitian
# - struktur data divalidasi sebelum diproses lebih lanjut
# ==========================================================

# ==========================================================
# 1. INSTALASI DAN IMPORT PUSTAKA
# ==========================================================
!pip -q install ucimlrepo joblib

import warnings
warnings.filterwarnings("ignore")

import io
import os
import zipfile
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score, f1_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)

import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("="*80)
print("PROGRAM ANALISIS KLASIFIKASI TINGKAT AKTIVITAS REVIEW")
print("="*80)

# ==========================================================
# 2. FUNGSI PEMUATAN DATASET YANG ROBUST
# ==========================================================
def load_dataset_uci_1276():
    """
    Memuat dataset UCI id=1276 dengan dua tahap:
    1) ucimlrepo
    2) fallback ke unduhan ZIP resmi UCI jika tahap 1 gagal
    """
    sumber_data = None

    # Tahap 1: mencoba jalur impor melalui ucimlrepo
    try:
        print("\n[2A] Mencoba memuat dataset melalui ucimlrepo...")
        dataset = fetch_ucirepo(id=1276)
        X_raw = dataset.data.features.copy()
        y_raw = dataset.data.targets.copy()

        df = X_raw.copy()
        # Antisipasi bentuk target bisa series atau dataframe
        if isinstance(y_raw, pd.DataFrame):
            df["y"] = y_raw.iloc[:, 0]
        else:
            df["y"] = y_raw

        sumber_data = "ucimlrepo"
        print("Pemuatan berhasil melalui ucimlrepo.")
        return df, sumber_data

    except Exception as e:
        print("Pemuatan melalui ucimlrepo gagal.")
        print("Alasan error:", str(e))

    # Tahap 2: fallback ke file ZIP resmi dari UCI
    try:
        print("\n[2B] Mencoba fallback dengan mengunduh file ZIP resmi dari UCI...")
        zip_url = "https://archive.ics.uci.edu/static/public/1276/amazon%2Bproduct%2Band%2Bgoogle%2Blocations%2Breviews.zip"
        response = requests.get(zip_url, timeout=120)
        response.raise_for_status()

        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            # Menampilkan isi ZIP agar transparan
            print("Isi file ZIP:", z.namelist())

            # Mencari file CSV utama
            csv_candidates = [name for name in z.namelist() if name.lower().endswith(".csv")]
            if len(csv_candidates) == 0:
                raise FileNotFoundError("Tidak ditemukan file CSV di dalam ZIP.")

            # Mengambil kandidat pertama; pada halaman UCI file utamanya adalah y_amazon-google-large.csv
            csv_name = csv_candidates[0]
            print("File CSV yang digunakan:", csv_name)

            with z.open(csv_name) as f:
                df = pd.read_csv(f)

        sumber_data = "zip_resmi_uci"
        print("Pemuatan berhasil melalui file ZIP resmi UCI.")
        return df, sumber_data

    except Exception as e:
        raise RuntimeError(
            "Dataset gagal dimuat baik melalui ucimlrepo maupun melalui ZIP resmi UCI. "
            f"Detail error terakhir: {str(e)}"
        )

# ==========================================================
# 3. MEMUAT DATASET
# ==========================================================
print("\n[3] Memuat dataset resmi dari UCI dengan strategi fallback...")
df, sumber_data = load_dataset_uci_1276()

print("\nSumber pemuatan data:", sumber_data)
print("Ukuran data awal:", df.shape)
print("\nLima baris pertama dataset:")
display(df.head())

print("\nNama kolom dataset:")
print(df.columns.tolist())

print("\nInformasi struktur data:")
print(df.info())

print("\nJumlah missing value per kolom:")
display(df.isnull().sum())

# ==========================================================
# 4. VALIDASI STRUKTUR DATA
# ==========================================================
print("\n[4] Melakukan validasi struktur kolom...")
kolom_wajib = ["ds", "unique_id", "y"]
for kolom in kolom_wajib:
    if kolom not in df.columns:
        raise ValueError(f"Kolom wajib '{kolom}' tidak ditemukan pada dataset hasil pemuatan.")
print("Validasi berhasil. Dataset memiliki kolom ds, unique_id, dan y.")

# ==========================================================
# 5. PRA-PEMROSESAN DATA
# ==========================================================
print("\n[5] Melakukan pra-pemrosesan data...")
df["ds"] = pd.to_datetime(df["ds"], errors="coerce")
df["y"] = pd.to_numeric(df["y"], errors="coerce")
df = df.dropna(subset=["ds", "unique_id", "y"]).copy()
df = df.sort_values(["unique_id", "ds"]).reset_index(drop=True)

print("Ukuran data setelah cleaning:", df.shape)

print("\nStatistik deskriptif target numerik (y):")
display(df["y"].describe())

# ==========================================================
# 6. VISUALISASI AWAL DATA
# ==========================================================
print("\n[6] Menampilkan visualisasi awal data...")

df_agg = df.groupby("ds", as_index=False)["y"].sum()

plt.figure(figsize=(14, 5))
plt.plot(df_agg["ds"], df_agg["y"])
plt.title("Tren Agregat Jumlah Review per Waktu")
plt.xlabel("Waktu")
plt.ylabel("Jumlah Review per Jam")
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(df["y"], bins=50)
plt.title("Distribusi Jumlah Review per Jam")
plt.xlabel("Jumlah Review")
plt.ylabel("Frekuensi")
plt.grid(True, alpha=0.3)
plt.show()

# ==========================================================
# 7. FEATURE ENGINEERING
# ==========================================================
print("\n[7] Melakukan feature engineering...")

df["hour"] = df["ds"].dt.hour
df["dayofweek"] = df["ds"].dt.dayofweek
df["day"] = df["ds"].dt.day
df["month"] = df["ds"].dt.month
df["quarter"] = df["ds"].dt.quarter
df["is_weekend"] = df["dayofweek"].isin([5, 6]).astype(int)

df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

df["lag_1"] = df.groupby("unique_id")["y"].shift(1)
df["lag_2"] = df.groupby("unique_id")["y"].shift(2)
df["lag_3"] = df.groupby("unique_id")["y"].shift(3)
df["lag_24"] = df.groupby("unique_id")["y"].shift(24)

df["rolling_mean_3"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=3).mean().reset_index(level=0, drop=True)
)
df["rolling_mean_24"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=24).mean().reset_index(level=0, drop=True)
)
df["rolling_std_24"] = (
    df.groupby("unique_id")["y"].shift(1).rolling(window=24).std().reset_index(level=0, drop=True)
)

uid_encoder = LabelEncoder()
df["unique_id_encoded"] = uid_encoder.fit_transform(df["unique_id"].astype(str))

# ==========================================================
# 8. PEMBENTUKAN LABEL KLASIFIKASI
# ==========================================================
print("\n[8] Membentuk label klasifikasi aktivitas review...")

q1 = df["y"].quantile(0.33)
q2 = df["y"].quantile(0.66)

def label_aktivitas(nilai):
    if nilai <= q1:
        return "rendah"
    elif nilai <= q2:
        return "sedang"
    else:
        return "tinggi"

df["aktivitas_review"] = df["y"].apply(label_aktivitas)

print(f"Batas kuantil: q1 = {q1:.4f}, q2 = {q2:.4f}")
display(df["aktivitas_review"].value_counts())

plt.figure(figsize=(7, 4))
df["aktivitas_review"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribusi Kelas Aktivitas Review")
plt.xlabel("Kelas")
plt.ylabel("Jumlah Data")
plt.grid(axis="y", alpha=0.3)
plt.show()

# ==========================================================
# 9. MENYIAPKAN DATA MODEL
# ==========================================================
print("\n[9] Menyiapkan data untuk pemodelan...")
df_model = df.dropna().copy()

fitur = [
    "unique_id_encoded",
    "hour", "dayofweek", "day", "month", "quarter", "is_weekend",
    "hour_sin", "hour_cos", "dow_sin", "dow_cos",
    "lag_1", "lag_2", "lag_3", "lag_24",
    "rolling_mean_3", "rolling_mean_24", "rolling_std_24"
]

X = df_model[fitur].copy()
y = df_model["aktivitas_review"].copy()

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# ==========================================================
# 10. SPLIT DATA BERBASIS WAKTU
# ==========================================================
print("\n[10] Membagi data secara kronologis...")
n = len(df_model)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

X_train = X.iloc[:train_end]
y_train = y_encoded[:train_end]
X_val = X.iloc[train_end:val_end]
y_val = y_encoded[train_end:val_end]
X_test = X.iloc[val_end:]
y_test = y_encoded[val_end:]

display(pd.DataFrame({
    "Subset": ["Train", "Validation", "Test"],
    "Jumlah Data": [len(X_train), len(X_val), len(X_test)],
    "Proporsi": [len(X_train)/n, len(X_val)/n, len(X_test)/n]
}))

# ==========================================================
# 11. STANDARDISASI DAN PENANGANAN IMBALANCE
# ==========================================================
print("\n[11] Melakukan standardisasi fitur dan menghitung class weight...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

classes = np.unique(y_train)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
class_weights = {int(c): float(w) for c, w in zip(classes, weights)}
print("Class weights:", class_weights)

# ==========================================================
# 12. MODEL BASELINE
# ==========================================================
print("\n[12] Melatih model baseline Logistic Regression...")
baseline = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=SEED)
baseline.fit(X_train_scaled, y_train)
y_pred_baseline = baseline.predict(X_test_scaled)

baseline_acc = accuracy_score(y_test, y_pred_baseline)
baseline_f1_macro = f1_score(y_test, y_pred_baseline, average="macro")
baseline_f1_weighted = f1_score(y_test, y_pred_baseline, average="weighted")

print("Accuracy baseline   :", round(baseline_acc, 4))
print("F1 macro baseline   :", round(baseline_f1_macro, 4))
print("F1 weighted baseline:", round(baseline_f1_weighted, 4))

# ==========================================================
# 13. MODEL MLP
# ==========================================================
print("\n[13] Melatih model MLP...")

def build_mlp(input_dim, n_classes):
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation="relu", kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        Dropout(0.30),

        Dense(64, activation="relu", kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        Dropout(0.25),

        Dense(32, activation="relu", kernel_regularizer=l2(1e-4)),
        Dropout(0.20),

        Dense(n_classes, activation="softmax")
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

mlp = build_mlp(X_train_scaled.shape[1], len(label_encoder.classes_))

early_stopping = EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6, verbose=1)

history = mlp.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=60,
    batch_size=256,
    class_weight=class_weights,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

history_df = pd.DataFrame(history.history)

plt.figure(figsize=(10, 4))
plt.plot(history_df["loss"], label="Train Loss")
plt.plot(history_df["val_loss"], label="Validation Loss")
plt.title("Kurva Loss Selama Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history_df["accuracy"], label="Train Accuracy")
plt.plot(history_df["val_accuracy"], label="Validation Accuracy")
plt.title("Kurva Accuracy Selama Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ==========================================================
# 14. EVALUASI MODEL
# ==========================================================
print("\n[14] Mengevaluasi model...")
y_prob_mlp = mlp.predict(X_test_scaled, verbose=0)
y_pred_mlp = np.argmax(y_prob_mlp, axis=1)

mlp_acc = accuracy_score(y_test, y_pred_mlp)
mlp_f1_macro = f1_score(y_test, y_pred_mlp, average="macro")
mlp_f1_weighted = f1_score(y_test, y_pred_mlp, average="weighted")

print("Accuracy MLP   :", round(mlp_acc, 4))
print("F1 macro MLP   :", round(mlp_f1_macro, 4))
print("F1 weighted MLP:", round(mlp_f1_weighted, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_mlp, target_names=label_encoder.classes_))

cm = confusion_matrix(y_test, y_pred_mlp)
fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_).plot(
    ax=ax, cmap="Blues", values_format="d"
)
plt.title("Confusion Matrix - Model MLP")
plt.show()

display(pd.DataFrame({
    "Model": ["Logistic Regression (Baseline)", "MLP"],
    "Accuracy": [baseline_acc, mlp_acc],
    "F1 Macro": [baseline_f1_macro, mlp_f1_macro],
    "F1 Weighted": [baseline_f1_weighted, mlp_f1_weighted]
}).style.format({
    "Accuracy": "{:.4f}",
    "F1 Macro": "{:.4f}",
    "F1 Weighted": "{:.4f}"
}))

# ==========================================================
# 15. INTERPRETASI SINGKAT
# ==========================================================
print("\n[15] Interpretasi singkat")
print("="*80)
print(f"Sumber data yang berhasil digunakan: {sumber_data}")
print("Jika sumber_data = 'zip_resmi_uci', berarti fallback berhasil mengatasi")
print("kegagalan impor melalui ucimlrepo tanpa mengubah sumber dataset.")
if mlp_f1_macro > baseline_f1_macro:
    print("MLP menunjukkan kinerja yang lebih baik daripada baseline pada F1 Macro.")
else:
    print("MLP belum melampaui baseline secara jelas pada F1 Macro.")
print("Kesimpulan harus tetap dibatasi pada klasifikasi tingkat aktivitas review,")
print("bukan isi atau sentimen teks review.")
print("="*80)
```
