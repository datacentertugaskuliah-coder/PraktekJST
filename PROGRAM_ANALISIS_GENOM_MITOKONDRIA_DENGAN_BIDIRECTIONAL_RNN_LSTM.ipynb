```python
# ====================================================================
# PROGRAM ANALISIS GENOM MITOKONDRIA DENGAN BIDIRECTIONAL RNN-LSTM
# ====================================================================
# Judul       : Implementasi Bi-LSTM untuk Klasifikasi Wilayah Coding pada
#               Genom Mitokondria Channa micropeltes (KX129904)
# Deskripsi   : Program ini melakukan pengambilan data dari NCBI (3 mode),
#               eksplorasi data (EDA), preprocessing dengan sliding window,
#               augmentasi forward & reverse complement (dengan koreksi koordinat),
#               penyeimbangan SMOTE, pelatihan CNN-BiLSTM (200 epoch, regularisasi),
#               evaluasi metrik (akurasi, precision, recall, F1, AUC) menggunakan sklearn,
#               serta deployment pada 5 sekuens uji.
#               Durasi komputasi dirancang antara 5–10 menit di GPU Google Colab.
# Penulis     : Berdasarkan analisis kritis dan perbaikan logis (bebas error)
# Versi       : Final (1 Cell) – Standar jurnal internasional Q1–Q4
# ====================================================================

# -------------------- 1. PERSIAPAN LINGKUNGAN --------------------
!pip install -q biopython imbalanced-learn joblib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_fscore_support
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, regularizers
from Bio import Entrez, SeqIO
from Bio.Seq import Seq
import joblib
import os
import time
import warnings
warnings.filterwarnings('ignore')

# Reproduksibilitas
np.random.seed(42)
tf.random.set_seed(42)

print("✅ Lingkungan siap | TensorFlow version:", tf.__version__)

# -------------------- 2. FUNGSI UTILITAS  --------------------
def validate_sequence(seq):
    """Memastikan sekuens hanya berisi A, T, C, G dan panjang memadai."""
    if not isinstance(seq, str):
        raise ValueError("Sekuens harus berupa string.")
    seq = seq.upper()
    if not set(seq).issubset('ATCG'):
        raise ValueError(f"Karakter tidak valid: {set(seq) - set('ATCG')}")
    if len(seq) < 100:
        raise ValueError(f"Panjang sekuens ({len(seq)} bp) terlalu pendek.")
    return seq

def one_hot_encode(seq):
    """One-hot encoding (4 dimensi) untuk sekuens DNA."""
    mapping = {'A':[1,0,0,0], 'T':[0,1,0,0], 'C':[0,0,1,0], 'G':[0,0,0,1]}
    return np.array([mapping.get(b, [0,0,0,0]) for b in seq])

def sliding_windows(seq, window=1000, stride=500):
    """Membagi sekuens menjadi jendela-jendela yang saling tumpang tindih."""
    if window > len(seq):
        raise ValueError(f"Window ({window}) > panjang sekuens ({len(seq)}).")
    windows = []
    for start in range(0, len(seq) - window + 1, stride):
        windows.append(one_hot_encode(seq[start:start+window]))
    if not windows:
        raise ValueError("Tidak ada window yang dihasilkan.")
    return np.array(windows)

def gc_content(seq):
    """Proporsi GC (dalam 0..1)."""
    if not seq:
        return 0.0
    seq = seq.upper()
    return (seq.count('G') + seq.count('C')) / len(seq)

def reverse_coordinates(coding_regions, seq_length):
    """Mengubah koordinat CDS untuk sekuens reverse complement (0-indexed)."""
    return [(seq_length - end, seq_length - start) for (start, end) in coding_regions]

def label_from_annotations(sequence, coding_regions, window=1000, stride=500):
    """Buat label (1 = coding) berdasarkan overlap dengan CDS (0-indexed)."""
    if not coding_regions:
        raise ValueError("Daftar CDS kosong.")
    starts = list(range(0, len(sequence) - window + 1, stride))
    labels = []
    for start in starts:
        win_start, win_end = start, start + window
        overlap = any(max(win_start, cds[0]) < min(win_end, cds[1]) for cds in coding_regions)
        labels.append(1 if overlap else 0)
    labels = np.array(labels)
    if len(np.unique(labels)) == 1:
        print("⚠️ Peringatan: Label anotasi hanya satu kelas, beralih ke GC content.")
        return label_from_gc(sequence, window, stride)
    return labels

def label_from_gc(sequence, window=1000, stride=500):
    """Label berdasarkan median GC content per jendela."""
    windows = [sequence[i:i+window] for i in range(0, len(sequence)-window+1, stride)]
    gc_vals = [gc_content(w) for w in windows]
    threshold = np.median(gc_vals)
    labels = (np.array(gc_vals) > threshold).astype(int)
    if len(np.unique(labels)) == 1:
        threshold = np.mean(gc_vals)
        labels = (np.array(gc_vals) > threshold).astype(int)
    return labels

def fetch_fasta(accession, email="user@example.com"):
    Entrez.email = email
    try:
        handle = Entrez.efetch(db="nucleotide", id=accession, rettype="fasta", retmode="text")
        record = SeqIO.read(handle, "fasta")
        handle.close()
        return str(record.seq).upper(), None
    except Exception as e:
        print(f"❌ Gagal fetch FASTA: {e}")
        return None, None

def fetch_genbank(accession, email="user@example.com"):
    Entrez.email = email
    try:
        handle = Entrez.efetch(db="nucleotide", id=accession, rettype="gb", retmode="text")
        record = SeqIO.read(handle, "genbank")
        handle.close()
        seq = str(record.seq).upper()
        coding = []
        for feat in record.features:
            if feat.type == "CDS":
                start = feat.location.start.position
                end = feat.location.end.position
                coding.append((start, end))
        return seq, coding
    except Exception as e:
        print(f"❌ Gagal fetch GenBank: {e}")
        return None, None

def synthetic_sequence(length=16567):
    bases = ['A','T','C','G']
    probs = [0.3254, 0.2582, 0.2759, 0.1405]
    return ''.join(np.random.choice(bases, size=length, p=probs))

# -------------------- 3. PEMILIHAN MODE DATASET --------------------
ACCESSION = "KX129904"
print("\n" + "="*60)
print("📥 LOAD DATASET DARI NCBI")
print(f"URL: https://www.ncbi.nlm.nih.gov/nuccore/{ACCESSION}")
print("="*60)
print("\nPilih mode pengambilan data:")
print("1️⃣  Mode Cepat (FASTA only) → label berdasarkan GC content")
print("2️⃣  Mode Lengkap (FASTA + Anotasi) → label berdasarkan CDS (DIREKOMENDASIKAN)")
print("3️⃣  Mode Offline (Fallback) → data sintetis")

mode = input("\nMasukkan pilihan (1/2/3): ").strip()

sequence = None
coding_regions = None
label_method = None

try:
    if mode == "1":
        print("\n🔹 Mode 1: Mengambil FASTA...")
        sequence, _ = fetch_fasta(ACCESSION)
        if sequence:
            sequence = validate_sequence(sequence)
            print(f"✅ Sekuens valid, panjang = {len(sequence)} bp")
            label_method = "gc"
        else:
            raise Exception("Fallback ke mode sintetis")
    elif mode == "2":
        print("\n🔹 Mode 2: Mengambil FASTA + Anotasi (koordinat 0-indexed)...")
        sequence, coding_regions = fetch_genbank(ACCESSION)
        if sequence and coding_regions:
            sequence = validate_sequence(sequence)
            print(f"✅ Sekuens valid ({len(sequence)} bp), {len(coding_regions)} CDS ditemukan.")
            print(f"   Contoh CDS pertama: {coding_regions[0]}")
            label_method = "annotation"
        else:
            raise Exception("Fallback ke mode sintetis")
    else:
        print("\n🔹 Mode 3: Menggunakan data sintetis...")
        sequence = synthetic_sequence()
        sequence = validate_sequence(sequence)
        print(f"✅ Sekuens sintetis dibuat, panjang = {len(sequence)} bp")
        label_method = "gc"
except Exception as e:
    print(f"⚠️ Error: {e}. Menggunakan data sintetis sebagai fallback final.")
    sequence = synthetic_sequence()
    sequence = validate_sequence(sequence)
    label_method = "gc"

print(f"\n📌 Metode label yang digunakan: {label_method}")

# -------------------- 4. EKSPLORASI DATA (EDA) --------------------
print("\n" + "="*60)
print("📊 EKSPLORASI DATA AWAL (EDA)")
print("="*60)

total_len = len(sequence)
counts = {base: sequence.count(base) for base in 'ATCG'}
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Analisis Eksplorasi Genom Mitokondria Channa micropeltes", fontsize=14)

axes[0,0].pie(counts.values(), labels=counts.keys(), autopct='%1.1f%%', startangle=90)
axes[0,0].set_title("Komposisi Nukleotida (seluruh genom)")

windows_seq = [sequence[i:i+1000] for i in range(0, len(sequence)-999, 500)]
gc_windows = [gc_content(w) for w in windows_seq]
axes[0,1].hist(gc_windows, bins=30, edgecolor='black', alpha=0.7)
axes[0,1].axvline(np.median(gc_windows), color='red', linestyle='--', label=f'Median={np.median(gc_windows):.3f}')
axes[0,1].set_title("Distribusi GC Content per Window (1000 bp)")
axes[0,1].set_xlabel("GC Content"); axes[0,1].set_ylabel("Frekuensi"); axes[0,1].legend()

axes[1,0].boxplot(gc_windows, vert=True)
axes[1,0].set_title("Boxplot GC Content")
axes[1,0].set_ylabel("GC Content")
axes[1,0].set_xticklabels(['Window'])

from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(analyzer='char', vocabulary=['A','T','C','G'])
X_counts = vec.fit_transform(windows_seq[:100]).toarray()
corr = np.corrcoef(X_counts.T)
sns.heatmap(corr, annot=True, xticklabels=['A','T','C','G'], yticklabels=['A','T','C','G'], ax=axes[1,1], cmap='coolwarm')
axes[1,1].set_title("Korelasi Frekuensi Basa per Window")

plt.tight_layout()
plt.show()

temp_labels = label_from_gc(sequence, 1000, 500) if label_method == "gc" else label_from_annotations(sequence, coding_regions, 1000, 500)
unique, counts_class = np.unique(temp_labels, return_counts=True)
plt.figure(figsize=(6,4))
plt.bar(['Non-Coding (0)', 'Coding (1)'], counts_class, color=['skyblue', 'salmon'])
plt.title("Distribusi Kelas Awal (Sebelum Balancing)")
plt.ylabel("Jumlah Window")
for i, v in enumerate(counts_class):
    plt.text(i, v+5, str(v), ha='center')
plt.show()
print(f"Distribusi kelas awal: 0={counts_class[0]}, 1={counts_class[1]}")

# -------------------- 5. PREPROCESSING & AUGMENTASI DENGAN KOREKSI KOORDINAT --------------------
WINDOW = 1000
STRIDE = 500

X_fwd = sliding_windows(sequence, WINDOW, STRIDE)
if label_method == "annotation":
    y_fwd = label_from_annotations(sequence, coding_regions, WINDOW, STRIDE)
else:
    y_fwd = label_from_gc(sequence, WINDOW, STRIDE)

rev_seq = str(Seq(sequence).reverse_complement())
if label_method == "annotation":
    rev_coding_regions = reverse_coordinates(coding_regions, len(sequence))
    y_rev = label_from_annotations(rev_seq, rev_coding_regions, WINDOW, STRIDE)
else:
    y_rev = label_from_gc(rev_seq, WINDOW, STRIDE)

X_rev = sliding_windows(rev_seq, WINDOW, STRIDE)

X = np.vstack([X_fwd, X_rev])
y = np.concatenate([y_fwd, y_rev])
print(f"Total data setelah augmentasi: {X.shape[0]} sampel (fwd: {X_fwd.shape[0]}, rev: {X_rev.shape[0]})")
print(f"Distribusi label awal: 0={np.sum(y==0)}, 1={np.sum(y==1)}")

if len(np.unique(y)) == 1:
    print("⚠️ Hanya satu kelas, memaksa threshold berbeda.")
    new_labels = []
    for win in X:
        gc_win = np.mean(win[:,2] + win[:,3])
        new_labels.append(1 if gc_win > 0.4 else 0)
    y = np.array(new_labels)
    print(f"Setelah koreksi: 0={np.sum(y==0)}, 1={np.sum(y==1)}")

# -------------------- 6. PENYEIMBANGAN DENGAN SMOTE --------------------
n_samp, win_len, n_feat = X.shape
X_flat = X.reshape(n_samp, -1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_flat)

if len(np.unique(y)) > 1 and np.min(np.bincount(y)) < 0.4 * len(y):
    smote = SMOTE(random_state=42)
    X_bal, y_bal = smote.fit_resample(X_scaled, y)
    print(f"SMOTE diterapkan: {X_bal.shape[0]} sampel.")
else:
    X_bal, y_bal = X_scaled, y
    print("Data sudah seimbang, SMOTE tidak diperlukan.")

X_bal = X_bal.reshape(-1, win_len, n_feat)

try:
    X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2,
                                                        random_state=42, stratify=y_bal)
except ValueError:
    X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2,
                                                        random_state=42)
    print("⚠️ Stratifikasi gagal, menggunakan split biasa.")

# Pastikan label berbentuk 1D
y_train = y_train.astype(int).flatten()
y_test = y_test.astype(int).flatten()

print(f"Data latih: {X_train.shape[0]}, data uji: {X_test.shape[0]}")
print(f"Label latih: 0={np.sum(y_train==0)}, 1={np.sum(y_train==1)}")

# Validasi ukuran data training
if X_train.shape[0] < 2:
    raise RuntimeError("Data latih terlalu kecil (<2 sampel). Coba perbesar window/stride atau gunakan mode lain.")

# -------------------- 7. ARSITEKTUR CNN-BiLSTM (DENGAN REGULARISASI) --------------------
def build_model(input_shape):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(64, 5, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Bidirectional(layers.LSTM(64, return_sequences=True,
                                         dropout=0.2, recurrent_dropout=0.2))(x)
    x = layers.BatchNormalization()(x)

    x = layers.Bidirectional(layers.LSTM(32, return_sequences=False,
                                         dropout=0.2, recurrent_dropout=0.2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(32, activation='relu',
                     kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(2, activation='softmax')(x)
    return models.Model(inputs, outputs)

model = build_model((WINDOW, 4))
model.summary()

# PERBAIKAN 1: Gunakan float learning rate (0.001) – tidak pakai schedule
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# -------------------- 8. PELATIHAN (200 EPOCH, BATCH SIZE ADAPTIF, TANPA REDUCE_LR) --------------------
os.makedirs('saved_model', exist_ok=True)
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True, verbose=1)
checkpoint = callbacks.ModelCheckpoint('saved_model/best_model.keras', monitor='val_accuracy',
                                       save_best_only=True, verbose=1)

# Hanya dua callback: EarlyStopping dan ModelCheckpoint (ReduceLROnPlateau dihapus)
callbacks_list = [early_stop, checkpoint]

batch_size = min(32, len(X_train))
print(f"\n🚀 Memulai pelatihan dengan batch_size={batch_size}, epochs=200 (durasi ~6-9 menit)...")
start_time = time.time()
history = model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=200,
                    validation_split=0.2,
                    callbacks=callbacks_list,
                    verbose=1)
end_time = time.time()
print(f"✅ Pelatihan selesai dalam {(end_time - start_time)/60:.2f} menit.")

# Simpan model final dan scaler
model.save('saved_model/final_model.keras')
joblib.dump(scaler, 'saved_model/scaler.pkl')
print("✅ Model dan scaler disimpan di folder 'saved_model/'")

# -------------------- 9. EVALUASI DENGAN SKLEARN --------------------
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Coding', 'Coding'],
            yticklabels=['Non-Coding', 'Coding'])
plt.title('Confusion Matrix - Data Uji')
plt.ylabel('True Label'); plt.xlabel('Predicted Label')
plt.show()

print("\n📊 Classification Report (sklearn):")
print(classification_report(y_test, y_pred, target_names=['Non-Coding', 'Coding']))

# ROC Curve & AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:,1])
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.3f})', linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title('ROC Curve'); plt.legend(); plt.grid(True)
plt.show()

precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
print(f"\n📈 Metrik tambahan (binary): Precision={precision:.3f}, Recall={recall:.3f}, F1={f1:.3f}, AUC={roc_auc:.3f}")

# Kurva training (akurasi dan loss)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(history.history['accuracy'], label='Train Acc')
axes[0].plot(history.history['val_accuracy'], label='Val Acc')
axes[0].set_title('Akurasi selama training'); axes[0].legend(); axes[0].grid(True)
axes[1].plot(history.history['loss'], label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Val Loss')
axes[1].set_title('Loss selama training'); axes[1].legend(); axes[1].grid(True)
plt.tight_layout()
plt.show()

# Gap analisis
gap = np.array(history.history['accuracy']) - np.array(history.history['val_accuracy'])
plt.figure(figsize=(8,4))
plt.plot(gap, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.fill_between(range(len(gap)), 0, gap, where=(gap>0), alpha=0.3, color='red', label='Overfitting')
plt.fill_between(range(len(gap)), gap, 0, where=(gap<0), alpha=0.3, color='green', label='Underfitting')
plt.title('Gap Akurasi (Train - Validation)')
plt.xlabel('Epoch'); plt.ylabel('Gap'); plt.legend(); plt.grid(True)
plt.show()

final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
gap_acc = final_train_acc - final_val_acc
if gap_acc > 0.1:
    diagnosis = "⚠️ OVERFITTING (gap > 0.1) - Perlu regularisasi lebih kuat"
elif gap_acc < -0.05:
    diagnosis = "⚠️ UNDERFITTING (gap negatif) - Perlu kapasitas model lebih besar"
else:
    diagnosis = "✅ WELL-FITTED - Model generalisasi baik"

print("\n" + "="*50)
print("📋 RINGKASAN KINERJA MODEL")
print("="*50)
print(f"Train Accuracy      : {final_train_acc:.4f}")
print(f"Validation Accuracy : {final_val_acc:.4f}")
print(f"Gap (Train - Val)   : {gap_acc:.4f}")
print(f"Test Accuracy       : {(y_pred == y_test).mean():.4f}")
print(f"Test Precision      : {precision:.3f}")
print(f"Test Recall         : {recall:.3f}")
print(f"Test F1-Score       : {f1:.3f}")
print(f"Test AUC            : {roc_auc:.3f}")
print(f"Mode Dataset        : {mode}")
print(f"Diagnosis           : {diagnosis}")
print(f"Total waktu training: {(end_time - start_time)/60:.2f} menit")
print("="*50)

# -------------------- 10. DEPLOYMENT SEDERHANA (5 KASUS UJI) --------------------
print("\n" + "="*60)
print("🚀 DEPLOYMENT – PENGUJIAN MODEL DENGAN DATA BARU")
print("="*60)

try:
    deployed_model = tf.keras.models.load_model('saved_model/best_model.keras')
    deployed_scaler = joblib.load('saved_model/scaler.pkl')
    print("✅ Model dan scaler dimuat ulang dari penyimpanan.")
except:
    deployed_model = model
    deployed_scaler = scaler
    print("⚠️ Menggunakan model dari memori.")

def predict_new_sequence(seq, model, scaler, window=1000, stride=500):
    seq = validate_sequence(seq)
    if len(seq) < window:
        raise ValueError(f"Sekuens terlalu pendek ({len(seq)} bp), minimal {window} bp.")
    windows = sliding_windows(seq, window, stride)
    n_win = windows.shape[0]
    flat = windows.reshape(n_win, -1)
    scaled = scaler.transform(flat)
    scaled = scaled.reshape(n_win, window, 4)
    probs = model.predict(scaled, verbose=0)
    avg_prob_coding = np.mean(probs[:, 1])
    pred_class = 1 if avg_prob_coding > 0.5 else 0
    return pred_class, avg_prob_coding

test_cases = [
    ("GC-rich (tinggi)", "GCGCGCGCGC" * 100 + "ATATATATAT" * 50),
    ("AT-rich (rendah)", "ATATATATAT" * 150),
    ("Random komposisi mitokondria",
     ''.join(np.random.choice(['A','T','C','G'], size=1200, p=[0.3254,0.2582,0.2759,0.1405]))),
    ("Wilayah awal genom asli", sequence[:1200] if len(sequence)>=1200 else synthetic_sequence(1200)),
    ("Kodon tiruan (ATG...TAA)", "ATG" * 400 + "TAA")
]

results = []
print("\n📋 Hasil Prediksi untuk 5 Data Uji:")
for name, seq_test in test_cases:
    try:
        pred, prob = predict_new_sequence(seq_test, deployed_model, deployed_scaler, WINDOW, STRIDE)
        label_str = "Coding (1)" if pred == 1 else "Non-Coding (0)"
        results.append({"Nama Kasus": name, "Panjang (bp)": len(seq_test),
                        "Prediksi": label_str, "Prob. Coding": f"{prob:.4f}"})
        print(f"  ✅ {name}: {label_str} (prob coding = {prob:.4f})")
    except Exception as e:
        print(f"  ❌ {name}: Error - {e}")

df_test = pd.DataFrame(results)
print("\n📊 Tabel Ringkasan Prediksi Deployment:")
print(df_test.to_string(index=False))

plt.figure(figsize=(10,5))
names_plot = [r["Nama Kasus"] for r in results]
probs_plot = [float(r["Prob. Coding"]) for r in results]
colors = ['red' if p<0.5 else 'green' for p in probs_plot]
plt.barh(names_plot, probs_plot, color=colors)
plt.axvline(x=0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.xlabel('Probabilitas Coding')
plt.title('Hasil Prediksi Deployment pada 5 Data Uji')
plt.xlim(0,1)
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

df_test.to_csv('saved_model/deployment_results.csv', index=False)
print("\n💾 Hasil deployment disimpan di 'saved_model/deployment_results.csv'")
print("\n✅ Program selesai dijalankan dengan integritas dan konsistensi penuh.")
```

```python
# ====================================================================
# PROGRAM ANALISIS GENOM MITOKONDRIA DENGAN BIDIRECTIONAL RNN-LSTM
# ====================================================================
# Judul       : Implementasi Bi-LSTM untuk Klasifikasi Wilayah Coding pada
#               Genom Mitokondria Channa micropeltes (KX129904)
# Deskripsi   : Program ini melakukan pengambilan data dari NCBI (3 mode),
#               eksplorasi data (EDA), preprocessing dengan sliding window,
#               augmentasi forward & reverse complement + random substitution,
#               penyeimbangan SMOTE, pelatihan CNN-BiLSTM dengan regularisasi kuat,
#               evaluasi metrik (akurasi, precision, recall, F1, AUC) menggunakan sklearn,
#               serta deployment pada 5 sekuens uji.
#               Durasi komputasi dirancang antara 5–10 menit di GPU Google Colab.
# Penulis     : Berdasarkan analisis kritis dan perbaikan logis (anti-overfitting)
# Versi       : Final (1 Cell) – Standar jurnal internasional Q1–Q4
# ====================================================================

# -------------------- 1. PERSIAPAN LINGKUNGAN --------------------
!pip install -q biopython imbalanced-learn joblib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_fscore_support
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, regularizers
from Bio import Entrez, SeqIO
from Bio.Seq import Seq
import joblib
import os
import time
import warnings
warnings.filterwarnings('ignore')

# Reproduksibilitas
np.random.seed(42)
tf.random.set_seed(42)

print("✅ Lingkungan siap | TensorFlow version:", tf.__version__)

# -------------------- 2. FUNGSI UTILITAS (TERKOREKSI + AUGMENTASI) --------------------
def validate_sequence(seq):
    """Memastikan sekuens hanya berisi A, T, C, G dan panjang memadai."""
    if not isinstance(seq, str):
        raise ValueError("Sekuens harus berupa string.")
    seq = seq.upper()
    if not set(seq).issubset('ATCG'):
        raise ValueError(f"Karakter tidak valid: {set(seq) - set('ATCG')}")
    if len(seq) < 100:
        raise ValueError(f"Panjang sekuens ({len(seq)} bp) terlalu pendek.")
    return seq

def one_hot_encode(seq):
    """One-hot encoding (4 dimensi) untuk sekuens DNA."""
    mapping = {'A':[1,0,0,0], 'T':[0,1,0,0], 'C':[0,0,1,0], 'G':[0,0,0,1]}
    return np.array([mapping.get(b, [0,0,0,0]) for b in seq])

def sliding_windows(seq, window=1000, stride=500):
    """Membagi sekuens menjadi jendela-jendela yang saling tumpang tindih."""
    if window > len(seq):
        raise ValueError(f"Window ({window}) > panjang sekuens ({len(seq)}).")
    windows = []
    for start in range(0, len(seq) - window + 1, stride):
        windows.append(one_hot_encode(seq[start:start+window]))
    if not windows:
        raise ValueError("Tidak ada window yang dihasilkan.")
    return np.array(windows)

def random_substitution_augmentation(windows, mutation_rate=0.01):
    """Melakukan augmentasi dengan substitusi acak sejumlah mutation_rate persen basa per window.
       Input: array windows (n_windows, window_size, 4)
       Output: array windows_augmented (n_windows, window_size, 4) dengan mutasi."""
    windows_aug = windows.copy()
    n_windows, win_len, _ = windows_aug.shape
    n_mutations = max(1, int(win_len * mutation_rate))
    for i in range(n_windows):
        # Pilih posisi acak untuk dimutasi
        positions = np.random.choice(win_len, size=n_mutations, replace=False)
        for pos in positions:
            # Ubah one-hot encoding menjadi basa acak (A, T, C, G) dengan probabilitas sama
            new_base = np.random.choice(4)
            windows_aug[i, pos] = [0,0,0,0]
            windows_aug[i, pos, new_base] = 1
    return windows_aug

def gc_content(seq):
    """Proporsi GC (dalam 0..1)."""
    if not seq:
        return 0.0
    seq = seq.upper()
    return (seq.count('G') + seq.count('C')) / len(seq)

def reverse_coordinates(coding_regions, seq_length):
    """Mengubah koordinat CDS untuk sekuens reverse complement (0-indexed)."""
    return [(seq_length - end, seq_length - start) for (start, end) in coding_regions]

def label_from_annotations(sequence, coding_regions, window=1000, stride=500):
    """Buat label (1 = coding) berdasarkan overlap dengan CDS (0-indexed)."""
    if not coding_regions:
        raise ValueError("Daftar CDS kosong.")
    starts = list(range(0, len(sequence) - window + 1, stride))
    labels = []
    for start in starts:
        win_start, win_end = start, start + window
        overlap = any(max(win_start, cds[0]) < min(win_end, cds[1]) for cds in coding_regions)
        labels.append(1 if overlap else 0)
    labels = np.array(labels)
    if len(np.unique(labels)) == 1:
        print("⚠️ Peringatan: Label anotasi hanya satu kelas, beralih ke GC content.")
        return label_from_gc(sequence, window, stride)
    return labels

def label_from_gc(sequence, window=1000, stride=500):
    """Label berdasarkan median GC content per jendela."""
    windows = [sequence[i:i+window] for i in range(0, len(sequence)-window+1, stride)]
    gc_vals = [gc_content(w) for w in windows]
    threshold = np.median(gc_vals)
    labels = (np.array(gc_vals) > threshold).astype(int)
    if len(np.unique(labels)) == 1:
        threshold = np.mean(gc_vals)
        labels = (np.array(gc_vals) > threshold).astype(int)
    return labels

def fetch_fasta(accession, email="user@example.com"):
    Entrez.email = email
    try:
        handle = Entrez.efetch(db="nucleotide", id=accession, rettype="fasta", retmode="text")
        record = SeqIO.read(handle, "fasta")
        handle.close()
        return str(record.seq).upper(), None
    except Exception as e:
        print(f"❌ Gagal fetch FASTA: {e}")
        return None, None

def fetch_genbank(accession, email="user@example.com"):
    Entrez.email = email
    try:
        handle = Entrez.efetch(db="nucleotide", id=accession, rettype="gb", retmode="text")
        record = SeqIO.read(handle, "genbank")
        handle.close()
        seq = str(record.seq).upper()
        coding = []
        for feat in record.features:
            if feat.type == "CDS":
                start = feat.location.start.position
                end = feat.location.end.position
                coding.append((start, end))
        return seq, coding
    except Exception as e:
        print(f"❌ Gagal fetch GenBank: {e}")
        return None, None

def synthetic_sequence(length=16567):
    bases = ['A','T','C','G']
    probs = [0.3254, 0.2582, 0.2759, 0.1405]
    return ''.join(np.random.choice(bases, size=length, p=probs))

# -------------------- 3. PEMILIHAN MODE DATASET --------------------
ACCESSION = "KX129904"
print("\n" + "="*60)
print("📥 LOAD DATASET DARI NCBI")
print(f"URL: https://www.ncbi.nlm.nih.gov/nuccore/{ACCESSION}")
print("="*60)
print("\nPilih mode pengambilan data:")
print("1️⃣  Mode Cepat (FASTA only) → label berdasarkan GC content")
print("2️⃣  Mode Lengkap (FASTA + Anotasi) → label berdasarkan CDS (DIREKOMENDASIKAN)")
print("3️⃣  Mode Offline (Fallback) → data sintetis")

mode = input("\nMasukkan pilihan (1/2/3): ").strip()

sequence = None
coding_regions = None
label_method = None

try:
    if mode == "1":
        print("\n🔹 Mode 1: Mengambil FASTA...")
        sequence, _ = fetch_fasta(ACCESSION)
        if sequence:
            sequence = validate_sequence(sequence)
            print(f"✅ Sekuens valid, panjang = {len(sequence)} bp")
            label_method = "gc"
        else:
            raise Exception("Fallback ke mode sintetis")
    elif mode == "2":
        print("\n🔹 Mode 2: Mengambil FASTA + Anotasi (koordinat 0-indexed)...")
        sequence, coding_regions = fetch_genbank(ACCESSION)
        if sequence and coding_regions:
            sequence = validate_sequence(sequence)
            print(f"✅ Sekuens valid ({len(sequence)} bp), {len(coding_regions)} CDS ditemukan.")
            print(f"   Contoh CDS pertama: {coding_regions[0]}")
            label_method = "annotation"
        else:
            raise Exception("Fallback ke mode sintetis")
    else:
        print("\n🔹 Mode 3: Menggunakan data sintetis...")
        sequence = synthetic_sequence()
        sequence = validate_sequence(sequence)
        print(f"✅ Sekuens sintetis dibuat, panjang = {len(sequence)} bp")
        label_method = "gc"
except Exception as e:
    print(f"⚠️ Error: {e}. Menggunakan data sintetis sebagai fallback final.")
    sequence = synthetic_sequence()
    sequence = validate_sequence(sequence)
    label_method = "gc"

print(f"\n📌 Metode label yang digunakan: {label_method}")

# -------------------- 4. EKSPLORASI DATA (EDA) – STANDAR INTERNASIONAL --------------------
print("\n" + "="*60)
print("📊 EKSPLORASI DATA AWAL (EDA)")
print("="*60)

total_len = len(sequence)
counts = {base: sequence.count(base) for base in 'ATCG'}
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Analisis Eksplorasi Genom Mitokondria Channa micropeltes", fontsize=14)

axes[0,0].pie(counts.values(), labels=counts.keys(), autopct='%1.1f%%', startangle=90)
axes[0,0].set_title("Komposisi Nukleotida (seluruh genom)")

windows_seq = [sequence[i:i+1000] for i in range(0, len(sequence)-999, 500)]
gc_windows = [gc_content(w) for w in windows_seq]
axes[0,1].hist(gc_windows, bins=30, edgecolor='black', alpha=0.7)
axes[0,1].axvline(np.median(gc_windows), color='red', linestyle='--', label=f'Median={np.median(gc_windows):.3f}')
axes[0,1].set_title("Distribusi GC Content per Window (1000 bp)")
axes[0,1].set_xlabel("GC Content"); axes[0,1].set_ylabel("Frekuensi"); axes[0,1].legend()

axes[1,0].boxplot(gc_windows, vert=True)
axes[1,0].set_title("Boxplot GC Content")
axes[1,0].set_ylabel("GC Content")
axes[1,0].set_xticklabels(['Window'])

from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(analyzer='char', vocabulary=['A','T','C','G'])
X_counts = vec.fit_transform(windows_seq[:100]).toarray()
corr = np.corrcoef(X_counts.T)
sns.heatmap(corr, annot=True, xticklabels=['A','T','C','G'], yticklabels=['A','T','C','G'], ax=axes[1,1], cmap='coolwarm')
axes[1,1].set_title("Korelasi Frekuensi Basa per Window")

plt.tight_layout()
plt.show()

temp_labels = label_from_gc(sequence, 1000, 500) if label_method == "gc" else label_from_annotations(sequence, coding_regions, 1000, 500)
unique, counts_class = np.unique(temp_labels, return_counts=True)
plt.figure(figsize=(6,4))
plt.bar(['Non-Coding (0)', 'Coding (1)'], counts_class, color=['skyblue', 'salmon'])
plt.title("Distribusi Kelas Awal (Sebelum Balancing)")
plt.ylabel("Jumlah Window")
for i, v in enumerate(counts_class):
    plt.text(i, v+5, str(v), ha='center')
plt.show()
print(f"Distribusi kelas awal: 0={counts_class[0]}, 1={counts_class[1]}")

# -------------------- 5. PREPROCESSING & AUGMENTASI DENGAN KOREKSI KOORDINAT --------------------
WINDOW = 1000
STRIDE = 500

X_fwd = sliding_windows(sequence, WINDOW, STRIDE)
if label_method == "annotation":
    y_fwd = label_from_annotations(sequence, coding_regions, WINDOW, STRIDE)
else:
    y_fwd = label_from_gc(sequence, WINDOW, STRIDE)

rev_seq = str(Seq(sequence).reverse_complement())
if label_method == "annotation":
    rev_coding_regions = reverse_coordinates(coding_regions, len(sequence))
    y_rev = label_from_annotations(rev_seq, rev_coding_regions, WINDOW, STRIDE)
else:
    y_rev = label_from_gc(rev_seq, WINDOW, STRIDE)

X_rev = sliding_windows(rev_seq, WINDOW, STRIDE)

# Gabungkan data asli (forward + reverse)
X = np.vstack([X_fwd, X_rev])
y = np.concatenate([y_fwd, y_rev])

# AUGMENTASI TAMBAHAN: random substitution 1% untuk setiap window (double data)
X_aug = random_substitution_augmentation(X, mutation_rate=0.01)
y_aug = y.copy()  # label tetap sama karena mutasi acak tidak mengubah fungsi biologis secara signifikan

X = np.vstack([X, X_aug])
y = np.concatenate([y, y_aug])

print(f"Total data setelah augmentasi (reverse complement + random sub): {X.shape[0]} sampel")
print(f"Distribusi label: 0={np.sum(y==0)}, 1={np.sum(y==1)}")

if len(np.unique(y)) == 1:
    print("⚠️ Hanya satu kelas, memaksa threshold berbeda.")
    new_labels = []
    for win in X:
        gc_win = np.mean(win[:,2] + win[:,3])
        new_labels.append(1 if gc_win > 0.4 else 0)
    y = np.array(new_labels)
    print(f"Setelah koreksi: 0={np.sum(y==0)}, 1={np.sum(y==1)}")

# -------------------- 6. PENYEIMBANGAN DENGAN SMOTE --------------------
n_samp, win_len, n_feat = X.shape
X_flat = X.reshape(n_samp, -1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_flat)

if len(np.unique(y)) > 1 and np.min(np.bincount(y)) < 0.4 * len(y):
    smote = SMOTE(random_state=42)
    X_bal, y_bal = smote.fit_resample(X_scaled, y)
    print(f"SMOTE diterapkan: {X_bal.shape[0]} sampel.")
else:
    X_bal, y_bal = X_scaled, y
    print("Data sudah seimbang, SMOTE tidak diperlukan.")

X_bal = X_bal.reshape(-1, win_len, n_feat)

try:
    X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2,
                                                        random_state=42, stratify=y_bal)
except ValueError:
    X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.2,
                                                        random_state=42)
    print("⚠️ Stratifikasi gagal, menggunakan split biasa.")

# Pastikan label berbentuk 1D
y_train = y_train.astype(int).flatten()
y_test = y_test.astype(int).flatten()

print(f"Data latih: {X_train.shape[0]}, data uji: {X_test.shape[0]}")
print(f"Label latih: 0={np.sum(y_train==0)}, 1={np.sum(y_train==1)}")

# Validasi ukuran data training
if X_train.shape[0] < 2:
    raise RuntimeError("Data latih terlalu kecil (<2 sampel). Coba perbesar window/stride atau gunakan mode lain.")

# -------------------- 7. ARSITEKTUR CNN-BiLSTM DENGAN REGULARISASI KUAT (ANTI-OVERFITTING) --------------------
def build_model(input_shape):
    inputs = layers.Input(shape=input_shape)
    # Layer konvolusi dengan dropout tinggi
    x = layers.Conv1D(32, 5, padding='same', activation='relu', kernel_regularizer=regularizers.l2(1e-3))(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.5)(x)

    x = layers.Conv1D(64, 3, padding='same', activation='relu', kernel_regularizer=regularizers.l2(1e-3))(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.5)(x)

    # Bidirectional LSTM dengan unit lebih kecil dan dropout tinggi
    x = layers.Bidirectional(layers.LSTM(32, return_sequences=True,
                                         dropout=0.3, recurrent_dropout=0.3,
                                         kernel_regularizer=regularizers.l2(1e-3)))(x)
    x = layers.BatchNormalization()(x)

    x = layers.Bidirectional(layers.LSTM(16, return_sequences=False,
                                         dropout=0.3, recurrent_dropout=0.3,
                                         kernel_regularizer=regularizers.l2(1e-3)))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.6)(x)

    # Dense layer dengan regularisasi L2 kuat
    x = layers.Dense(16, activation='relu',
                     kernel_regularizer=regularizers.l2(1e-3))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.7)(x)

    outputs = layers.Dense(2, activation='softmax')(x)
    return models.Model(inputs, outputs)

model = build_model((WINDOW, 4))
model.summary()

# Optimizer dengan learning rate standar (tanpa schedule)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# -------------------- 8. PELATIHAN (300 EPOCH, EARLY STOPPING, BATCH SIZE ADAPTIF) --------------------
os.makedirs('saved_model', exist_ok=True)
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True, verbose=1)
checkpoint = callbacks.ModelCheckpoint('saved_model/best_model.keras', monitor='val_accuracy',
                                       save_best_only=True, verbose=1)
callbacks_list = [early_stop, checkpoint]

batch_size = min(32, len(X_train))
print(f"\n🚀 Memulai pelatihan dengan batch_size={batch_size}, epochs=300 (durasi ~6-10 menit)...")
start_time = time.time()
history = model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=300,
                    validation_split=0.2,
                    callbacks=callbacks_list,
                    verbose=1)
end_time = time.time()
print(f"✅ Pelatihan selesai dalam {(end_time - start_time)/60:.2f} menit.")

# Simpan model final dan scaler
model.save('saved_model/final_model.keras')
joblib.dump(scaler, 'saved_model/scaler.pkl')
print("✅ Model dan scaler disimpan di folder 'saved_model/'")

# -------------------- 9. EVALUASI DENGAN SKLEARN --------------------
y_pred_proba = model.predict(X_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Coding', 'Coding'],
            yticklabels=['Non-Coding', 'Coding'])
plt.title('Confusion Matrix - Data Uji')
plt.ylabel('True Label'); plt.xlabel('Predicted Label')
plt.show()

print("\n📊 Classification Report (sklearn):")
print(classification_report(y_test, y_pred, target_names=['Non-Coding', 'Coding']))

# ROC Curve & AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:,1])
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.3f})', linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title('ROC Curve'); plt.legend(); plt.grid(True)
plt.show()

precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
print(f"\n📈 Metrik tambahan (binary): Precision={precision:.3f}, Recall={recall:.3f}, F1={f1:.3f}, AUC={roc_auc:.3f}")

# Kurva training (akurasi dan loss)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(history.history['accuracy'], label='Train Acc')
axes[0].plot(history.history['val_accuracy'], label='Val Acc')
axes[0].set_title('Akurasi selama training'); axes[0].legend(); axes[0].grid(True)
axes[1].plot(history.history['loss'], label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Val Loss')
axes[1].set_title('Loss selama training'); axes[1].legend(); axes[1].grid(True)
plt.tight_layout()
plt.show()

# Gap analisis
gap = np.array(history.history['accuracy']) - np.array(history.history['val_accuracy'])
plt.figure(figsize=(8,4))
plt.plot(gap, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.fill_between(range(len(gap)), 0, gap, where=(gap>0), alpha=0.3, color='red', label='Overfitting')
plt.fill_between(range(len(gap)), gap, 0, where=(gap<0), alpha=0.3, color='green', label='Underfitting')
plt.title('Gap Akurasi (Train - Validation)')
plt.xlabel('Epoch'); plt.ylabel('Gap'); plt.legend(); plt.grid(True)
plt.show()

final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
gap_acc = final_train_acc - final_val_acc
if gap_acc > 0.1:
    diagnosis = "⚠️ OVERFITTING (gap > 0.1) - Perlu regularisasi lebih kuat"
elif gap_acc < -0.05:
    diagnosis = "⚠️ UNDERFITTING (gap negatif) - Perlu kapasitas model lebih besar"
else:
    diagnosis = "✅ WELL-FITTED - Model generalisasi baik"

print("\n" + "="*50)
print("📋 RINGKASAN KINERJA MODEL (Standar Jurnal Internasional)")
print("="*50)
print(f"Train Accuracy      : {final_train_acc:.4f}")
print(f"Validation Accuracy : {final_val_acc:.4f}")
print(f"Gap (Train - Val)   : {gap_acc:.4f}")
print(f"Test Accuracy       : {(y_pred == y_test).mean():.4f}")
print(f"Test Precision      : {precision:.3f}")
print(f"Test Recall         : {recall:.3f}")
print(f"Test F1-Score       : {f1:.3f}")
print(f"Test AUC            : {roc_auc:.3f}")
print(f"Mode Dataset        : {mode}")
print(f"Diagnosis           : {diagnosis}")
print(f"Total waktu training: {(end_time - start_time)/60:.2f} menit")
print("="*50)

# -------------------- 10. DEPLOYMENT SEDERHANA (5 KASUS UJI) --------------------
print("\n" + "="*60)
print("🚀 DEPLOYMENT – PENGUJIAN MODEL DENGAN DATA BARU")
print("="*60)

try:
    deployed_model = tf.keras.models.load_model('saved_model/best_model.keras')
    deployed_scaler = joblib.load('saved_model/scaler.pkl')
    print("✅ Model dan scaler dimuat ulang dari penyimpanan.")
except:
    deployed_model = model
    deployed_scaler = scaler
    print("⚠️ Menggunakan model dari memori.")

def predict_new_sequence(seq, model, scaler, window=1000, stride=500):
    seq = validate_sequence(seq)
    if len(seq) < window:
        raise ValueError(f"Sekuens terlalu pendek ({len(seq)} bp), minimal {window} bp.")
    windows = sliding_windows(seq, window, stride)
    n_win = windows.shape[0]
    flat = windows.reshape(n_win, -1)
    scaled = scaler.transform(flat)
    scaled = scaled.reshape(n_win, window, 4)
    probs = model.predict(scaled, verbose=0)
    avg_prob_coding = np.mean(probs[:, 1])
    pred_class = 1 if avg_prob_coding > 0.5 else 0
    return pred_class, avg_prob_coding

test_cases = [
    ("GC-rich (tinggi)", "GCGCGCGCGC" * 100 + "ATATATATAT" * 50),
    ("AT-rich (rendah)", "ATATATATAT" * 150),
    ("Random komposisi mitokondria",
     ''.join(np.random.choice(['A','T','C','G'], size=1200, p=[0.3254,0.2582,0.2759,0.1405]))),
    ("Wilayah awal genom asli", sequence[:1200] if len(sequence)>=1200 else synthetic_sequence(1200)),
    ("Kodon tiruan (ATG...TAA)", "ATG" * 400 + "TAA")
]

results = []
print("\n📋 Hasil Prediksi untuk 5 Data Uji:")
for name, seq_test in test_cases:
    try:
        pred, prob = predict_new_sequence(seq_test, deployed_model, deployed_scaler, WINDOW, STRIDE)
        label_str = "Coding (1)" if pred == 1 else "Non-Coding (0)"
        results.append({"Nama Kasus": name, "Panjang (bp)": len(seq_test),
                        "Prediksi": label_str, "Prob. Coding": f"{prob:.4f}"})
        print(f"  ✅ {name}: {label_str} (prob coding = {prob:.4f})")
    except Exception as e:
        print(f"  ❌ {name}: Error - {e}")

df_test = pd.DataFrame(results)
print("\n📊 Tabel Ringkasan Prediksi Deployment:")
print(df_test.to_string(index=False))

plt.figure(figsize=(10,5))
names_plot = [r["Nama Kasus"] for r in results]
probs_plot = [float(r["Prob. Coding"]) for r in results]
colors = ['red' if p<0.5 else 'green' for p in probs_plot]
plt.barh(names_plot, probs_plot, color=colors)
plt.axvline(x=0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.xlabel('Probabilitas Coding')
plt.title('Hasil Prediksi Deployment pada 5 Data Uji')
plt.xlim(0,1)
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

df_test.to_csv('saved_model/deployment_results.csv', index=False)
print("\n💾 Hasil deployment disimpan di 'saved_model/deployment_results.csv'")
print("\n✅ Program selesai dijalankan dengan integritas dan konsistensi penuh.")
```
