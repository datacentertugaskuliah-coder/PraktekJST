```python
# ============================================================
# JUDUL PROGRAM
# Prediksi Beban Listrik Menggunakan PGCB Hourly Generation Dataset
# Berbasis Upload File di Google Colab dengan ANN (ReLU dan MSE)
# ============================================================

# ============================================================
# DESKRIPSI SINGKAT
# Program ini disusun untuk membangun model prediksi beban listrik
# menggunakan dataset PGCB Hourly Generation Dataset yang diunggah
# langsung ke Google Colab.
#
# Alur analisis meliputi:
# 1. Upload file dataset (.zip, .xlsx, .xls, atau .csv) dari komputer lokal.
# 2. Pembacaan dataset secara otomatis.
# 3. Pemeriksaan struktur data.
# 4. Rekayasa fitur waktu.
# 5. Pra-pemrosesan data.
# 6. Pembagian data secara kronologis untuk menjaga integritas time series.
# 7. Pelatihan model ANN dengan ReLU pada hidden layer dan MSE sebagai loss.
# 8. Evaluasi model.
# 9. Penyimpanan model dan artefak preprocessing.
# 10. Simulasi deployment dengan 3 pengujian sederhana.
#
# Catatan akademik:
# Karena permasalahan ini merupakan regresi time series, fokus perbaikan
# analitik diarahkan pada konsistensi data, pencegahan data leakage,
# pengendalian outlier fitur, serta generalisasi model.
# ============================================================

!pip -q install openpyxl joblib tensorflow scikit-learn seaborn

import warnings, os, io, json, zipfile, random, joblib
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ------------------------------------------------------------
# 1. PENGATURAN AWAL
# ------------------------------------------------------------
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

plt.rcParams["figure.figsize"] = (12, 5)
sns.set_style("whitegrid")

print("="*90)
print("PROGRAM PREDIKSI BEBAN LISTRIK - PGCB HOURLY GENERATION DATASET")
print("="*90)

# ------------------------------------------------------------
# 2. UPLOAD FILE DATASET DARI GOOGLE COLAB
# ------------------------------------------------------------
from google.colab import files

print("\nSilakan unggah file dataset Anda (.zip, .xlsx, .xls, atau .csv).")
uploaded = files.upload()

if len(uploaded) == 0:
    raise ValueError("Tidak ada file yang diunggah.")

uploaded_filename = list(uploaded.keys())[0]
print(f'File yang diunggah: "{uploaded_filename}"')

# ------------------------------------------------------------
# 3. MEMBACA DATASET SECARA OTOMATIS
# ------------------------------------------------------------
print("\n[1] Membaca dataset secara otomatis...")

def baca_dataset_otomatis(file_name, file_bytes):
    lower_name = file_name.lower()

    # Jika file berupa ZIP, cari file Excel atau CSV di dalamnya
    if lower_name.endswith(".zip"):
        with zipfile.ZipFile(io.BytesIO(file_bytes), "r") as zf:
            daftar_file = zf.namelist()
            print("Isi file ZIP:", daftar_file)

            excel_files = [f for f in daftar_file if f.lower().endswith((".xlsx", ".xls"))]
            csv_files = [f for f in daftar_file if f.lower().endswith(".csv")]

            if len(excel_files) > 0:
                target_file = excel_files[0]
                print(f"File data yang dibaca dari ZIP: {target_file}")
                with zf.open(target_file) as f:
                    df = pd.read_excel(f)
                return df, target_file

            elif len(csv_files) > 0:
                target_file = csv_files[0]
                print(f"File data yang dibaca dari ZIP: {target_file}")
                with zf.open(target_file) as f:
                    df = pd.read_csv(f)
                return df, target_file

            else:
                raise FileNotFoundError("ZIP tidak berisi file .xlsx, .xls, atau .csv.")

    # Jika file langsung berupa Excel
    elif lower_name.endswith((".xlsx", ".xls")):
        df = pd.read_excel(io.BytesIO(file_bytes))
        return df, file_name

    # Jika file langsung berupa CSV
    elif lower_name.endswith(".csv"):
        df = pd.read_csv(io.BytesIO(file_bytes))
        return df, file_name

    else:
        raise ValueError("Format file tidak didukung. Gunakan .zip, .xlsx, .xls, atau .csv.")

df, sumber_file = baca_dataset_otomatis(uploaded_filename, uploaded[uploaded_filename])

print(f"Sumber data berhasil dibaca dari: {sumber_file}")
print("Ukuran dataset:", df.shape)

# ------------------------------------------------------------
# 4. PEMERIKSAAN AWAL DATA
# ------------------------------------------------------------
print("\n[2] Pemeriksaan awal data...")
print("Daftar kolom:", list(df.columns))

display(df.head())
display(df.dtypes.to_frame("Tipe_Data"))
display(df.isna().sum().to_frame("Jumlah_Missing"))

# ------------------------------------------------------------
# 5. PENETAPAN TARGET DAN KOLOM WAKTU SECARA LOGIS
# ------------------------------------------------------------
print("\n[3] Menetapkan target dan kolom waktu...")

# Target utama prediksi beban listrik
if "demand_mw" in df.columns:
    target_col = "demand_mw"
else:
    # fallback logis jika nama kolom sedikit berbeda
    kandidat_target = [c for c in df.columns if "demand" in str(c).lower() or "load" in str(c).lower()]
    if len(kandidat_target) == 0:
        raise ValueError(
            "Kolom target untuk beban listrik tidak ditemukan. "
            "Pastikan dataset memiliki kolom seperti 'demand_mw'."
        )
    target_col = kandidat_target[0]

print("Kolom target yang digunakan:", target_col)

# Deteksi kolom waktu
datetime_col = None
for c in df.columns:
    if "datetime" in str(c).lower():
        datetime_col = c
        break

if datetime_col is None:
    for c in df.columns:
        if "date" in str(c).lower():
            datetime_col = c
            break

if datetime_col is not None:
    df[datetime_col] = pd.to_datetime(df[datetime_col], errors="coerce")
    df = df.sort_values(datetime_col).reset_index(drop=True)
    print("Kolom waktu yang digunakan:", datetime_col)
else:
    print("Kolom waktu tidak terdeteksi eksplisit; urutan baris dipertahankan.")

# ------------------------------------------------------------
# 6. REKAYASA FITUR WAKTU
# ------------------------------------------------------------
print("\n[4] Rekayasa fitur waktu...")

if datetime_col is not None:
    df["hour"] = df[datetime_col].dt.hour
    df["day"] = df[datetime_col].dt.day
    df["dayofweek"] = df[datetime_col].dt.dayofweek
    df["month"] = df[datetime_col].dt.month
    df["weekofyear"] = df[datetime_col].dt.isocalendar().week.astype(int)
    df["is_weekend"] = (df["dayofweek"] >= 5).astype(int)

    # Fitur siklikal
    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
    df["dayofweek_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
    df["dayofweek_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)
    df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
    df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)

# ------------------------------------------------------------
# 7. PRA-PEMROSESAN DATA
# ------------------------------------------------------------
print("\n[5] Pra-pemrosesan data...")

# Hapus kolom yang tidak tepat dimasukkan langsung ke model
drop_cols = [target_col]
if datetime_col is not None:
    drop_cols.append(datetime_col)
if "remarks" in df.columns:
    drop_cols.append("remarks")

feature_cols = [c for c in df.columns if c not in drop_cols]

# Ubah fitur menjadi numerik bila memungkinkan
for c in feature_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Simpan hanya fitur numerik
feature_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(df[c])]

# Susun dataset akhir
df_model = df[feature_cols + [target_col]].copy()

# Target harus numerik
df_model[target_col] = pd.to_numeric(df_model[target_col], errors="coerce")
df_model = df_model.dropna(subset=[target_col]).reset_index(drop=True)

# Imputasi median untuk missing pada fitur
for c in feature_cols:
    df_model[c] = df_model[c].fillna(df_model[c].median())

print("Jumlah fitur yang digunakan:", len(feature_cols))
print("Daftar fitur:")
print(feature_cols)

# ------------------------------------------------------------
# 8. ANALISIS DESKRIPTIF DAN VISUALISASI
# ------------------------------------------------------------
print("\n[6] Analisis deskriptif dan visualisasi...")

display(df_model.describe().T)

ringkasan_target = pd.DataFrame({
    "Statistik": ["Minimum", "Maksimum", "Rata-rata", "Median", "Standar Deviasi", "Skewness"],
    "Nilai": [
        df_model[target_col].min(),
        df_model[target_col].max(),
        df_model[target_col].mean(),
        df_model[target_col].median(),
        df_model[target_col].std(),
        df_model[target_col].skew()
    ]
})
display(ringkasan_target)

plt.figure(figsize=(14, 5))
plt.plot(df_model[target_col].values, linewidth=0.7)
plt.title("Perubahan Target terhadap Urutan Waktu")
plt.xlabel("Urutan Observasi")
plt.ylabel(target_col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
sns.histplot(df_model[target_col], bins=50, kde=True)
plt.title(f"Distribusi Target ({target_col})")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 2.5))
sns.boxplot(x=df_model[target_col])
plt.title("Boxplot Target")
plt.tight_layout()
plt.show()

corr = df_model.corr(numeric_only=True)
top_cols = corr[target_col].abs().sort_values(ascending=False).head(min(12, len(corr))).index
plt.figure(figsize=(10, 8))
sns.heatmap(df_model[top_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Heatmap Korelasi Fitur Teratas")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 9. PEMBAGIAN DATA SECARA KRONOLOGIS
# ------------------------------------------------------------
print("\n[7] Pembagian data train, validation, dan test secara kronologis...")

X_all = df_model[feature_cols].copy()
y_all = df_model[target_col].copy()

n = len(df_model)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

X_train = X_all.iloc[:train_end].copy()
y_train = y_all.iloc[:train_end].copy()

X_val = X_all.iloc[train_end:val_end].copy()
y_val = y_all.iloc[train_end:val_end].copy()

X_test = X_all.iloc[val_end:].copy()
y_test = y_all.iloc[val_end:].copy()

print("Train      :", X_train.shape)
print("Validation :", X_val.shape)
print("Test       :", X_test.shape)

# ------------------------------------------------------------
# 10. PENANGANAN OUTLIER FITUR
# ------------------------------------------------------------
print("\n[8] Penanganan outlier fitur dengan clipping berbasis train...")

lower_bounds = X_train.quantile(0.01)
upper_bounds = X_train.quantile(0.99)

X_train = X_train.clip(lower=lower_bounds, upper=upper_bounds, axis=1)
X_val   = X_val.clip(lower=lower_bounds, upper=upper_bounds, axis=1)
X_test  = X_test.clip(lower=lower_bounds, upper=upper_bounds, axis=1)

# ------------------------------------------------------------
# 11. NORMALISASI FITUR
# ------------------------------------------------------------
print("\n[9] Normalisasi fitur...")
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

# ------------------------------------------------------------
# 12. MEMBANGUN MODEL ANN
# ------------------------------------------------------------
print("\n[10] Membangun model ANN...")

model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),
    Dense(128, activation="relu"),
    BatchNormalization(),
    Dropout(0.20),

    Dense(64, activation="relu"),
    BatchNormalization(),
    Dropout(0.15),

    Dense(32, activation="relu"),
    Dense(1, activation="linear")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="mse",
    metrics=["mae"]
)

model.summary()

# ------------------------------------------------------------
# 13. CALLBACK UNTUK PENCEGAHAN OVERFITTING
# ------------------------------------------------------------
print("\n[11] Menyiapkan callback...")

callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    ),
    ModelCheckpoint(
        filepath="best_pgcb_model.keras",
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    )
]

# ------------------------------------------------------------
# 14. PELATIHAN MODEL
# ------------------------------------------------------------
print("\n[12] Melatih model...")

history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=200,
    batch_size=256,
    callbacks=callbacks,
    verbose=1
)

# ------------------------------------------------------------
# 15. VISUALISASI KURVA PELATIHAN
# ------------------------------------------------------------
print("\n[13] Visualisasi kurva pelatihan...")

plt.figure(figsize=(12, 4))
plt.plot(history.history["loss"], label="Train Loss (MSE)")
plt.plot(history.history["val_loss"], label="Validation Loss (MSE)")
plt.title("Kurva Loss Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(history.history["mae"], label="Train MAE")
plt.plot(history.history["val_mae"], label="Validation MAE")
plt.title("Kurva MAE Pelatihan")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 16. EVALUASI MODEL
# ------------------------------------------------------------
print("\n[14] Evaluasi model...")

y_pred_train = model.predict(X_train_scaled, verbose=0).ravel()
y_pred_val   = model.predict(X_val_scaled, verbose=0).ravel()
y_pred_test  = model.predict(X_test_scaled, verbose=0).ravel()

def hitung_metrik(y_true, y_pred, label):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / np.where(np.abs(y_true) < 1e-8, 1e-8, np.abs(y_true)))) * 100
    return {
        "Dataset": label,
        "MSE": mse,
        "RMSE": rmse,
        "MAE": mae,
        "R2": r2,
        "MAPE(%)": mape
    }

tabel_metrik = pd.DataFrame([
    hitung_metrik(y_train.values, y_pred_train, "Train"),
    hitung_metrik(y_val.values, y_pred_val, "Validation"),
    hitung_metrik(y_test.values, y_pred_test, "Test")
])

display(tabel_metrik)

plt.figure(figsize=(14, 5))
plt.plot(y_test.values[:500], label="Aktual", linewidth=1)
plt.plot(y_pred_test[:500], label="Prediksi", linewidth=1)
plt.title("Perbandingan Aktual vs Prediksi pada 500 Data Uji Pertama")
plt.xlabel("Urutan Observasi")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 17. MENYIMPAN MODEL DAN ARTEFAK
# ------------------------------------------------------------
print("\n[15] Menyimpan model dan artefak...")

model.save("pgcb_ann_final_model.keras")
joblib.dump(scaler, "pgcb_feature_scaler.gz")
joblib.dump(feature_cols, "pgcb_feature_columns.gz")
joblib.dump(lower_bounds, "pgcb_lower_bounds.gz")
joblib.dump(upper_bounds, "pgcb_upper_bounds.gz")

konfigurasi = {
    "uploaded_filename": uploaded_filename,
    "source_data_file": sumber_file,
    "target_column": target_col,
    "feature_columns": feature_cols,
    "datetime_column": datetime_col,
    "train_rows": int(len(X_train)),
    "val_rows": int(len(X_val)),
    "test_rows": int(len(X_test)),
    "seed": SEED
}

with open("pgcb_model_config.json", "w") as f:
    json.dump(konfigurasi, f, indent=2)

print("File yang tersimpan:")
for file_name in [
    "best_pgcb_model.keras",
    "pgcb_ann_final_model.keras",
    "pgcb_feature_scaler.gz",
    "pgcb_feature_columns.gz",
    "pgcb_lower_bounds.gz",
    "pgcb_upper_bounds.gz",
    "pgcb_model_config.json"
]:
    print(f"- {file_name}: {os.path.exists(file_name)}")

# ------------------------------------------------------------
# 18. FUNGSI DEPLOYMENT
# ------------------------------------------------------------
def prediksi_deployment(input_df):
    loaded_model = tf.keras.models.load_model("pgcb_ann_final_model.keras")
    loaded_scaler = joblib.load("pgcb_feature_scaler.gz")
    loaded_features = joblib.load("pgcb_feature_columns.gz")
    loaded_lower = joblib.load("pgcb_lower_bounds.gz")
    loaded_upper = joblib.load("pgcb_upper_bounds.gz")

    data = input_df.copy()

    missing_cols = [c for c in loaded_features if c not in data.columns]
    if len(missing_cols) > 0:
        raise ValueError(f"Kolom input deployment belum lengkap: {missing_cols}")

    data = data[loaded_features].copy()

    for c in data.columns:
        data[c] = pd.to_numeric(data[c], errors="coerce")
        data[c] = data[c].fillna(0)

    data = data.clip(lower=loaded_lower, upper=loaded_upper, axis=1)
    data_scaled = loaded_scaler.transform(data)

    pred = loaded_model.predict(data_scaled, verbose=0).ravel()
    return pred

# ------------------------------------------------------------
# 19. PENGUJIAN DEPLOYMENT SEDERHANA (3 SKENARIO)
# ------------------------------------------------------------
print("\n[16] Pengujian deployment sederhana...")

uji_1 = X_test.iloc[[0]].copy()
aktual_1 = y_test.iloc[0]

uji_2 = X_test.iloc[[1]].copy()
aktual_2 = y_test.iloc[1]

uji_3 = X_train.mean().to_frame().T
for c in uji_3.columns[:min(3, len(uji_3.columns))]:
    uji_3[c] = uji_3[c] * 1.05
aktual_3 = np.nan

gabungan_uji = pd.concat([uji_1, uji_2, uji_3], ignore_index=True)
pred_uji = prediksi_deployment(gabungan_uji)

hasil_deployment = gabungan_uji.copy()
hasil_deployment.insert(0, "Skenario", [
    "Uji 1 - Data nyata pertama dari test set",
    "Uji 2 - Data nyata kedua dari test set",
    "Uji 3 - Data sintetis rata-rata train (+5%)"
])
hasil_deployment["Prediksi_Model"] = pred_uji
hasil_deployment["Aktual"] = [aktual_1, aktual_2, aktual_3]
hasil_deployment["Abs_Error"] = np.where(
    pd.isna(hasil_deployment["Aktual"]),
    np.nan,
    np.abs(hasil_deployment["Aktual"] - hasil_deployment["Prediksi_Model"])
)

display(hasil_deployment)

# ------------------------------------------------------------
# 20. INTERPRETASI HASIL DEPLOYMENT
# ------------------------------------------------------------
print("\n[17] Interpretasi hasil deployment...")

for i in range(len(hasil_deployment)):
    nama = hasil_deployment.loc[i, "Skenario"]
    pred = hasil_deployment.loc[i, "Prediksi_Model"]
    aktual = hasil_deployment.loc[i, "Aktual"]
    error = hasil_deployment.loc[i, "Abs_Error"]

    print(f"\n{nama}")
    print(f"Prediksi model : {pred:.4f}")
    if pd.notna(aktual):
        print(f"Nilai aktual   : {aktual:.4f}")
        print(f"Absolute error : {error:.4f}")
    else:
        print("Nilai aktual   : tidak tersedia karena data ini bersifat simulatif.")
        print("Interpretasi   : skenario ini digunakan untuk menguji kestabilan keluaran model.")

# ------------------------------------------------------------
# 21. KESIMPULAN OTOMATIS
# ------------------------------------------------------------
print("\n[18] Kesimpulan otomatis...")

hasil_test = tabel_metrik[tabel_metrik["Dataset"] == "Test"].iloc[0]

print(f'''
KESIMPULAN:
1. Program berhasil menggabungkan proses upload file dengan analisis dalam satu cell.
2. Dataset dibaca secara otomatis dari file upload lokal.
3. Model menggunakan ReLU pada hidden layer dan MSE sebagai loss.
4. Pembagian data dilakukan secara kronologis agar integritas time series terjaga.
5. Model dan artefak preprocessing berhasil disimpan.
6. Deployment sederhana telah diuji dengan 3 skenario.
7. Kinerja model pada data uji:
   - MSE   : {hasil_test["MSE"]:.4f}
   - RMSE  : {hasil_test["RMSE"]:.4f}
   - MAE   : {hasil_test["MAE"]:.4f}
   - R2    : {hasil_test["R2"]:.4f}
   - MAPE  : {hasil_test["MAPE(%)"]:.4f}%
''')
```
