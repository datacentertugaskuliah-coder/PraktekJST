```python
# ============================================================
# JUDUL PROGRAM
# ============================================================
# OPTIMASI JUMLAH NEURON HIDDEN LAYER DENGAN RANDOM SEARCH
# PADA DATASET DRY BEAN, DENGAN PENGENDALIAN IMBALANCE,
# UNDERFITTING, OVERFITTING, PENYIMPANAN MODEL,
# DAN PENGUJIAN DEPLOYMENT SEDERHANA
#
# Deskripsi singkat:
# Program ini mengimplementasikan pipeline machine learning lengkap untuk
# klasifikasi multiclass pada dataset Dry Bean. Program mendukung tiga
# sumber data (upload manual, file lokal, atau UCI), melakukan validasi
# struktur dataset, EDA, preprocessing, pencarian hyperparameter berbasis
# Random Search dengan  evaluasi model dengan
# metrik yang sesuai untuk data tidak seimbang, revisi logis berbasis
# resampling bila diperlukan, penyimpanan artefak model, dan pengujian
# deployment sederhana menggunakan 5 data uji.
#
# Catatan penting:
# - Program dirancang agar replikatif, defensif terhadap error umum, dan
#   konsisten secara metodologis.
# - Pemilihan model terbaik tidak hanya didasarkan pada accuracy, tetapi
#   pada skor komposit validasi yang mempertimbangkan balanced accuracy
#   dan macro-F1.
#   dengan pekerjaan komputasi yang relevan, yaitu Random Search bertahap.
# ============================================================

# ------------------------------------------------------------
# A. INSTALASI PUSTAKA
# ------------------------------------------------------------
!pip -q install ucimlrepo imbalanced-learn seaborn joblib openpyxl

# ------------------------------------------------------------
# B. IMPOR PUSTAKA
# ------------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")

import os
import json
import time
import math
import random
import traceback
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
    recall_score
)
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN

# ------------------------------------------------------------
# C. PENGATURAN GLOBAL DAN GAYA VISUAL
# ------------------------------------------------------------
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.keras.utils.set_random_seed(SEED)

# Gaya visual yang bersih dan mudah dibaca
sns.set_theme(style="whitegrid", context="talk")
plt.rcParams["figure.dpi"] = 140
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10
plt.rcParams["legend.fontsize"] = 10

# ------------------------------------------------------------
# D. KONFIGURASI EKSEKUSI
# ------------------------------------------------------------
# Sumber data:
# 1 = upload manual di Colab
# 2 = file lokal
# 3 = langsung dari UCI
DATA_SOURCE = 1

# Path file jika DATA_SOURCE = 2
LOCAL_FILE_PATH = "/content/dry_bean_dataset.csv"

# Folder keluaran
OUTPUT_DIR = "/content/output_drybean_final"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Anggaran waktu Random Search
MIN_RUNTIME_MINUTES = 5
MAX_RUNTIME_MINUTES = 10
MIN_RUNTIME_SECONDS = MIN_RUNTIME_MINUTES * 60
MAX_RUNTIME_SECONDS = MAX_RUNTIME_MINUTES * 60

# Jumlah trial minimum dan maksimum
MIN_TRIALS = 8
MAX_TRIALS = 18

# Epoch maksimum per trial
MAX_EPOCHS_PER_TRIAL = 35
PATIENCE = 6

# ------------------------------------------------------------
# E. FUNGSI UTILITAS
# ------------------------------------------------------------
def print_section(title):
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)

def safe_display(obj, title=None):
    if title is not None:
        print(title)
    try:
        display(obj)
    except:
        print(obj)

def standardize_target_column(df):
    """
    Menyeragamkan nama kolom target menjadi 'Class' agar seluruh pipeline
    berjalan konsisten.
    """
    target_candidates = ["Class", "class", "CLASS", "target", "Target", "label", "Label"]
    found = None
    for col in df.columns:
        if col in target_candidates:
            found = col
            break

    if found is not None and found != "Class":
        df = df.rename(columns={found: "Class"})

    if "Class" not in df.columns:
        raise ValueError(
            "Kolom target tidak ditemukan. Pastikan file memiliki kolom target "
            "seperti 'Class', 'class', 'target', atau 'label'."
        )
    return df

def validate_dataframe(df):
    """
    Memeriksa kualitas struktural dataset agar pipeline tidak gagal
    pada tahap training atau inferensi.
    """
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Objek data bukan pandas DataFrame.")

    if df.empty:
        raise ValueError("Dataset kosong.")

    # Hapus kolom indeks buatan bila ada
    df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

    if "Class" not in df.columns:
        raise ValueError("Kolom target 'Class' tidak ditemukan.")

    # Pastikan target tidak kosong
    if df["Class"].isnull().sum() > 0:
        raise ValueError("Kolom target mengandung missing values.")

    # Pastikan fitur numerik
    feature_cols = [c for c in df.columns if c != "Class"]
    if len(feature_cols) == 0:
        raise ValueError("Tidak ada kolom fitur.")

    for col in feature_cols:
        if not pd.api.types.is_numeric_dtype(df[col]):
            df[col] = pd.to_numeric(df[col], errors="coerce")

    missing_feature_total = df[feature_cols].isnull().sum().sum()
    if missing_feature_total > 0:
        missing_map = df[feature_cols].isnull().sum()
        missing_map = missing_map[missing_map > 0].to_dict()
        raise ValueError(
            f"Fitur mengandung missing values setelah konversi numerik: {missing_map}"
        )

    if df["Class"].nunique() < 2:
        raise ValueError("Jumlah kelas target kurang dari 2.")

    return df

def load_dataset(data_source=1, local_path=None):
    """
    Memuat dataset dari tiga sumber.
    """
    if data_source == 1:
        try:
            from google.colab import files
        except Exception:
            raise RuntimeError("Mode upload manual hanya dapat digunakan di Google Colab.")

        print("Silakan upload file dataset dalam format CSV atau XLSX.")
        uploaded = files.upload()
        if len(uploaded) == 0:
            raise ValueError("Tidak ada file yang diupload.")

        filename = list(uploaded.keys())[0]
        print(f"File terdeteksi: {filename}")

        if filename.lower().endswith(".csv"):
            df = pd.read_csv(filename)
        elif filename.lower().endswith(".xlsx"):
            df = pd.read_excel(filename)
        else:
            raise ValueError("Format file tidak didukung. Gunakan CSV atau XLSX.")

        df = standardize_target_column(df)
        df = validate_dataframe(df)
        return df

    elif data_source == 2:
        if local_path is None:
            raise ValueError("local_path harus diisi untuk DATA_SOURCE = 2.")
        if not os.path.exists(local_path):
            raise FileNotFoundError(f"File tidak ditemukan: {local_path}")

        if local_path.lower().endswith(".csv"):
            df = pd.read_csv(local_path)
        elif local_path.lower().endswith(".xlsx"):
            df = pd.read_excel(local_path)
        else:
            raise ValueError("Format file tidak didukung. Gunakan CSV atau XLSX.")

        df = standardize_target_column(df)
        df = validate_dataframe(df)
        return df

    elif data_source == 3:
        from ucimlrepo import fetch_ucirepo
        dry_bean = fetch_ucirepo(id=602)
        X_uci = dry_bean.data.features.copy()
        y_uci = dry_bean.data.targets.copy()

        if isinstance(y_uci, pd.DataFrame):
            y_uci = y_uci.iloc[:, 0]

        df = X_uci.copy()
        df["Class"] = y_uci
        df = standardize_target_column(df)
        df = validate_dataframe(df)
        return df

    else:
        raise ValueError("DATA_SOURCE harus bernilai 1, 2, atau 3.")

def build_model(config, input_dim, n_classes):
    """
    Membangun model neural network berdasarkan konfigurasi acak.
    """
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=(input_dim,)))

    model.add(tf.keras.layers.Dense(
        config["units_1"],
        activation="relu",
        kernel_regularizer=tf.keras.regularizers.l2(config["l2_1"])
    ))
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.Dropout(config["dropout_1"]))

    if config["n_hidden"] == 2:
        model.add(tf.keras.layers.Dense(
            config["units_2"],
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(config["l2_2"])
        ))
        model.add(tf.keras.layers.BatchNormalization())
        model.add(tf.keras.layers.Dropout(config["dropout_2"]))

    model.add(tf.keras.layers.Dense(n_classes, activation="softmax"))

    optimizer = tf.keras.optimizers.Adam(learning_rate=config["learning_rate"])
    model.compile(
        optimizer=optimizer,
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

def sample_random_config():
    """
    Menghasilkan konfigurasi acak untuk Random Search.
    """
    n_hidden = random.choice([1, 2])

    config = {
        "n_hidden": n_hidden,
        "units_1": random.choice(list(range(32, 257, 32))),
        "dropout_1": random.choice([0.10, 0.15, 0.20, 0.25, 0.30, 0.35]),
        "l2_1": random.choice([1e-5, 1e-4, 1e-3]),
        "learning_rate": random.choice([1e-2, 5e-3, 1e-3, 5e-4]),
        "batch_size": random.choice([16, 32, 64]),
    }

    if n_hidden == 2:
        config["units_2"] = random.choice(list(range(16, 129, 16)))
        config["dropout_2"] = random.choice([0.10, 0.15, 0.20, 0.25, 0.30, 0.35])
        config["l2_2"] = random.choice([1e-5, 1e-4, 1e-3])
    else:
        config["units_2"] = 0
        config["dropout_2"] = 0.0
        config["l2_2"] = 0.0

    return config

def config_to_key(config):
    """
    Mengubah konfigurasi menjadi representasi string agar trial tidak duplikat.
    """
    return json.dumps(config, sort_keys=True)

def evaluate_predictions(y_true, y_pred):
    """
    Menghitung metrik evaluasi utama.
    """
    acc = accuracy_score(y_true, y_pred)
    bal_acc = balanced_accuracy_score(y_true, y_pred)
    macro_f1 = f1_score(y_true, y_pred, average="macro")
    return acc, bal_acc, macro_f1

def composite_score(acc, bal_acc, macro_f1):
    """
    Skor komposit untuk pemilihan model.
    Penekanan terbesar diletakkan pada balanced accuracy.
    """
    return (0.15 * acc) + (0.50 * bal_acc) + (0.35 * macro_f1)

def run_single_trial(config, X_train_in, y_train_in, X_val_in, y_val_in, class_weight_dict, input_dim, n_classes):
    """
    Menjalankan satu trial training dan mengembalikan model, history, dan metrik validasi.
    """
    model = build_model(config, input_dim, n_classes)

    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor="val_loss",
            patience=PATIENCE,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=max(2, PATIENCE // 2),
            min_lr=1e-6,
            verbose=0
        )
    ]

    history = model.fit(
        X_train_in, y_train_in,
        validation_data=(X_val_in, y_val_in),
        epochs=MAX_EPOCHS_PER_TRIAL,
        batch_size=config["batch_size"],
        callbacks=callbacks,
        class_weight=class_weight_dict,
        verbose=0
    )

    # Evaluasi validation
    val_pred = np.argmax(model.predict(X_val_in, verbose=0), axis=1)
    train_pred = np.argmax(model.predict(X_train_in, verbose=0), axis=1)

    val_acc, val_bal_acc, val_macro_f1 = evaluate_predictions(y_val_in, val_pred)
    train_acc, train_bal_acc, train_macro_f1 = evaluate_predictions(y_train_in, train_pred)
    score = composite_score(val_acc, val_bal_acc, val_macro_f1)

    result = {
        "model": model,
        "history": history.history,
        "config": config,
        "epochs_run": len(history.history["loss"]),
        "train_accuracy": train_acc,
        "train_balanced_accuracy": train_bal_acc,
        "train_macro_f1": train_macro_f1,
        "val_accuracy": val_acc,
        "val_balanced_accuracy": val_bal_acc,
        "val_macro_f1": val_macro_f1,
        "val_score": score,
        "train_val_gap": train_acc - val_acc
    }
    return result

def evaluate_model_full(model, X_data, y_true, class_names, label="DATA"):
    """
    Evaluasi lengkap model pada satu subset data.
    """
    y_prob = model.predict(X_data, verbose=0)
    y_pred = np.argmax(y_prob, axis=1)

    acc, bal_acc, macro_f1 = evaluate_predictions(y_true, y_pred)
    recall_each = recall_score(y_true, y_pred, average=None)

    print(f"\n=== EVALUASI {label} ===")
    print(f"Accuracy          : {acc:.4f}")
    print(f"Balanced Accuracy : {bal_acc:.4f}")
    print(f"Macro F1          : {macro_f1:.4f}")

    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    report_df = pd.DataFrame(report).T
    safe_display(report_df, "Classification Report:")

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f"Confusion Matrix - {label}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    return {
        "accuracy": acc,
        "balanced_accuracy": bal_acc,
        "macro_f1": macro_f1,
        "per_class_recall": recall_each,
        "y_pred": y_pred,
        "report_df": report_df
    }

def diagnose_model(train_eval, val_eval, class_names):
    """
    Diagnosis kritis terhadap overfitting, underfitting, dan imbalance.
    """
    diagnostics = []

    gap = train_eval["accuracy"] - val_eval["accuracy"]
    min_recall = float(np.min(val_eval["per_class_recall"]))
    weak_class = class_names[int(np.argmin(val_eval["per_class_recall"]))]

    if gap > 0.05:
        diagnostics.append(f"Overfitting cukup kuat (gap train-validation = {gap:.4f}).")
    elif gap > 0.03:
        diagnostics.append(f"Overfitting ringan (gap train-validation = {gap:.4f}).")
    else:
        diagnostics.append(f"Overfitting relatif terkendali (gap = {gap:.4f}).")

    if train_eval["accuracy"] < 0.90 and val_eval["accuracy"] < 0.90 and val_eval["macro_f1"] < 0.90:
        diagnostics.append("Underfitting terindikasi karena performa train dan validation sama-sama belum tinggi.")
    else:
        diagnostics.append("Underfitting tidak dominan.")

    if (val_eval["accuracy"] - val_eval["balanced_accuracy"]) > 0.03 or min_recall < 0.80:
        diagnostics.append(
            f"Dampak imbalance masih terlihat; recall kelas terlemah ({weak_class}) = {min_recall:.4f}."
        )
    else:
        diagnostics.append("Dampak imbalance relatif terkendali.")

    return diagnostics

def save_artifacts(model, scaler, label_encoder, feature_names, metadata, output_dir):
    """
    Menyimpan model dan seluruh artefak yang diperlukan untuk deployment.
    """
    model_path = os.path.join(output_dir, "drybean_final_model.keras")
    scaler_path = os.path.join(output_dir, "drybean_scaler.joblib")
    le_path = os.path.join(output_dir, "drybean_label_encoder.joblib")
    feature_path = os.path.join(output_dir, "drybean_feature_names.json")
    meta_path = os.path.join(output_dir, "drybean_metadata.json")

    model.save(model_path)
    joblib.dump(scaler, scaler_path)
    joblib.dump(label_encoder, le_path)

    with open(feature_path, "w") as f:
        json.dump(list(feature_names), f, indent=2)

    with open(meta_path, "w") as f:
        json.dump(metadata, f, indent=2)

    return {
        "model_path": model_path,
        "scaler_path": scaler_path,
        "label_encoder_path": le_path,
        "feature_names_path": feature_path,
        "metadata_path": meta_path
    }

def load_artifacts(output_dir):
    """
    Memuat ulang artefak untuk simulasi deployment.
    """
    model = tf.keras.models.load_model(os.path.join(output_dir, "drybean_final_model.keras"))
    scaler = joblib.load(os.path.join(output_dir, "drybean_scaler.joblib"))
    label_encoder = joblib.load(os.path.join(output_dir, "drybean_label_encoder.joblib"))

    with open(os.path.join(output_dir, "drybean_feature_names.json"), "r") as f:
        feature_names = json.load(f)

    with open(os.path.join(output_dir, "drybean_metadata.json"), "r") as f:
        metadata = json.load(f)

    return model, scaler, label_encoder, feature_names, metadata

def predict_single_sample(input_dict, model, scaler, label_encoder, feature_names):
    """
    Prediksi satu data baru dengan validasi input yang ketat.
    """
    missing_features = [f for f in feature_names if f not in input_dict]
    if len(missing_features) > 0:
        raise ValueError(f"Fitur input kurang: {missing_features}")

    row = pd.DataFrame([[input_dict[f] for f in feature_names]], columns=feature_names)
    for col in feature_names:
        row[col] = pd.to_numeric(row[col], errors="raise")

    row_scaled = scaler.transform(row)
    probs = model.predict(row_scaled, verbose=0)[0]
    pred_idx = int(np.argmax(probs))
    pred_class = label_encoder.inverse_transform([pred_idx])[0]
    confidence = float(np.max(probs))

    prob_df = pd.DataFrame({
        "Class": list(label_encoder.classes_),
        "Probability": probs
    }).sort_values("Probability", ascending=False)

    return pred_class, confidence, prob_df

# ------------------------------------------------------------
# F. MEMUAT DATASET
# ------------------------------------------------------------
print_section("PEMUATAN DATASET")
df = load_dataset(DATA_SOURCE, LOCAL_FILE_PATH)
print("Ukuran dataset:", df.shape)
safe_display(df.head(), "Contoh lima baris pertama:")
print("\nInformasi dataset:")
df.info()
safe_display(df.isnull().sum(), "\nJumlah missing values:")

# ------------------------------------------------------------
# G. PEMISAHAN FITUR DAN TARGET
# ------------------------------------------------------------
print_section("PEMISAHAN FITUR DAN TARGET")
X = df.drop(columns=["Class"]).copy()
y = df["Class"].copy()

feature_names = list(X.columns)
print("Ukuran fitur:", X.shape)
print("Jumlah kelas:", y.nunique())
print("Daftar kelas:", sorted(y.unique()))

# ------------------------------------------------------------
# H. EXPLORATORY DATA ANALYSIS
# ------------------------------------------------------------
print_section("EXPLORATORY DATA ANALYSIS")

class_counts = y.value_counts().sort_values(ascending=False)
class_percent = (class_counts / len(y) * 100).round(2)
class_table = pd.DataFrame({"Jumlah": class_counts, "Persentase (%)": class_percent})
safe_display(class_table, "Distribusi kelas:")

plt.figure(figsize=(10, 5))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title("Distribusi Kelas Dry Bean")
plt.xlabel("Kelas")
plt.ylabel("Jumlah Sampel")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

safe_display(X.describe().T, "Statistik deskriptif fitur:")

plt.figure(figsize=(12, 8))
sns.heatmap(X.corr(), cmap="coolwarm", center=0)
plt.title("Heatmap Korelasi Antarfitur")
plt.tight_layout()
plt.show()

candidate_features = ["Area", "Perimeter", "MajorAxisLength", "MinorAxisLength"]
selected_features = [c for c in candidate_features if c in X.columns]
if len(selected_features) > 0:
    n_plot = len(selected_features)
    n_rows = math.ceil(n_plot / 2)
    fig, axes = plt.subplots(n_rows, 2, figsize=(14, 5 * n_rows))
    axes = np.array(axes).reshape(-1)

    for i, col in enumerate(selected_features):
        sns.boxplot(data=df, x="Class", y=col, ax=axes[i])
        axes[i].set_title(f"Boxplot {col} per Kelas")
        axes[i].tick_params(axis="x", rotation=45)

    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# I. PREPROCESSING
# ------------------------------------------------------------
print_section("PREPROCESSING")

le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = list(le.classes_)

# Split stratified
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y_encoded, test_size=0.15, random_state=SEED, stratify=y_encoded
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full,
    test_size=0.1765,  # kira-kira 15% dari total
    random_state=SEED,
    stratify=y_train_full
)

print("Ukuran data setelah split:")
print("Train      :", X_train.shape, y_train.shape)
print("Validation :", X_val.shape, y_val.shape)
print("Test       :", X_test.shape, y_test.shape)

# Standardisasi
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Class weights
classes_unique = np.unique(y_train)
weights = compute_class_weight(class_weight="balanced", classes=classes_unique, y=y_train)
class_weight_dict = {int(c): float(w) for c, w in zip(classes_unique, weights)}
print("\nClass weights:")
print(class_weight_dict)

input_dim = X_train_scaled.shape[1]
n_classes = len(class_names)

# ------------------------------------------------------------
# J. RANDOM SEARCH
# ------------------------------------------------------------
print_section("RANDOM SEARCH")

start_time = time.time()
seen_configs = set()
trial_results = []
best_result = None
trial_no = 0

while True:
    elapsed = time.time() - start_time

    # Hentikan jika melebihi batas maksimum
    if elapsed >= MAX_RUNTIME_SECONDS:
        print(f"Menghentikan pencarian karena mencapai batas waktu maksimum ({MAX_RUNTIME_MINUTES} menit).")
        break

    # Hentikan jika sudah melewati waktu minimum dan trial minimum
    if elapsed >= MIN_RUNTIME_SECONDS and trial_no >= MIN_TRIALS:
        print(f"Menghentikan pencarian karena target minimum waktu ({MIN_RUNTIME_MINUTES} menit) dan trial minimum telah tercapai.")
        break

    # Hentikan jika trial maksimum tercapai
    if trial_no >= MAX_TRIALS:
        print(f"Menghentikan pencarian karena trial maksimum ({MAX_TRIALS}) tercapai.")
        break

    # Ambil konfigurasi unik
    config = sample_random_config()
    key = config_to_key(config)
    if key in seen_configs:
        continue
    seen_configs.add(key)
    trial_no += 1

    trial_start = time.time()
    result = run_single_trial(
        config=config,
        X_train_in=X_train_scaled,
        y_train_in=y_train,
        X_val_in=X_val_scaled,
        y_val_in=y_val,
        class_weight_dict=class_weight_dict,
        input_dim=input_dim,
        n_classes=n_classes
    )
    trial_time = time.time() - trial_start

    result_row = {
        "trial": trial_no,
        "n_hidden": result["config"]["n_hidden"],
        "units_1": result["config"]["units_1"],
        "units_2": result["config"]["units_2"],
        "dropout_1": result["config"]["dropout_1"],
        "dropout_2": result["config"]["dropout_2"],
        "l2_1": result["config"]["l2_1"],
        "l2_2": result["config"]["l2_2"],
        "learning_rate": result["config"]["learning_rate"],
        "batch_size": result["config"]["batch_size"],
        "epochs_run": result["epochs_run"],
        "train_accuracy": result["train_accuracy"],
        "val_accuracy": result["val_accuracy"],
        "val_balanced_accuracy": result["val_balanced_accuracy"],
        "val_macro_f1": result["val_macro_f1"],
        "train_val_gap": result["train_val_gap"],
        "val_score": result["val_score"],
        "trial_seconds": trial_time
    }
    trial_results.append(result_row)

    # Simpan hasil terbaik
    if best_result is None or result["val_score"] > best_result["val_score"]:
        best_result = result

    print(
        f"Trial {trial_no:02d} | score={result['val_score']:.4f} | "
        f"val_acc={result['val_accuracy']:.4f} | "
        f"val_bal_acc={result['val_balanced_accuracy']:.4f} | "
        f"val_macro_f1={result['val_macro_f1']:.4f} | "
        f"waktu={trial_time:.1f} detik"
    )

search_elapsed = time.time() - start_time
print(f"\nTotal waktu Random Search: {search_elapsed/60:.2f} menit")
if search_elapsed < MIN_RUNTIME_SECONDS:
    print("Catatan: pada runtime ini target minimum waktu belum tercapai. Hal ini dapat terjadi jika lingkungan komputasi sangat cepat.")
elif search_elapsed > MAX_RUNTIME_SECONDS:
    print("Catatan: runtime melewati batas maksimum karena satu trial yang sedang berjalan diselesaikan terlebih dahulu.")
else:
    print("Random Search selesai dalam rentang waktu target.")

trial_df = pd.DataFrame(trial_results).sort_values("val_score", ascending=False).reset_index(drop=True)
safe_display(trial_df.head(10), "Sepuluh konfigurasi terbaik:")

plt.figure(figsize=(10, 5))
sns.scatterplot(data=trial_df, x="units_1", y="val_score", hue="n_hidden", size="val_balanced_accuracy", sizes=(80, 220))
plt.title("Skor Validasi Komposit terhadap Jumlah Neuron Hidden Layer")
plt.xlabel("Jumlah Neuron Hidden Layer 1")
plt.ylabel("Skor Validasi Komposit")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
trial_plot_df = trial_df.sort_values("trial")
plt.plot(trial_plot_df["trial"], trial_plot_df["val_accuracy"], marker="o", label="Validation Accuracy")
plt.plot(trial_plot_df["trial"], trial_plot_df["val_balanced_accuracy"], marker="s", label="Validation Balanced Accuracy")
plt.plot(trial_plot_df["trial"], trial_plot_df["val_macro_f1"], marker="^", label="Validation Macro-F1")
plt.plot(trial_plot_df["trial"], trial_plot_df["val_score"], marker="D", label="Validation Composite Score")
plt.title("Perubahan Kinerja Antar-Trial Random Search")
plt.xlabel("Trial")
plt.ylabel("Nilai Metrik")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# K. MODEL TERBAIK DARI RANDOM SEARCH
# ------------------------------------------------------------
print_section("MODEL TERBAIK DARI RANDOM SEARCH")
print("Konfigurasi terbaik:")
for k, v in best_result["config"].items():
    print(f"- {k}: {v}")

best_model = best_result["model"]
best_history = pd.DataFrame(best_result["history"])

plt.figure(figsize=(10, 5))
plt.plot(best_history["accuracy"], label="Train Accuracy")
plt.plot(best_history["val_accuracy"], label="Validation Accuracy")
plt.title("Learning Curve Accuracy - Model Terbaik")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(best_history["loss"], label="Train Loss")
plt.plot(best_history["val_loss"], label="Validation Loss")
plt.title("Learning Curve Loss - Model Terbaik")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# L. EVALUASI MODEL TERBAIK
# ------------------------------------------------------------
print_section("EVALUASI MODEL TERBAIK")
train_eval = evaluate_model_full(best_model, X_train_scaled, y_train, class_names, label="TRAIN")
val_eval = evaluate_model_full(best_model, X_val_scaled, y_val, class_names, label="VALIDATION")
test_eval = evaluate_model_full(best_model, X_test_scaled, y_test, class_names, label="TEST")

# ------------------------------------------------------------
# M. DIAGNOSIS KRITIS
# ------------------------------------------------------------
print_section("DIAGNOSIS KRITIS")
diagnostics = diagnose_model(train_eval, val_eval, class_names)
for i, item in enumerate(diagnostics, 1):
    print(f"{i}. {item}")

# ------------------------------------------------------------
# N.  OPSIONAL UNTUK IMBALANCE
# ------------------------------------------------------------
print_section(" OPSIONAL")
use_resampling = False
resampling_note = "Resampling tambahan tidak diterapkan."
final_model = best_model
final_model_name = "Model Terbaik Awal"
final_test_eval = test_eval

val_gap_imb = val_eval["accuracy"] - val_eval["balanced_accuracy"]
min_recall = float(np.min(val_eval["per_class_recall"]))

# Resampling hanya dilakukan bila benar-benar diperlukan
if val_gap_imb > 0.03 or min_recall < 0.80:
    use_resampling = True

if use_resampling:
    print("Model menunjukkan indikasi ketidakseimbangan performa antarkelas. Program mencoba revisi dengan resampling.")
    X_train_res, y_train_res = None, None
    sampler_used = None

    try:
        sampler = SMOTEENN(random_state=SEED)
        X_train_res, y_train_res = sampler.fit_resample(X_train_scaled, y_train)
        sampler_used = "SMOTEENN"
    except Exception as e:
        print(f"SMOTEENN gagal: {str(e)}")
        try:
            # k_neighbors diturunkan agar lebih aman untuk kelas kecil
            sampler = SMOTE(random_state=SEED, k_neighbors=3)
            X_train_res, y_train_res = sampler.fit_resample(X_train_scaled, y_train)
            sampler_used = "SMOTE"
        except Exception as e2:
            print(f"SMOTE juga gagal: {str(e2)}")
            sampler_used = None

    if sampler_used is not None:
        print(f"Resampling berhasil menggunakan: {sampler_used}")
        safe_display(pd.Series(y_train_res).value_counts().sort_index(), "Distribusi kelas setelah resampling:")

        revised_result = run_single_trial(
            config=best_result["config"],
            X_train_in=X_train_res,
            y_train_in=y_train_res,
            X_val_in=X_val_scaled,
            y_val_in=y_val,
            class_weight_dict=None,  # saat sudah di-resample, class weight tidak dipakai
            input_dim=input_dim,
            n_classes=n_classes
        )

        revised_model = revised_result["model"]
        revised_val_pred = np.argmax(revised_model.predict(X_val_scaled, verbose=0), axis=1)
        revised_val_acc, revised_val_bal_acc, revised_val_macro_f1 = evaluate_predictions(y_val, revised_val_pred)
        revised_val_score = composite_score(revised_val_acc, revised_val_bal_acc, revised_val_macro_f1)

        old_val_score = composite_score(val_eval["accuracy"], val_eval["balanced_accuracy"], val_eval["macro_f1"])

        print(f"Skor validasi model awal    : {old_val_score:.4f}")
        print(f"Skor validasi model revisi  : {revised_val_score:.4f}")

        if revised_val_score >= old_val_score:
            final_model = revised_model
            final_model_name = f"Model Revisi dengan {sampler_used}"
            final_test_eval = evaluate_model_full(final_model, X_test_scaled, y_test, class_names, label=f"TEST - {sampler_used}")
            resampling_note = f"Model revisi dengan {sampler_used} dipilih karena skor validasi komposit lebih baik."
        else:
            resampling_note = f"Model awal dipertahankan karena revisi {sampler_used} tidak meningkatkan skor validasi komposit."
    else:
        resampling_note = "Seluruh metode resampling gagal dijalankan, sehingga model awal dipertahankan."

print(resampling_note)

# ------------------------------------------------------------
# O. PENYIMPANAN MODEL
# ------------------------------------------------------------
print_section("PENYIMPANAN MODEL ")
metadata = {
    "program_title": "Optimasi Hidden Layer Dry Bean dengan Random Search",
    "data_source": int(DATA_SOURCE),
    "n_samples": int(df.shape[0]),
    "n_features": int(X.shape[1]),
    "class_names": list(class_names),
    "random_search_min_runtime_minutes": MIN_RUNTIME_MINUTES,
    "random_search_max_runtime_minutes": MAX_RUNTIME_MINUTES,
    "random_search_actual_runtime_minutes": round(search_elapsed / 60, 4),
    "best_config": best_result["config"],
    "final_model_name": final_model_name,
    "final_test_accuracy": float(final_test_eval["accuracy"]),
    "final_test_balanced_accuracy": float(final_test_eval["balanced_accuracy"]),
    "final_test_macro_f1": float(final_test_eval["macro_f1"]),
    "resampling_note": resampling_note,
    "seed": SEED
}

artifact_paths = save_artifacts(
    model=final_model,
    scaler=scaler,
    label_encoder=le,
    feature_names=feature_names,
    metadata=metadata,
    output_dir=OUTPUT_DIR
)

print("Artefak tersimpan:")
for k, v in artifact_paths.items():
    print(f"- {k}: {v}")

# ------------------------------------------------------------
# P. PENGUJIAN DEPLOYMENT SEDERHANA (5 DATA)
# ------------------------------------------------------------
print_section("PENGUJIAN DEPLOYMENT SEDERHANA (5 DATA)")
loaded_model, loaded_scaler, loaded_le, loaded_features, loaded_metadata = load_artifacts(OUTPUT_DIR)
safe_display(pd.DataFrame([loaded_metadata]), "Metadata model yang dimuat ulang:")

# Data mentah test untuk simulasi deployment
X_test_raw = X_test.reset_index(drop=True)
y_test_labels = pd.Series(le.inverse_transform(y_test)).reset_index(drop=True)

n_dev_tests = min(5, len(X_test_raw))
sample_indices = np.random.choice(len(X_test_raw), size=n_dev_tests, replace=False)

dev_rows = []
for i, idx in enumerate(sample_indices, 1):
    sample_input = X_test_raw.loc[idx, loaded_features].to_dict()
    true_label = y_test_labels.loc[idx]

    pred_label, confidence, prob_df = predict_single_sample(
        input_dict=sample_input,
        model=loaded_model,
        scaler=loaded_scaler,
        label_encoder=loaded_le,
        feature_names=loaded_features
    )

    dev_rows.append({
        "No": i,
        "Index_Test": int(idx),
        "Label_Aktual": true_label,
        "Prediksi_Model": pred_label,
        "Confidence": round(confidence, 4),
        "Benar": pred_label == true_label
    })

    print(f"\nPengujian ke-{i}")
    print(f"Label aktual   : {true_label}")
    print(f"Prediksi model : {pred_label}")
    print(f"Confidence     : {confidence:.4f}")
    safe_display(prob_df.head(3), "Top-3 probabilitas kelas:")

dev_results_df = pd.DataFrame(dev_rows)
safe_display(dev_results_df, "Ringkasan pengujian deployment sederhana:")

dev_success = dev_results_df["Benar"].mean()
print(f"Tingkat keberhasilan pada {n_dev_tests} pengujian sederhana: {dev_success:.2%}")

plt.figure(figsize=(9, 4.5))
sns.barplot(data=dev_results_df, x="No", y="Confidence", hue="Benar")
plt.title("Confidence Prediksi pada Pengujian Deployment Sederhana")
plt.xlabel("Nomor Pengujian")
plt.ylabel("Confidence")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# Probabilitas kelas untuk satu contoh
example_input = X_test_raw.loc[sample_indices[0], loaded_features].to_dict()
pred_label, confidence, prob_df = predict_single_sample(
    input_dict=example_input,
    model=loaded_model,
    scaler=loaded_scaler,
    label_encoder=loaded_le,
    feature_names=loaded_features
)

plt.figure(figsize=(8, 4.5))
sns.barplot(data=prob_df, x="Class", y="Probability")
plt.title("Distribusi Probabilitas Kelas untuk Satu Sampel Uji")
plt.xlabel("Kelas")
plt.ylabel("Probability")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Q. KESIMPULAN
# ------------------------------------------------------------
print_section("KESIMPULAN ")
print("1. Program telah memuat, memvalidasi, dan memproses dataset secara defensif untuk mengurangi error umum.")
print("2. Random Search dijalankan dengan anggaran waktu, sehingga runtime tetap terkendali dan relevan secara metodologis.")
print("3. Model terbaik dipilih berdasarkan skor komposit validasi, bukan accuracy tunggal.")
print("4. Evaluasi dilakukan dengan accuracy, balanced accuracy, macro-F1, classification report, confusion matrix, tabel, dan grafik.")
print("5. Artefak model untuk deployment telah disimpan dan diuji kembali pada lima data uji.")
print("6. Seluruh alur dirancang agar dapat direproduksi pada Google Colab dengan sumber data yang sama.")
```
