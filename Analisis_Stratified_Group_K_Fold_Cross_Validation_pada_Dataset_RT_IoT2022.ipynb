```python
# ============================================================
# JUDUL PROGRAM
# Analisis Stratified Group K-Fold Cross-Validation pada Dataset RT-IoT2022
# dengan Pengendalian Imbalance, Underfitting, Overfitting,
# Penyimpanan Model, dan Pengujian Deployment Sederhana
# ============================================================

!pip -q install ucimlrepo joblib

import warnings
warnings.filterwarnings("ignore")

import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedGroupKFold
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.ensemble import RandomForestClassifier

# ------------------------------------------------------------
# BAGIAN 1. FUNGSI PEMUATAN DATASET (TETAP SAMA)
# ------------------------------------------------------------
def load_dataset_flexible():
    print("Pilih sumber dataset:")
    print("1. Upload manual file CSV")
    print("2. Ambil dataset langsung dari UCI")
    print("3. Baca file CSV dari path lokal/runtime")
    source_choice = input("Masukkan pilihan (1/2/3): ").strip()

    if source_choice == "1":
        print("\nSilakan upload file CSV Anda.")
        from google.colab import files
        uploaded = files.upload()
        if not uploaded: raise ValueError("Tidak ada file yang di-upload.")
        fn = list(uploaded.keys())[0]
        df = pd.read_csv(fn)
        if df.empty: raise ValueError("Dataset hasil upload kosong.")
        print(f"\nFile berhasil di-upload: {fn}")
        print(f"Ukuran dataset: {df.shape}")
        return df, "upload_manual", fn
    elif source_choice == "2":
        print("\nMengambil dataset dari UCI...")
        from ucimlrepo import fetch_ucirepo
        rt_iot = fetch_ucirepo(id=942)
        X = rt_iot.data.features.copy()
        y_raw = rt_iot.data.targets.copy()
        if isinstance(y_raw, pd.DataFrame) and y_raw.shape[1] >= 1:
            y = y_raw.iloc[:, 0].copy()
        else:
            y = pd.Series(y_raw, name="target")
        target_name = y.name if y.name is not None else "target"
        df = X.copy().reset_index(drop=True)
        df[target_name] = pd.Series(y).reset_index(drop=True)
        if df.empty: raise ValueError("Dataset dari UCI kosong.")
        print("\nDataset berhasil diambil dari UCI.")
        print(f"Ukuran dataset: {df.shape}")
        print(f"Nama target terdeteksi: {target_name}")
        return df, "uci", target_name
    elif source_choice == "3":
        file_path = input("Masukkan path file CSV: ").strip()
        if not os.path.exists(file_path): raise FileNotFoundError(f"File tidak ditemukan: {file_path}")
        df = pd.read_csv(file_path)
        if df.empty: raise ValueError("Dataset dari path lokal kosong.")
        print(f"\nFile berhasil dibaca dari path: {file_path}")
        print(f"Ukuran dataset: {df.shape}")
        return df, "local_path", file_path
    else:
        raise ValueError("Pilihan tidak valid. Gunakan 1, 2, atau 3.")

# ------------------------------------------------------------
# BAGIAN 2. IDENTIFIKASI TARGET & GRUP LOGIS
# ------------------------------------------------------------
def identify_target_column(df):
    if df.empty: raise ValueError("DataFrame kosong.")
    print("\nDaftar kolom dataset:")
    print(df.columns.tolist())
    kandidat = ["label", "Label", "target", "Target", "class", "Class", "Attack_type", "attack_type", "Attack", "attack"]
    detected = [c for c in kandidat if c in df.columns]
    if detected:
        print(f"\nKolom target terdeteksi: {detected[0]}")
        if input("Gunakan kolom ini sebagai target? (y/n): ").strip().lower() == "y":
            return detected[0]
    col = input("Masukkan nama kolom target secara manual: ").strip()
    if col not in df.columns: raise ValueError(f"Kolom '{col}' tidak ditemukan.")
    return col

def create_logical_group(df, proto_col="proto", resp_port_col="id.resp_p"):
    """
    Membuat kolom grup logis untuk mencegah data leakage.
    Pada dataset IoT/Network, grouping berdasarkan protokol + port tujuan
    adalah pendekatan yang aman secara metodologis (tidak mengandung label).
    """
    group_id = df[proto_col].astype(str) + "_" + df[resp_port_col].astype(str)
    df["group_id"] = group_id
    unique_groups = df["group_id"].nunique()
    print(f"\nKolom grup logis dibuat: 'group_id' | Jumlah grup unik: {unique_groups}")
    return df

# ------------------------------------------------------------
# BAGIAN 3. PREPROCESSING
# ------------------------------------------------------------
def build_preprocessor(X):
    num_cols = X.select_dtypes(include=["number"]).columns.tolist()
    cat_cols = X.select_dtypes(exclude=["number"]).columns.tolist()
    transformers = []
    if num_cols:
        transformers.append(("num", Pipeline([("imp", SimpleImputer(strategy="median")), ("sc", StandardScaler())]), num_cols))
    if cat_cols:
        transformers.append(("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")), ("oh", OneHotEncoder(handle_unknown="ignore"))]), cat_cols))
    if not transformers: raise ValueError("Tidak ada kolom numerik maupun kategorikal.")
    return ColumnTransformer(transformers), num_cols, cat_cols

# ------------------------------------------------------------
# BAGIAN 4. EVALUASI STRATIFIED GROUP K-FOLD
# ------------------------------------------------------------
def evaluate_stratified_group_kfold(X, y_encoded, groups, preprocessor, class_names, model_params, n_splits=5, random_state=42):
    if len(X) != len(y_encoded) != len(groups):
        raise ValueError("Jumlah baris fitur, target, dan grup harus sama.")
    if len(np.unique(y_encoded)) < 2:
        raise ValueError("Target hanya memiliki satu kelas.")
    if groups.nunique() < n_splits:
        raise ValueError(f"Jumlah grup unik ({groups.nunique()}) lebih kecil dari n_splits ({n_splits}).")

    sgkf = StratifiedGroupKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    fold_results = []
    oof_true, oof_pred = [], []

    for fold, (train_idx, test_idx) in enumerate(sgkf.split(X, y_encoded, groups), start=1):
        X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
        y_tr, y_te = y_encoded[train_idx], y_encoded[test_idx]

        pipe = Pipeline([("prep", preprocessor), ("clf", RandomForestClassifier(**model_params))])
        pipe.fit(X_tr, y_tr)
        p_tr = pipe.predict(X_tr)
        p_te = pipe.predict(X_te)

        oof_true.extend(y_te)
        oof_pred.extend(p_te)

        fold_results.append({
            "fold": fold,
            "train_accuracy": accuracy_score(y_tr, p_tr),
            "test_accuracy": accuracy_score(y_te, p_te),
            "train_f1_macro": f1_score(y_tr, p_tr, average="macro", zero_division=0),
            "test_f1_macro": f1_score(y_te, p_te, average="macro", zero_division=0),
            "train_f1_weighted": f1_score(y_tr, p_tr, average="weighted", zero_division=0),
            "test_f1_weighted": f1_score(y_te, p_te, average="weighted", zero_division=0),
            "train_precision_macro": precision_score(y_tr, p_tr, average="macro", zero_division=0),
            "test_precision_macro": precision_score(y_te, p_te, average="macro", zero_division=0),
            "train_recall_macro": recall_score(y_tr, p_tr, average="macro", zero_division=0),
            "test_recall_macro": recall_score(y_te, p_te, average="macro", zero_division=0),
        })

    fold_df = pd.DataFrame(fold_results)
    summary_df = fold_df.drop(columns=["fold"]).agg(["mean", "std"]).T.reset_index()
    summary_df.columns = ["metric", "mean", "std"]

    oof_true, oof_pred = np.array(oof_true), np.array(oof_pred)
    report_df = pd.DataFrame(classification_report(oof_true, oof_pred, target_names=class_names, output_dict=True, zero_division=0)).T
    cm = confusion_matrix(oof_true, oof_pred)

    avg_train = fold_df["train_f1_macro"].mean()
    avg_test = fold_df["test_f1_macro"].mean()
    gap = avg_train - avg_test

    if gap > 0.05 and avg_train > avg_test:
        fit_status = "Indikasi overfitting"
    elif avg_train < 0.75 and avg_test < 0.75:
        fit_status = "Indikasi underfitting"
    else:
        fit_status = "Relatif seimbang"

    diagnostics = {"avg_train_f1_macro": avg_train, "avg_test_f1_macro": avg_test, "generalization_gap": gap, "fit_status": fit_status}
    return fold_df, summary_df, report_df, cm, diagnostics

# ------------------------------------------------------------
# BAGIAN 5 & 6. SIMPAN MODEL & DEPLOYMENT (SAMA SEBELUMNYA)
# ------------------------------------------------------------
def save_model(model, label_encoder, target_col, feature_columns, model_path="model_sgkfold.pkl", meta_path="model_sgkfold_metadata.json"):
    joblib.dump(model, model_path)
    with open(meta_path, "w", encoding="utf-8") as f:
        json.dump({"target_column": target_col, "feature_columns": feature_columns, "class_labels": list(label_encoder.classes_)}, f, indent=4)
    print(f"\nModel disimpan ke: {model_path}")
    print(f"Metadata disimpan ke: {meta_path}")

def simple_deployment_test(model, X_source, y_source, label_encoder, n_samples=5, random_state=42):
    if n_samples not in [3, 5]: raise ValueError("Sampel harus 3 atau 5.")
    if len(X_source) < n_samples: raise ValueError("Data tidak cukup.")
    samp = X_source.sample(n=n_samples, random_state=random_state)
    y_true = y_source.loc[samp.index].astype(str).values
    y_pred = label_encoder.inverse_transform(model.predict(samp))
    res = samp.copy()
    res["Aktual"], res["Prediksi"] = y_true, y_pred
    res["Kesesuaian"] = np.where(res["Aktual"] == res["Prediksi"], "Benar", "Salah")
    return res, (res["Kesesuaian"] == "Benar").mean()

# ============================================================
# EKSEKUSI UTAMA
# ============================================================
if __name__ == "__main__":
    df, src_type, src_info = load_dataset_flexible()
    target_col = identify_target_column(df)
    df = create_logical_group(df) # Membuat grup logis

    print("\n===== INFORMASI SUMBER DATA =====")
    print(f"Sumber: {src_type} | Info: {src_info} | Target: {target_col}")

    X = df.drop(columns=[target_col, "group_id"]).copy()
    y = df[target_col].copy()
    groups = df["group_id"].copy()

    # Label Encoding
    le = LabelEncoder()
    y_enc = le.fit_transform(y.astype(str))
    class_names = le.classes_

    # Distribusi & Imbalance
    class_dist = pd.Series(y.astype(str)).value_counts().sort_values(ascending=False)
    imb_ratio = class_dist.iloc[0] / class_dist.iloc[-1] if len(class_dist) > 1 else 1.0
    print("\n===== DISTRIBUSI KELAS & IMBALANCE =====")
    print(class_dist)
    print(f"Imbalance Ratio: {imb_ratio:.2f}")

    # Preprocessing
    preprocessor, num_cols, cat_cols = build_preprocessor(X)
    print(f"\nKolom Numerik: {len(num_cols)} | Kategorikal: {len(cat_cols)}")

    # Model Config
    model_params = {
        "n_estimators": 300, "max_depth": 20, "min_samples_split": 6,
        "min_samples_leaf": 2, "max_features": "sqrt", "class_weight": "balanced_subsample",
        "random_state": 42, "n_jobs": -1
    }

    # Evaluasi Stratified Group K-Fold
    fold_df, summary_df, report_df, cm, diagnostics = evaluate_stratified_group_kfold(
        X, y_enc, groups, preprocessor, class_names, model_params, n_splits=5, random_state=42
    )

    print("\n===== HASIL PER FOLD =====")
    display(fold_df)
    print("\n===== RINGKASAN HASIL =====")
    display(summary_df)
    print("\n===== CLASSIFICATION REPORT (OUT-OF-FOLD) =====")
    display(report_df)
    print("\n===== DIAGNOSTIK MODEL =====")
    for k, v in diagnostics.items(): print(f"{k}: {v}")

    # Visualisasi
    plt.figure(figsize=(10, 5))
    plt.plot(fold_df["fold"], fold_df["train_f1_macro"], marker='o', label="Train F1 Macro")
    plt.plot(fold_df["fold"], fold_df["test_f1_macro"], marker='o', label="Validation F1 Macro")
    plt.title("Stratified Group K-Fold: Train vs Validation F1 Macro")
    plt.xlabel("Fold"); plt.ylabel("F1 Macro"); plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

    fig, ax = plt.subplots(figsize=(10, 8))
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names).plot(ax=ax, xticks_rotation=45, colorbar=False)
    plt.title("Confusion Matrix (Stratified Group K-Fold)"); plt.tight_layout(); plt.show()

    # Model Final
    final_model = Pipeline([("prep", preprocessor), ("clf", RandomForestClassifier(**model_params))])
    final_model.fit(X, y_enc)

    # Feature Importance
    feat_names = final_model.named_steps["prep"].get_feature_names_out()
    imp = final_model.named_steps["clf"].feature_importances_
    imp_df = pd.DataFrame({"feature": feat_names, "importance": imp}).sort_values("importance", ascending=False)
    print("\n===== TOP 10 FEATURE IMPORTANCE =====")
    display(imp_df.head(10))

    # Simpan & Deployment Test
    save_model(final_model, le, target_col, X.columns.tolist())
    n_test = 5
    dep_df, dep_rate = simple_deployment_test(final_model, X, y, le, n_samples=n_test, random_state=42)
    print(f"\n===== PENGUJIAN DEPLOYMENT ({n_test} SAMPEL) =====")
    display(dep_df)
    print(f"Tingkat keberhasilan: {dep_rate:.2%}")

    # Interpretasi Logis
    print("\n===== INTERPRETASI METODOLOGIS =====")
    print(f"1. Teknik CV: Stratified Group K-Fold (mencegah leakage antar grup & menjaga proporsi kelas).")
    print(f"2. Imbalance Ratio: {imb_ratio:.2f}. Penanganan: class_weight='balanced_subsample'.")
    print(f"3. Train F1 Macro: {diagnostics['avg_train_f1_macro']:.4f} | Validation F1 Macro: {diagnostics['avg_test_f1_macro']:.4f}")
    print(f"4. Generalization Gap: {diagnostics['generalization_gap']:.4f} | Status: {diagnostics['fit_status']}")
    if diagnostics['fit_status'] == "Relatif seimbang":
        print("5. Logika: Gap kecil & skor stabil mengindikasikan model tidak overfit/underfit. Evaluasi realistis untuk deployment.")
    else:
        print("5. Logika: Gap besar atau skor rendah menunjukkan perlunya tuning kapasitas model atau redistribusi grup.")
    print(f"6. Deployment test ({n_test} sampel): {dep_rate:.2%}. Model siap untuk uji integrasi sistem lebih lanjut.")
```
