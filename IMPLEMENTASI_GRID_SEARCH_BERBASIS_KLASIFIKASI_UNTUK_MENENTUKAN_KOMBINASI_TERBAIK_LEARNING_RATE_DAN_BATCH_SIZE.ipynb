```python
# ================================================================
# JUDUL PROGRAM
# ================================================================
# IMPLEMENTASI GRID SEARCH BERBASIS KLASIFIKASI UNTUK MENENTUKAN
# KOMBINASI TERBAIK LEARNING RATE DAN BATCH SIZE PADA MODEL
# NEURAL NETWORK DENGAN PENGENDALIAN CLASS IMBALANCE,
# REGULARISASI, EARLY STOPPING, PENYIMPANAN MODEL,
# DAN UJI DEPLOYMENT SEDERHANA PADA DATASET AI4I 2020
#
# ================================================================
# DESKRIPSI SINGKAT
# ================================================================
# Program ini disusun untuk analisis predictive maintenance berbasis
# machine learning dengan target klasifikasi biner "Machine failure".
# Alur program mencakup:
# 1. Pemilihan sumber data secara fleksibel
# 2. Validasi konteks dataset agar konsisten dengan AI4I 2020
# 3. Audit kualitas data
# 4. Pra-pemrosesan yang aman
# 5. Penanganan ketidakseimbangan kelas
# 6. Grid search learning rate dan batch size
# 7. Penyusunan 3 rekomendasi perbaikan
# 8. Penerapan perbaikan otomatis secara logis
# 9. Evaluasi model akhir dengan multi-metrik
# 10. Penyimpanan model dan artefak
# 11. Simulasi deployment sederhana
#
# Catatan metodologis:
# - Model dipilih tidak hanya berdasarkan MSE, tetapi juga binary
#   cross-entropy, ROC-AUC, F1-score, precision, recall, dan stabilitas
#   generalisasi.
# - Konfigurasi grid search dibatasi secara rasional agar durasi proses
#   pada Colab yang umum cenderung berada pada kisaran 5–10 menit.
# - Program ini berhenti otomatis bila dataset yang dimuat tidak sesuai
#   dengan konteks AI4I predictive maintenance.
# ================================================================

# ================================================================
# 1. INSTALASI PUSTAKA
# ================================================================
!pip -q install ucimlrepo imbalanced-learn tensorflow scikit-learn seaborn joblib

# ================================================================
# 2. IMPOR PUSTAKA
# ================================================================
import os
import json
import time
import joblib
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    mean_squared_error, log_loss, roc_curve, precision_recall_curve
)

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

from ucimlrepo import fetch_ucirepo
from imblearn.over_sampling import SMOTE

# ================================================================
# 3. PENGATURAN AWAL
# ================================================================
np.random.seed(42)
tf.random.set_seed(42)
sns.set_theme(style="whitegrid", context="talk")
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 220)
plt.rcParams["figure.dpi"] = 120

print("="*100)
print("PROGRAM ANALISIS MACHINE LEARNING PREDICTIVE MAINTENANCE (AI4I 2020)")
print("="*100)

# ================================================================
# 4. FUNGSI BANTU
# ================================================================
def show_section(title):
    print("\n" + "="*100)
    print(title)
    print("="*100)

def robust_read_csv(path):
    try:
        return pd.read_csv(path)
    except Exception:
        try:
            return pd.read_csv(path, sep=";")
        except Exception as e:
            raise ValueError(f"Gagal membaca file CSV. Detail: {e}")

def sanitize_dataframe(df):
    df = df.copy()
    df = df.dropna(how="all")
    if df.columns.duplicated().sum() > 0:
        df = df.loc[:, ~df.columns.duplicated()]
    return df

def normalize_columns(df):
    return [str(c).strip().lower() for c in df.columns]

def detect_dataset_context(df):
    cols_lower = normalize_columns(df)

    ai4i_signals = [
        "air temperature", "process temperature",
        "rotational speed", "torque", "tool wear",
        "machine failure"
    ]
    ai4i_match = sum(1 for s in ai4i_signals if any(s in c for c in cols_lower))

    ecommerce_signals = [
        "event_type", "product_id", "category_id", "category_code",
        "brand", "price", "user_id", "user_session"
    ]
    ecommerce_match = sum(1 for s in ecommerce_signals if any(s in c for c in cols_lower))

    if ai4i_match >= 4:
        return "predictive_maintenance_ai4i"
    elif ecommerce_match >= 4:
        return "ecommerce"
    return "unknown"

def validate_required_features(df):
    cols_lower = normalize_columns(df)
    required = [
        "air temperature",
        "process temperature",
        "rotational speed",
        "torque",
        "tool wear"
    ]
    missing = [s for s in required if not any(s in c for c in cols_lower)]
    if len(missing) > 1:
        raise ValueError(f"Fitur utama AI4I tidak cukup. Sinyal yang tidak ditemukan: {missing}")

def find_target_column(df):
    candidates = [c for c in df.columns if "machine" in str(c).lower() and "failure" in str(c).lower()]
    if len(candidates) == 0:
        raise ValueError("Kolom target 'Machine failure' tidak ditemukan.")
    return candidates[0]

def validate_binary_target(y, target_name):
    unique_vals = sorted(pd.Series(y).dropna().unique().tolist())
    if len(unique_vals) != 2:
        raise ValueError(f"Kolom {target_name} tidak bersifat biner. Nilai unik: {unique_vals}")

def safe_clip_probs(probs):
    return np.clip(np.array(probs).ravel(), 1e-7, 1 - 1e-7)

def detect_fit_status(history):
    train_loss = history.history["loss"][-1]
    val_loss = history.history["val_loss"][-1]
    gap = val_loss - train_loss

    if train_loss > 0.45 and val_loss > 0.45:
        return "underfitting"
    elif gap > 0.08:
        return "overfitting"
    else:
        return "good_fit"

def evaluate_metrics(y_true, y_prob, threshold=0.5):
    y_true = np.array(y_true).astype(int)
    y_prob = safe_clip_probs(y_prob)
    y_pred = (y_prob >= threshold).astype(int)

    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1_score": f1_score(y_true, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_true, y_prob),
        "mse": mean_squared_error(y_true, y_prob),
        "binary_crossentropy": log_loss(y_true, y_prob, labels=[0, 1])
    }

def tune_threshold(y_true, y_prob):
    y_prob = safe_clip_probs(y_prob)
    thresholds = np.arange(0.10, 0.90, 0.02)
    best_thr = 0.50
    best_f1 = -1
    rows = []

    for thr in thresholds:
        y_pred = (y_prob >= thr).astype(int)
        p = precision_score(y_true, y_pred, zero_division=0)
        r = recall_score(y_true, y_pred, zero_division=0)
        f1 = f1_score(y_true, y_pred, zero_division=0)
        rows.append([thr, p, r, f1])
        if f1 > best_f1:
            best_f1 = f1
            best_thr = thr

    thr_df = pd.DataFrame(rows, columns=["threshold", "precision", "recall", "f1_score"])
    return best_thr, thr_df

def generalization_score(metrics_dict, fit_status):
    score = (
        0.35 * metrics_dict["roc_auc"] +
        0.30 * metrics_dict["f1_score"] +
        0.20 * metrics_dict["recall"] +
        0.10 * metrics_dict["precision"] -
        0.03 * metrics_dict["mse"] -
        0.02 * metrics_dict["binary_crossentropy"]
    )
    if fit_status in ["overfitting", "underfitting"]:
        score -= 0.05
    return score

def build_model(input_dim, learning_rate=0.001, dropout_rate=0.30, l2_lambda=1e-4, hidden_units=(64, 32)):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))
    model.add(Dense(hidden_units[0], activation="relu", kernel_regularizer=l2(l2_lambda)))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate))

    for units in hidden_units[1:]:
        model.add(Dense(units, activation="relu", kernel_regularizer=l2(l2_lambda)))
        model.add(BatchNormalization())
        model.add(Dropout(dropout_rate))

    model.add(Dense(1, activation="sigmoid"))

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=[
            tf.keras.metrics.BinaryAccuracy(name="accuracy"),
            tf.keras.metrics.AUC(name="auc"),
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.MeanSquaredError(name="mse")
        ]
    )
    return model

def load_dataset_flexible():
    print("\nPilih sumber data:")
    print("1 = Upload manual file CSV")
    print("2 = Gunakan file lokal di lingkungan Colab")
    print("3 = Ambil dataset AI4I 2020 dari UCI Repository")

    choice = input("Masukkan pilihan (1/2/3): ").strip()

    if choice == "1":
        from google.colab import files
        uploaded = files.upload()
        if len(uploaded) == 0:
            raise ValueError("Tidak ada file yang diunggah.")
        file_name = list(uploaded.keys())[0]
        df = robust_read_csv(file_name)
        print(f"\nDataset berhasil diunggah: {file_name}")
        return sanitize_dataframe(df)

    elif choice == "2":
        local_path = input("Masukkan path file CSV lokal: ").strip()
        if not os.path.exists(local_path):
            raise FileNotFoundError(f"File tidak ditemukan: {local_path}")
        df = robust_read_csv(local_path)
        print(f"\nDataset berhasil dibaca dari file lokal: {local_path}")
        return sanitize_dataframe(df)

    elif choice == "3":
        dataset = fetch_ucirepo(id=601)
        X_uci = dataset.data.features.copy()
        y_uci = dataset.data.targets.copy()
        df = pd.concat([X_uci, y_uci], axis=1)
        print("\nDataset berhasil diambil dari UCI Repository: AI4I 2020 Predictive Maintenance Dataset")
        return sanitize_dataframe(df)

    else:
        raise ValueError("Pilihan tidak valid. Gunakan 1, 2, atau 3.")

def save_artifacts(model, preprocessor, metadata, save_dir="saved_ai4i_model"):
    os.makedirs(save_dir, exist_ok=True)
    model_path = os.path.join(save_dir, "ai4i_nn_model.keras")
    preprocessor_path = os.path.join(save_dir, "preprocessor.joblib")
    metadata_path = os.path.join(save_dir, "metadata.json")

    model.save(model_path)
    joblib.dump(preprocessor, preprocessor_path)
    with open(metadata_path, "w") as f:
        json.dump(metadata, f, indent=4)

    return model_path, preprocessor_path, metadata_path

def load_saved_artifacts(model_path, preprocessor_path, metadata_path):
    loaded_model = load_model(model_path)
    loaded_preprocessor = joblib.load(preprocessor_path)
    with open(metadata_path, "r") as f:
        loaded_metadata = json.load(f)
    return loaded_model, loaded_preprocessor, loaded_metadata

def predict_single_input(input_df, loaded_model, loaded_preprocessor, loaded_metadata):
    required_num = loaded_metadata["numeric_columns"]
    required_cat = loaded_metadata["categorical_columns"]
    all_required = required_num + required_cat

    missing_cols = [c for c in all_required if c not in input_df.columns]
    if len(missing_cols) > 0:
        raise ValueError(f"Input deployment tidak lengkap. Kolom yang kurang: {missing_cols}")

    input_df = input_df[all_required].copy()

    for col in required_num:
        input_df[col] = pd.to_numeric(input_df[col], errors="coerce")
        if input_df[col].isnull().sum() > 0:
            raise ValueError(f"Kolom numerik {col} mengandung nilai tidak valid.")

    X_infer = loaded_preprocessor.transform(input_df)
    if hasattr(X_infer, "toarray"):
        X_infer = X_infer.toarray()

    probs = safe_clip_probs(loaded_model.predict(X_infer, verbose=0))
    threshold = loaded_metadata["best_threshold"]
    preds = (probs >= threshold).astype(int)

    output_df = input_df.copy()
    output_df["probabilitas_failure"] = probs
    output_df["prediksi_failure"] = preds
    return output_df

# ================================================================
# 5. MEMUAT DAN MEMVALIDASI DATASET
# ================================================================
show_section("MEMUAT DAN MEMVALIDASI DATASET")
df = load_dataset_flexible()

print("\nUkuran dataset:", df.shape)
display(df.head())

print("\nInformasi dataset:")
df.info()

print("\nMissing values per kolom:")
display(df.isnull().sum().to_frame("Jumlah Missing"))

dataset_context = detect_dataset_context(df)
print("\nKonteks dataset terdeteksi:", dataset_context)

if dataset_context != "predictive_maintenance_ai4i":
    raise ValueError("Dataset tidak sesuai dengan konteks AI4I predictive maintenance. Program dihentikan.")

validate_required_features(df)

# ================================================================
# 6. IDENTIFIKASI TARGET DAN AUDIT DATA
# ================================================================
show_section("IDENTIFIKASI TARGET DAN AUDIT DATA")

target_col = find_target_column(df)
y = df[target_col].astype(int)
validate_binary_target(y, target_col)
X = df.drop(columns=[target_col]).copy()

print("Kolom target:", target_col)
print("Ukuran fitur:", X.shape)
print("Ukuran target:", y.shape)

duplicate_rows = X.duplicated().sum()
print("Jumlah baris duplikat:", duplicate_rows)

if duplicate_rows > 0:
    print("Baris duplikat dihapus untuk menjaga kualitas data.")
    tmp_df = X.copy()
    tmp_df["_target_temp_"] = y.values
    tmp_df = tmp_df.drop_duplicates()
    y = tmp_df["_target_temp_"].astype(int)
    X = tmp_df.drop(columns=["_target_temp_"]).copy()

class_count = y.value_counts().sort_index()
class_ratio = y.value_counts(normalize=True).sort_index()
summary_target_df = pd.DataFrame({"Jumlah": class_count, "Proporsi": class_ratio})
display(summary_target_df)

plt.figure(figsize=(7,4.5))
sns.countplot(x=y)
plt.title("Distribusi Kelas Target: Machine Failure")
plt.xlabel("Machine Failure (0 = Tidak Gagal, 1 = Gagal)")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()

imbalance_ratio = class_count.min() / class_count.max()
print(f"Rasio ketidakseimbangan kelas (minority/majority): {imbalance_ratio:.4f}")

# ================================================================
# 7. PEMBERSIHAN FITUR DAN PRA-PEMROSESAN
# ================================================================
show_section("PEMBERSIHAN FITUR DAN PRA-PEMROSESAN")

drop_candidates = []
for col in X.columns:
    c = str(col).lower().strip()
    if c in ["udi", "product id"] or c == "id":
        drop_candidates.append(col)

X_model = X.drop(columns=drop_candidates, errors="ignore").copy()

categorical_cols = [col for col in X_model.columns if X_model[col].dtype == "object"]
numeric_cols = [col for col in X_model.columns if col not in categorical_cols]

print("Kolom identitas yang dihapus:", drop_candidates)
print("Kolom kategorik:", categorical_cols)
print("Kolom numerik:", numeric_cols)

if len(numeric_cols) == 0:
    raise ValueError("Tidak ada fitur numerik. Pipeline menjadi tidak valid.")

if X_model.isnull().sum().sum() > 0:
    print("Ditemukan missing values. Program melakukan imputasi sederhana.")
    for col in numeric_cols:
        X_model[col] = X_model[col].fillna(X_model[col].median())
    for col in categorical_cols:
        mode_val = X_model[col].mode(dropna=True)
        fill_val = mode_val.iloc[0] if len(mode_val) > 0 else "Unknown"
        X_model[col] = X_model[col].fillna(fill_val)

desc_df = X_model[numeric_cols].describe().T
display(desc_df)

plt.figure(figsize=(10,6))
corr = X_model[numeric_cols].corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Koefisien Korelasi"})
plt.title("Heatmap Korelasi Fitur Numerik")
plt.tight_layout()
plt.show()

# ================================================================
# 8. PEMBAGIAN DATA
# ================================================================
show_section("PEMBAGIAN DATA")

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_model, y, test_size=0.15, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.1765, stratify=y_train_full, random_state=42
)

split_df = pd.DataFrame({
    "Subset": ["Train", "Validation", "Test"],
    "Jumlah Data": [len(X_train), len(X_val), len(X_test)],
    "Proporsi": [len(X_train)/len(X_model), len(X_val)/len(X_model), len(X_test)/len(X_model)]
})
display(split_df)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ],
    remainder="drop"
)

X_train_proc = preprocessor.fit_transform(X_train)
X_val_proc = preprocessor.transform(X_val)
X_test_proc = preprocessor.transform(X_test)

if hasattr(X_train_proc, "toarray"):
    X_train_proc = X_train_proc.toarray()
    X_val_proc = X_val_proc.toarray()
    X_test_proc = X_test_proc.toarray()

input_dim = X_train_proc.shape[1]
if input_dim == 0:
    raise ValueError("Dimensi input setelah preprocessing bernilai 0.")

print("Dimensi input setelah preprocessing:", input_dim)

# ================================================================
# 9. PENANGANAN CLASS IMBALANCE
# ================================================================
show_section("PENANGANAN CLASS IMBALANCE")

use_smote = imbalance_ratio < 0.50
print("Apakah SMOTE digunakan?:", use_smote)

if use_smote:
    minority_count = y_train.value_counts().min()
    if minority_count < 6:
        print("Jumlah kelas minoritas terlalu kecil untuk SMOTE yang stabil. SMOTE dinonaktifkan.")
        X_train_bal, y_train_bal = X_train_proc, y_train
    else:
        smote = SMOTE(random_state=42, k_neighbors=min(5, minority_count - 1))
        X_train_bal, y_train_bal = smote.fit_resample(X_train_proc, y_train)
else:
    X_train_bal, y_train_bal = X_train_proc, y_train

classes = np.unique(y_train_bal)
class_weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train_bal)
class_weight_dict = {int(c): float(w) for c, w in zip(classes, class_weights)}
print("Class weight:", class_weight_dict)

plt.figure(figsize=(7,4.5))
sns.countplot(x=y_train_bal)
plt.title("Distribusi Kelas Setelah Penanganan Imbalance (Data Train)")
plt.xlabel("Machine Failure")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()

# ================================================================
# 10. GRID SEARCH DASAR
# ================================================================
show_section("GRID SEARCH DASAR")

# Grid dibatasi secara rasional agar durasi cenderung berada pada kisaran 5–10 menit.
learning_rates = [5e-4, 1e-3, 5e-3]
batch_sizes = [32, 64, 128]
max_epochs_grid = 60

grid_results = []
threshold_histories = {}
fit_status_counts = {"good_fit": 0, "overfitting": 0, "underfitting": 0}

start_time = time.time()

for lr in learning_rates:
    for bs in batch_sizes:
        try:
            print(f"Melatih model dengan learning_rate={lr} dan batch_size={bs}")

            base_model = build_model(
                input_dim=input_dim,
                learning_rate=lr,
                dropout_rate=0.30,
                l2_lambda=1e-4,
                hidden_units=(64, 32)
            )

            callbacks = [
                EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True),
                ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6)
            ]

            history = base_model.fit(
                X_train_bal, y_train_bal,
                validation_data=(X_val_proc, y_val),
                epochs=max_epochs_grid,
                batch_size=bs,
                class_weight=class_weight_dict,
                callbacks=callbacks,
                verbose=0
            )

            y_val_prob = base_model.predict(X_val_proc, verbose=0).ravel()
            best_thr, thr_df = tune_threshold(y_val, y_val_prob)
            metrics_val = evaluate_metrics(y_val, y_val_prob, threshold=best_thr)
            fit_status = detect_fit_status(history)
            fit_status_counts[fit_status] += 1
            g_score = generalization_score(metrics_val, fit_status)

            grid_results.append({
                "learning_rate": lr,
                "batch_size": bs,
                "best_threshold": best_thr,
                "val_accuracy": metrics_val["accuracy"],
                "val_precision": metrics_val["precision"],
                "val_recall": metrics_val["recall"],
                "val_f1_score": metrics_val["f1_score"],
                "val_roc_auc": metrics_val["roc_auc"],
                "val_mse": metrics_val["mse"],
                "val_binary_crossentropy": metrics_val["binary_crossentropy"],
                "fit_status": fit_status,
                "epochs_ran": len(history.history["loss"]),
                "final_train_loss": history.history["loss"][-1],
                "final_val_loss": history.history["val_loss"][-1],
                "generalization_score": g_score
            })

            threshold_histories[(lr, bs)] = thr_df

        except Exception as e:
            print(f"Konfigurasi lr={lr}, bs={bs} gagal. Detail: {e}")

if len(grid_results) == 0:
    raise ValueError("Seluruh konfigurasi grid search gagal.")

results_df = pd.DataFrame(grid_results).sort_values(by="generalization_score", ascending=False).reset_index(drop=True)
display(results_df)

# ================================================================
# 11. VISUALISASI GRID SEARCH
# ================================================================
show_section("VISUALISASI HASIL GRID SEARCH")

for metric_name, cmap, title in [
    ("val_mse", "YlGnBu", "Heatmap Validation MSE"),
    ("val_binary_crossentropy", "Oranges", "Heatmap Validation Binary Cross-Entropy"),
    ("val_roc_auc", "Reds", "Heatmap Validation ROC-AUC"),
    ("val_f1_score", "Greens", "Heatmap Validation F1-Score"),
    ("generalization_score", "Purples", "Heatmap Generalization Score")
]:
    pivot_df = results_df.pivot(index="batch_size", columns="learning_rate", values=metric_name)
    plt.figure(figsize=(8,5))
    sns.heatmap(pivot_df, annot=True, fmt=".4f", cmap=cmap, cbar_kws={"label": metric_name})
    plt.title(title)
    plt.xlabel("Learning Rate")
    plt.ylabel("Batch Size")
    plt.tight_layout()
    plt.show()

top_configs = results_df.head(min(5, len(results_df))).copy()
top_configs["Konfigurasi"] = top_configs["learning_rate"].astype(str) + " | " + top_configs["batch_size"].astype(str)

plt.figure(figsize=(10,5))
sns.barplot(data=top_configs, x="Konfigurasi", y="generalization_score")
plt.title("Lima Konfigurasi Terbaik Berdasarkan Generalization Score")
plt.xlabel("Learning Rate | Batch Size")
plt.ylabel("Generalization Score")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

# ================================================================
# 12. TIGA REKOMENDASI PERBAIKAN
# ================================================================
show_section("TIGA REKOMENDASI PERBAIKAN")

recommendations = []
max_recall = results_df["val_recall"].max()
has_overfitting = fit_status_counts["overfitting"] > 0
has_underfitting = fit_status_counts["underfitting"] > 0

if max_recall < 0.75:
    recommendations.append("Meningkatkan sensitivitas model terhadap kelas failure melalui threshold tuning yang ketat dan penguatan penanganan class imbalance.")
else:
    recommendations.append("Mempertahankan sensitivitas deteksi failure dengan threshold tuning berbasis validation set agar model tidak bias pada threshold default.")

if has_overfitting:
    recommendations.append("Memperkuat regularisasi model melalui peningkatan dropout, penyesuaian L2 regularization, dan pengendalian pelatihan dengan early stopping.")
else:
    recommendations.append("Mempertahankan regularisasi yang ada sambil terus memonitor validation loss agar kestabilan generalisasi tetap terjaga.")

if has_underfitting:
    recommendations.append("Meningkatkan kapasitas representasi model secara moderat dengan penambahan unit terseleksi agar pola kegagalan lebih tertangkap.")
else:
    recommendations.append("Mempertahankan kapasitas model yang efisien agar kompleksitas tidak berlebihan dan risiko generalisasi buruk tetap terkendali.")

recommendation_df = pd.DataFrame({
    "No": [1, 2, 3],
    "Rekomendasi Perbaikan": recommendations
})
display(recommendation_df)

# ================================================================
# 13. PENERAPAN PERBAIKAN OTOMATIS SECARA LOGIS
# ================================================================
show_section("PENERAPAN PERBAIKAN OTOMATIS")

best_row = results_df.iloc[0]
best_lr = float(best_row["learning_rate"])
best_bs = int(best_row["batch_size"])

# Perbaikan otomatis 1: threshold tuning tetap digunakan
# Perbaikan otomatis 2: regularisasi diperkuat bila ada overfitting
if has_overfitting:
    final_dropout = 0.40
    final_l2 = 5e-4
else:
    final_dropout = 0.30
    final_l2 = 1e-4

# Perbaikan otomatis 3: kapasitas model dinaikkan moderat bila ada underfitting
if has_underfitting:
    final_hidden_units = (128, 64, 32)
else:
    final_hidden_units = (64, 32)

auto_fix_df = pd.DataFrame({
    "Komponen": ["Learning Rate", "Batch Size", "Dropout", "L2 Regularization", "Hidden Units"],
    "Nilai Final": [best_lr, best_bs, final_dropout, final_l2, str(final_hidden_units)]
})
display(auto_fix_df)

# ================================================================
# 14. PELATIHAN MODEL AKHIR
# ================================================================
show_section("PELATIHAN MODEL AKHIR")

# Epoch model akhir dibuat moderat agar kualitas lebih stabil,
# namun tetap menjaga durasi total pada kisaran yang rasional.
final_epochs = 90

final_model = build_model(
    input_dim=input_dim,
    learning_rate=best_lr,
    dropout_rate=final_dropout,
    l2_lambda=final_l2,
    hidden_units=final_hidden_units
)

final_callbacks = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6)
]

history_final = final_model.fit(
    X_train_bal, y_train_bal,
    validation_data=(X_val_proc, y_val),
    epochs=final_epochs,
    batch_size=best_bs,
    class_weight=class_weight_dict,
    callbacks=final_callbacks,
    verbose=0
)

y_val_prob_final = final_model.predict(X_val_proc, verbose=0).ravel()
best_thr_final, thr_df_final = tune_threshold(y_val, y_val_prob_final)
val_metrics_final = evaluate_metrics(y_val, y_val_prob_final, threshold=best_thr_final)
fit_status_final = detect_fit_status(history_final)
gs_final = generalization_score(val_metrics_final, fit_status_final)

final_summary_df = pd.DataFrame({
    "Metrik": list(val_metrics_final.keys()) + ["fit_status", "generalization_score", "best_threshold"],
    "Nilai": list(val_metrics_final.values()) + [fit_status_final, gs_final, best_thr_final]
})
display(final_summary_df)

# ================================================================
# 15. EVALUASI PADA TEST SET
# ================================================================
show_section("EVALUASI PADA TEST SET")

y_test_prob = final_model.predict(X_test_proc, verbose=0).ravel()
y_test_pred = (safe_clip_probs(y_test_prob) >= best_thr_final).astype(int)
test_metrics = evaluate_metrics(y_test, y_test_prob, threshold=best_thr_final)

test_metrics_df = pd.DataFrame({
    "Metrik": list(test_metrics.keys()),
    "Nilai": list(test_metrics.values())
})
display(test_metrics_df)

print("\nClassification report:")
print(classification_report(y_test, y_test_pred, digits=4))

# ================================================================
# 16. TABEL PERBANDINGAN MODEL
# ================================================================
show_section("TABEL PERBANDINGAN MODEL")

comparison_df = pd.DataFrame([
    {
        "Model": "Best Grid Search Model",
        "Threshold": float(best_row["best_threshold"]),
        "Validation Accuracy": best_row["val_accuracy"],
        "Validation Precision": best_row["val_precision"],
        "Validation Recall": best_row["val_recall"],
        "Validation F1-Score": best_row["val_f1_score"],
        "Validation ROC-AUC": best_row["val_roc_auc"],
        "Validation MSE": best_row["val_mse"],
        "Validation Binary Cross-Entropy": best_row["val_binary_crossentropy"],
        "Fit Status": best_row["fit_status"],
        "Generalization Score": best_row["generalization_score"]
    },
    {
        "Model": "Final Model",
        "Threshold": float(best_thr_final),
        "Validation Accuracy": val_metrics_final["accuracy"],
        "Validation Precision": val_metrics_final["precision"],
        "Validation Recall": val_metrics_final["recall"],
        "Validation F1-Score": val_metrics_final["f1_score"],
        "Validation ROC-AUC": val_metrics_final["roc_auc"],
        "Validation MSE": val_metrics_final["mse"],
        "Validation Binary Cross-Entropy": val_metrics_final["binary_crossentropy"],
        "Fit Status": fit_status_final,
        "Generalization Score": gs_final
    }
])
display(comparison_df)

# ================================================================
# 17. VISUALISASI MODEL AKHIR
# ================================================================
show_section("VISUALISASI MODEL AKHIR")

plt.figure(figsize=(8,4.5))
plt.plot(history_final.history["loss"], label="Train Loss")
plt.plot(history_final.history["val_loss"], label="Validation Loss")
plt.title("Training vs Validation Loss - Final Model")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4.5))
plt.plot(history_final.history["auc"], label="Train AUC")
plt.plot(history_final.history["val_auc"], label="Validation AUC")
plt.title("Training vs Validation AUC - Final Model")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4.5))
plt.plot(thr_df_final["threshold"], thr_df_final["precision"], label="Precision")
plt.plot(thr_df_final["threshold"], thr_df_final["recall"], label="Recall")
plt.plot(thr_df_final["threshold"], thr_df_final["f1_score"], label="F1-Score")
plt.axvline(best_thr_final, linestyle="--", label=f"Best Threshold = {best_thr_final:.2f}")
plt.title("Threshold Tuning pada Validation Set")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

fpr, tpr, _ = roc_curve(y_test, safe_clip_probs(y_test_prob))
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {test_metrics['roc_auc']:.4f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("ROC Curve - Test Set")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.show()

precision_curve, recall_curve, _ = precision_recall_curve(y_test, safe_clip_probs(y_test_prob))
plt.figure(figsize=(6,5))
plt.plot(recall_curve, precision_curve)
plt.title("Precision-Recall Curve - Test Set")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.tight_layout()
plt.show()

cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(5.8,4.8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix - Final Model")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()

# ================================================================
# 18. PENYIMPANAN MODEL DAN ARTEFAK
# ================================================================
show_section("PENYIMPANAN MODEL DAN ARTEFAK")

metadata = {
    "target_column": target_col,
    "numeric_columns": numeric_cols,
    "categorical_columns": categorical_cols,
    "best_threshold": float(best_thr_final),
    "learning_rate": float(best_lr),
    "batch_size": int(best_bs),
    "dropout_rate": float(final_dropout),
    "l2_regularization": float(final_l2),
    "hidden_units": list(final_hidden_units),
    "test_metrics": {k: float(v) for k, v in test_metrics.items()}
}

model_path, preprocessor_path, metadata_path = save_artifacts(
    final_model, preprocessor, metadata, save_dir="saved_ai4i_model"
)

artifact_df = pd.DataFrame({
    "Artefak": ["Model", "Preprocessor", "Metadata"],
    "Lokasi": [model_path, preprocessor_path, metadata_path]
})
display(artifact_df)

# ================================================================
# 19. SIMULASI DEPLOYMENT SEDERHANA
# ================================================================
show_section("SIMULASI DEPLOYMENT SEDERHANA")

loaded_model, loaded_preprocessor, loaded_metadata = load_saved_artifacts(
    model_path, preprocessor_path, metadata_path
)

n_test_samples = min(5, len(X_test))
simple_test_samples = X_test.iloc[:n_test_samples].copy()
simple_test_labels = y_test.iloc[:n_test_samples].copy()

deployment_result = predict_single_input(
    simple_test_samples,
    loaded_model,
    loaded_preprocessor,
    loaded_metadata
)

deployment_result["actual_machine_failure"] = simple_test_labels.values
deployment_result["status_kecocokan"] = np.where(
    deployment_result["prediksi_failure"] == deployment_result["actual_machine_failure"],
    "Benar",
    "Salah"
)

display(deployment_result)

simple_test_accuracy = (deployment_result["prediksi_failure"] == deployment_result["actual_machine_failure"]).mean()
print(f"Akurasi sederhana pada 5 data uji deployment: {simple_test_accuracy:.4f}")

plt.figure(figsize=(8,4.5))
sns.barplot(
    x=[f"Data-{i+1}" for i in range(len(deployment_result))],
    y=deployment_result["probabilitas_failure"]
)
plt.axhline(loaded_metadata["best_threshold"], linestyle="--", label=f"Threshold = {loaded_metadata['best_threshold']:.2f}")
plt.title("Probabilitas Failure pada 5 Data Uji Deployment")
plt.xlabel("Data Uji")
plt.ylabel("Probabilitas Failure")
plt.legend()
plt.tight_layout()
plt.show()

# ================================================================
# 20. RINGKASAN EKSEKUTIF HASIL
# ================================================================
show_section("RINGKASAN EKSEKUTIF HASIL")

elapsed_minutes = (time.time() - start_time) / 60

executive_summary_df = pd.DataFrame({
    "Aspek": [
        "Model Terbaik Grid Search",
        "Model Akhir",
        "Threshold Akhir",
        "Status Fit Akhir",
        "ROC-AUC Test",
        "F1-Score Test",
        "Recall Test",
        "Akurasi Uji Deployment Sederhana",
        "Durasi Proses (menit)"
    ],
    "Nilai": [
        f"lr={best_lr}, bs={best_bs}",
        str(final_hidden_units),
        round(best_thr_final, 4),
        fit_status_final,
        round(test_metrics["roc_auc"], 4),
        round(test_metrics["f1_score"], 4),
        round(test_metrics["recall"], 4),
        round(simple_test_accuracy, 4),
        round(elapsed_minutes, 2)
    ]
})
display(executive_summary_df)

# ================================================================
# 21. INTERPRETASI HASIL
# ================================================================
show_section("INTERPRETASI HASIL SECARA LOGIS")

if comparison_df.loc[1, "Validation F1-Score"] >= comparison_df.loc[0, "Validation F1-Score"]:
    print("- Model akhir mempertahankan atau meningkatkan F1-score, sehingga keseimbangan precision dan recall menjadi lebih baik atau tetap stabil.")
else:
    print("- Model akhir belum meningkatkan F1-score, sehingga rekayasa fitur atau penyesuaian arsitektur masih dapat dipertimbangkan.")

if comparison_df.loc[1, "Validation ROC-AUC"] >= comparison_df.loc[0, "Validation ROC-AUC"]:
    print("- ROC-AUC model akhir mempertahankan atau meningkatkan kemampuan pemisahan antara kelas gagal dan tidak gagal.")
else:
    print("- ROC-AUC model akhir menurun, sehingga kemampuan generalisasi perlu ditinjau kembali.")

if fit_status_final == "good_fit":
    print("- Kurva train dan validation menunjukkan kondisi pembelajaran yang relatif seimbang, sehingga risiko underfitting dan overfitting lebih terkendali.")
elif fit_status_final == "overfitting":
    print("- Masih terdapat indikasi overfitting, sehingga penguatan regularisasi atau penyederhanaan model dapat dipertimbangkan.")
else:
    print("- Masih terdapat indikasi underfitting, sehingga kapasitas model atau kualitas fitur dapat ditingkatkan.")

print("\nInterpretasi pengujian deployment sederhana:")
if simple_test_accuracy >= 0.80:
    print("- Pada 5 data uji sederhana, model menunjukkan tingkat keberhasilan yang baik sebagai validasi awal inferensi.")
elif simple_test_accuracy >= 0.60:
    print("- Pada 5 data uji sederhana, model menunjukkan keberhasilan yang cukup, namun pengujian tambahan tetap diperlukan.")
else:
    print("- Pada 5 data uji sederhana, keberhasilan model masih rendah, sehingga implementasi lanjutan perlu dilakukan dengan hati-hati.")

print(f"\nDurasi proses tercatat sekitar {elapsed_minutes:.2f} menit.")
print("Dengan konfigurasi grid dan epoch yang dibatasi secara rasional, proses ini dirancang agar pada runtime Colab yang umum cenderung berada pada kisaran 5–10 menit.")

print("\nKesimpulan akhir:")
print("- Model dipilih melalui evaluasi multi-metrik, bukan hanya MSE.")
print("- Tiga rekomendasi perbaikan telah dihasilkan dan diterapkan secara otomatis berdasarkan hasil evaluasi awal.")
print("- Model, preprocessor, dan metadata telah disimpan untuk mendukung tahap penggunaan lanjutan.")
```
