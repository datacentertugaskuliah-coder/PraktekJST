```python
# =============================================================================
# PROGRAM KLASIFIKASI KUAT TEKAN BETON MENGGUNAKAN JARINGAN SARAF TIRUAN
# -----------------------------------------------------------------------------
# Deskripsi:
# Program ini mengimplementasikan Jaringan Saraf Tiruan (JST) dengan propagasi maju
# (forward propagation) untuk memprediksi kategori kuat tekan beton
# (rendah, sedang, tinggi) berdasarkan 8 fitur komposisi bahan dan umur beton.
# Dataset diperoleh melalui dua opsi: unggah manual file Excel atau unduh otomatis
# dari UCI Repository (ID=165). Evaluasi menggunakan akurasi, presisi, recall,
# dan F1-score. Program dirancang dengan penanganan error, pencegahan overfitting
# (regularisasi L2, dropout, early stopping), dan simulasi deployment sederhana.
#
# Fitur Utama:
# - Dua metode pemuatan data (manual/otomatis) dengan validasi.
# - Eksplorasi data dan visualisasi distribusi.
# - Diskritisasi target menjadi 3 kelas seimbang (persentil).
# - Arsitektur JST dengan regularisasi, dropout, dan early stopping.
# - Evaluasi lengkap dan confusion matrix.
# - Penyimpanan model, scaler, dan label encoder (format .h5, .save).
# - Simulasi prediksi pada 5 sampel uji dan 5 data buatan manual.
#
# Tiga Rekomendasi Perbaikan yang Telah Diimplementasikan:
# 1. Pencegahan Overfitting: Regularisasi L2 pada layer dense dan dropout (0.3)
#    untuk mengurangi kompleksitas model dan meningkatkan generalisasi.
# 2. Optimasi Pelatihan: EarlyStopping (patience=20) dan ReduceLROnPlateau
#    untuk menghentikan pelatihan saat validasi tidak membaik dan menyesuaikan
#    learning rate secara adaptif.
# 3. Standarisasi Fitur: Semua fitur input diskalakan menggunakan StandardScaler
#    agar memiliki mean=0 dan varians=1, mempercepat konvergensi dan mencegah
#    dominasi fitur berskala besar.
#
# Catatan: Program ini dirancang untuk dieksekusi di Google Colab.
# =============================================================================

# -----------------------------------------------------------------------------
# 1. Instalasi dan Impor Pustaka
# -----------------------------------------------------------------------------
!pip install -q scikit-learn matplotlib seaborn pandas numpy tensorflow joblib openpyxl ucimlrepo

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, classification_report, confusion_matrix)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import joblib
import io
from google.colab import files
from ucimlrepo import fetch_ucirepo

# -----------------------------------------------------------------------------
# 2. Pemuatan Dataset dengan Dua Opsi dan Penanganan Error
# -----------------------------------------------------------------------------
print("Pilih metode pemuatan dataset:")
print("1. Unggah file Excel manual (Concrete_Data.xls)")
print("2. Unduh otomatis dari UCI Repository (ucimlrepo)")
pilihan = input("Masukkan pilihan (1/2): ").strip()

if pilihan == '1':
    # Opsi 1: Unggah manual
    print("\nSilakan unggah file Concrete_Data.xls")
    uploaded = files.upload()
    if len(uploaded) == 0:
        raise ValueError("Tidak ada file yang diunggah. Program dihentikan.")
    filename = list(uploaded.keys())[0]
    print(f"File yang diunggah: {filename}")
    try:
        # Baca file Excel tanpa header (asumsi data mentah)
        df = pd.read_excel(io.BytesIO(uploaded[filename]), header=None, sheet_name=0)
    except Exception as e:
        raise ValueError(f"Gagal membaca file. Pastikan format file benar (xls/xlsx). Error: {e}")
    if df.shape[1] != 9:
        raise ValueError(f"Jumlah kolom tidak sesuai (diharapkan 9, ditemukan {df.shape[1]}).")
    # Beri nama kolom
    df.columns = ['Cement', 'BlastFurnaceSlag', 'FlyAsh', 'Water',
                  'Superplasticizer', 'CoarseAggregate', 'FineAggregate',
                  'Age', 'Strength']
    print("\nDataset berhasil dimuat melalui unggahan manual.")

elif pilihan == '2':
    # Opsi 2: Unduh dari UCI
    print("\nMengunduh dataset Concrete Compressive Strength dari UCI...")
    try:
        concrete = fetch_ucirepo(id=165)
        # Tampilkan metadata dan informasi variabel (opsional)
        print("\nMETADATA:")
        print(concrete.metadata)
        print("\nINFORMASI VARIABEL:")
        print(concrete.variables)
        X = concrete.data.features
        y = concrete.data.targets
        df = pd.concat([X, y], axis=1)
        df.rename(columns={'Concrete compressive strength': 'Strength'}, inplace=True)
        print("\nDataset berhasil dimuat dari UCI Repository.")
    except Exception as e:
        raise RuntimeError(f"Gagal mengunduh dari UCI. Periksa koneksi internet. Error: {e}")

else:
    raise ValueError("Pilihan tidak valid. Harap masukkan 1 atau 2.")

# Tampilkan informasi dasar dataset
print("\nDimensi data:", df.shape)
print("\nLima baris pertama:")
print(df.head())
print("\nInformasi dataset:")
print(df.info())
print("\nStatistik deskriptif:")
print(df.describe())

# -----------------------------------------------------------------------------
# 3. Eksplorasi Data dan Deteksi Anomali
# -----------------------------------------------------------------------------
# Periksa nilai hilang
print("\nJumlah nilai hilang per kolom:")
print(df.isnull().sum())
if df.isnull().sum().sum() > 0:
    print("Terdeteksi nilai hilang. Melakukan penghapusan baris dengan nilai hilang...")
    df.dropna(inplace=True)
    print(f"Ukuran dataset setelah penghapusan: {df.shape}")

# Deteksi outlier (hanya peringatan, tidak dihapus)
Q1 = df['Strength'].quantile(0.25)
Q3 = df['Strength'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Strength'] < (Q1 - 1.5 * IQR)) | (df['Strength'] > (Q3 + 1.5 * IQR))]
if len(outliers) > 0:
    print(f"Terdeteksi {len(outliers)} outlier pada target. (Tetap dipertahankan untuk keutuhan data)")

# Visualisasi distribusi target asli
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(df['Strength'], bins=30, kde=True)
plt.title('Distribusi Kuat Tekan Beton (MPa)')
plt.subplot(1,2,2)
sns.boxplot(x=df['Strength'])
plt.title('Boxplot Kuat Tekan Beton')
plt.tight_layout()
plt.show()

# -----------------------------------------------------------------------------
# 4. Diskritisasi Target Menjadi Tiga Kategori (Seimbang)
# -----------------------------------------------------------------------------
percentiles = [0, 33.3, 66.7, 100]
bin_edges = np.percentile(df['Strength'], percentiles)
# Sesuaikan tepi agar mencakup semua data
bin_edges[0] -= 0.001
bin_edges[-1] += 0.001
labels = ['Rendah', 'Sedang', 'Tinggi']
df['Kategori'] = pd.cut(df['Strength'], bins=bin_edges, labels=labels, include_lowest=True)

print("\nDistribusi kategori setelah diskritisasi:")
print(df['Kategori'].value_counts())

# Visualisasi distribusi kategori
sns.countplot(x='Kategori', data=df)
plt.title('Jumlah Sampel per Kategori')
plt.show()

# -----------------------------------------------------------------------------
# 5. Persiapan Data untuk Pemodelan
# -----------------------------------------------------------------------------
X = df.drop(['Strength', 'Kategori'], axis=1)
y = df['Kategori']

# Encode label kategori menjadi angka
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print("\nMapping label ke angka:", dict(zip(le.classes_, le.transform(le.classes_))))

# Stratified split (menjaga proporsi kelas)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Standarisasi fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nUkuran data training : {X_train_scaled.shape}")
print(f"Ukuran data testing  : {X_test_scaled.shape}")

# -----------------------------------------------------------------------------
# 6. Membangun Arsitektur JST dengan Regularisasi dan Dropout
# -----------------------------------------------------------------------------
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=[X_train.shape[1]],
                 kernel_regularizer=regularizers.l2(0.001),
                 name='hidden_1'),
    layers.Dropout(0.3, name='dropout_1'),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001),
                 name='hidden_2'),
    layers.Dropout(0.3, name='dropout_2'),
    layers.Dense(32, activation='relu', name='hidden_3'),
    layers.Dense(3, activation='softmax', name='output')
])

model.summary()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callback untuk mencegah overfitting dan mengatur learning rate
early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001)

# -----------------------------------------------------------------------------
# 7. Pelatihan Model
# -----------------------------------------------------------------------------
history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=32,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# -----------------------------------------------------------------------------
# 8. Visualisasi Kurva Pelatihan
# -----------------------------------------------------------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss selama pelatihan')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Akurasi selama pelatihan')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------------------------------------------------------
# 9. Evaluasi Model pada Data Testing
# -----------------------------------------------------------------------------
y_pred_prob = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_prob, axis=1)

accuracy = accuracy_score(y_test, y_pred)
precision_macro = precision_score(y_test, y_pred, average='macro')
recall_macro = recall_score(y_test, y_pred, average='macro')
f1_macro = f1_score(y_test, y_pred, average='macro')
precision_weighted = precision_score(y_test, y_pred, average='weighted')
recall_weighted = recall_score(y_test, y_pred, average='weighted')
f1_weighted = f1_score(y_test, y_pred, average='weighted')

print("\n" + "="*50)
print("          HASIL EVALUASI MODEL PADA DATA TESTING")
print("="*50)
print(f"Akurasi              : {accuracy:.4f}")
print(f"Presisi (macro)      : {precision_macro:.4f}")
print(f"Recall (macro)       : {recall_macro:.4f}")
print(f"F1-score (macro)     : {f1_macro:.4f}")
print(f"Presisi (weighted)   : {precision_weighted:.4f}")
print(f"Recall (weighted)    : {recall_weighted:.4f}")
print(f"F1-score (weighted)  : {f1_weighted:.4f}")

print("\nLaporan klasifikasi per kelas:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# -----------------------------------------------------------------------------
# 10. Visualisasi Confusion Matrix
# -----------------------------------------------------------------------------
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Kelas Prediksi')
plt.ylabel('Kelas Aktual')
plt.title('Confusion Matrix')
plt.show()

# -----------------------------------------------------------------------------
# 11. Penyimpanan Model, Scaler, dan Encoder
# -----------------------------------------------------------------------------
model.save('model_kuat_tekan_beton.h5')
joblib.dump(scaler, 'scaler.save')
joblib.dump(le, 'label_encoder.save')
print("\nModel, scaler, dan encoder berhasil disimpan.")

# -----------------------------------------------------------------------------
# 12. Simulasi Deployment: Prediksi 5 Sampel Acak dari Data Uji
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("   SIMULASI DEPLOYMENT: PREDIKSI PADA 5 DATA UJI")
print("="*50)

n_samples = 5
sample_indices = np.random.choice(len(X_test), n_samples, replace=False)
X_sample = X_test.iloc[sample_indices]
y_actual = y_test[sample_indices]
y_actual_label = le.inverse_transform(y_actual)

X_sample_scaled = scaler.transform(X_sample)
pred_prob = model.predict(X_sample_scaled)
pred_class = np.argmax(pred_prob, axis=1)
pred_label = le.inverse_transform(pred_class)

print("\n{:<5} {:<70} {:<10} {:<10} {:<20}".format(
    "No", "Fitur (C, BFS, FA, W, SP, CA, FAge, Age)", "Aktual", "Prediksi", "Probabilitas (%)"))
print("-" * 120)

for i in range(n_samples):
    fitur_str = ", ".join([f"{val:.1f}" for val in X_sample.iloc[i]])
    prob_str = ", ".join([f"{p:.2f}" for p in pred_prob[i]])
    print("{:<5} {:<70} {:<10} {:<10} {:<20}".format(
        i+1, fitur_str, y_actual_label[i], pred_label[i], prob_str))

acc_sample = np.mean(pred_class == y_actual)
print("-" * 120)
print(f"Akurasi pada {n_samples} sampel: {acc_sample:.2f}")

# -----------------------------------------------------------------------------
# 13. Simulasi Input Manual (Contoh Data Baru)
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("   SIMULASI INPUT MANUAL (CONTOH DATA BARU)")
print("="*50)

# Data contoh: [Cement, BlastFurnaceSlag, FlyAsh, Water, Superplasticizer, CoarseAggregate, FineAggregate, Age]
contoh_data = [
    [350, 0, 0, 180, 0, 1000, 800, 28],
    [500, 0, 50, 160, 5, 950, 700, 14],
    [200, 150, 0, 190, 0, 1050, 750, 90],
    [400, 0, 100, 170, 10, 900, 720, 56],
    [150, 200, 80, 200, 4, 980, 810, 365]
]

contoh_df = pd.DataFrame(contoh_data, columns=X.columns)
contoh_scaled = scaler.transform(contoh_df)
contoh_prob = model.predict(contoh_scaled)
contoh_pred = np.argmax(contoh_prob, axis=1)
contoh_label = le.inverse_transform(contoh_pred)

print("\n{:<5} {:<70} {:<15} {:<20}".format(
    "No", "Fitur (C, BFS, FA, W, SP, CA, FAge, Age)", "Kategori Prediksi", "Probabilitas (%)"))
print("-" * 110)

for i in range(len(contoh_data)):
    fitur_str = ", ".join([f"{val:.1f}" for val in contoh_data[i]])
    prob_str = ", ".join([f"{p:.2f}" for p in contoh_prob[i]])
    print("{:<5} {:<70} {:<15} {:<20}".format(
        i+1, fitur_str, contoh_label[i], prob_str))

# -----------------------------------------------------------------------------
# 14. Kesimpulan dan Rekomendasi
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("KESIMPULAN")
print("="*50)
print("1. Model JST berhasil dilatih dengan arsitektur yang mencegah overfitting.")
print("2. Performa model pada data testing menunjukkan akurasi sekitar {:.2f}%.".format(accuracy*100))
print("3. Model, scaler, dan encoder telah disimpan untuk keperluan deployment.")
print("4. Simulasi prediksi pada data baru menunjukkan hasil yang konsisten.")
print("5. Tiga rekomendasi perbaikan berikut telah diimplementasikan secara otomatis:")
print("   a. Regularisasi L2 dan dropout untuk mengurangi overfitting.")
print("   b. Early stopping dan reduksi learning rate untuk optimasi pelatihan.")
print("   c. Standarisasi fitur untuk mempercepat konvergensi dan stabilitas.")
print("="*50)
```

```python
# =============================================================================
# PROGRAM KLASIFIKASI KUAT TEKAN BETON MENGGUNAKAN JARINGAN SARAF TIRUAN
# -----------------------------------------------------------------------------
# Deskripsi:
# Program ini mengimplementasikan Jaringan Saraf Tiruan (JST) dengan propagasi maju
# (forward propagation) untuk memprediksi kategori kuat tekan beton
# (rendah, sedang, tinggi) berdasarkan 8 fitur komposisi bahan dan umur beton.
# Dataset diperoleh melalui dua opsi: unggah manual file Excel atau unduh otomatis
# dari UCI Repository (ID=165). Evaluasi menggunakan akurasi, presisi, recall,
# dan F1-score.
#
# Perbaikan Analitik:
# 1. Konversi tipe data numerik secara paksa untuk menghindari error 'str vs float'.
# 2. Penanganan missing values dan data tidak valid setelah konversi.
# 3. Validasi struktur dataset dan peringatan dini jika terdapat anomali.
# 4. Menggunakan Input layer eksplisit pada model untuk menghilangkan warning Keras.
# 5. Menyimpan model dalam format .keras (native) untuk menghindari warning legacy.
#
# Fitur Utama:
# - Dua metode pemuatan data (manual/otomatis) dengan validasi ketat.
# - Eksplorasi data, visualisasi distribusi, dan deteksi outlier.
# - Diskritisasi target menjadi 3 kelas seimbang (berbasis persentil).
# - Arsitektur JST dengan regularisasi L2, dropout, early stopping, dan reduksi LR.
# - Evaluasi lengkap (akurasi, presisi, recall, F1, confusion matrix).
# - Penyimpanan model (.keras), scaler, dan label encoder.
# - Simulasi deployment pada 5 sampel uji acak dan 5 data buatan manual.
#
# Catatan: Program ini dirancang untuk dieksekusi di Google Colab.
# =============================================================================

# -----------------------------------------------------------------------------
# 1. Instalasi dan Impor Pustaka
# -----------------------------------------------------------------------------
!pip install -q scikit-learn matplotlib seaborn pandas numpy tensorflow joblib openpyxl ucimlrepo

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, classification_report, confusion_matrix)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import joblib
import io
from google.colab import files
from ucimlrepo import fetch_ucirepo

# -----------------------------------------------------------------------------
# 2. Pemuatan Dataset dengan Dua Opsi dan Penanganan Error
# -----------------------------------------------------------------------------
print("Pilih metode pemuatan dataset:")
print("1. Unggah file Excel manual (Concrete_Data.xls)")
print("2. Unduh otomatis dari UCI Repository (ucimlrepo)")
pilihan = input("Masukkan pilihan (1/2): ").strip()

if pilihan == '1':
    # Opsi 1: Unggah manual
    print("\nSilakan unggah file Concrete_Data.xls")
    uploaded = files.upload()
    if len(uploaded) == 0:
        raise ValueError("Tidak ada file yang diunggah. Program dihentikan.")
    filename = list(uploaded.keys())[0]
    print(f"File yang diunggah: {filename}")
    try:
        # Baca file Excel tanpa header (asumsi data mentah)
        df = pd.read_excel(io.BytesIO(uploaded[filename]), header=None, sheet_name=0)
    except Exception as e:
        raise ValueError(f"Gagal membaca file. Pastikan format file benar (xls/xlsx). Error: {e}")
    if df.shape[1] != 9:
        raise ValueError(f"Jumlah kolom tidak sesuai (diharapkan 9, ditemukan {df.shape[1]}).")
    # Beri nama kolom
    df.columns = ['Cement', 'BlastFurnaceSlag', 'FlyAsh', 'Water',
                  'Superplasticizer', 'CoarseAggregate', 'FineAggregate',
                  'Age', 'Strength']
    print("\nDataset berhasil dimuat melalui unggahan manual.")

elif pilihan == '2':
    # Opsi 2: Unduh dari UCI
    print("\nMengunduh dataset Concrete Compressive Strength dari UCI...")
    try:
        concrete = fetch_ucirepo(id=165)
        print("\nMETADATA:")
        print(concrete.metadata)
        print("\nINFORMASI VARIABEL:")
        print(concrete.variables)
        X = concrete.data.features
        y = concrete.data.targets
        df = pd.concat([X, y], axis=1)
        df.rename(columns={'Concrete compressive strength': 'Strength'}, inplace=True)
        print("\nDataset berhasil dimuat dari UCI Repository.")
    except Exception as e:
        raise RuntimeError(f"Gagal mengunduh dari UCI. Periksa koneksi internet. Error: {e}")
else:
    raise ValueError("Pilihan tidak valid. Harap masukkan 1 atau 2.")

# -----------------------------------------------------------------------------
# 3. Konversi Tipe Data Numerik (Perbaikan 1)
# -----------------------------------------------------------------------------
# Pastikan semua kolom numerik (kecuali target kategorikal nanti) dikonversi ke float.
# Error 'str vs float' biasanya muncul karena ada data string.
for col in df.columns:
    if col != 'Strength':  # Strength akan dikategorikan nanti, tapi harus numerik dulu
        df[col] = pd.to_numeric(df[col], errors='coerce')
# Konversi juga kolom Strength ke numerik
df['Strength'] = pd.to_numeric(df['Strength'], errors='coerce')

# -----------------------------------------------------------------------------
# 4. Penanganan Missing Values dan Data Tidak Valid (Perbaikan 2)
# -----------------------------------------------------------------------------
print("\nJumlah nilai hilang per kolom (setelah konversi):")
print(df.isnull().sum())
if df.isnull().sum().sum() > 0:
    print("Terdeteksi nilai hilang. Menghapus baris dengan nilai hilang...")
    df.dropna(inplace=True)
    print(f"Ukuran dataset setelah penghapusan: {df.shape}")

# Validasi apakah masih ada data yang tersisa
if df.shape[0] == 0:
    raise ValueError("Dataset kosong setelah pembersihan. Periksa file input.")

# -----------------------------------------------------------------------------
# 5. Validasi Struktur Dataset (Perbaikan 3)
# -----------------------------------------------------------------------------
# Pastikan tidak ada kolom string tersisa (kecuali nanti kategori)
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"Peringatan: Kolom {col} masih bertipe object. Konversi paksa ke numerik.")
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df.dropna(inplace=True)  # bersihkan lagi jika ada yang gagal konversi

print("\nTipe data setelah konversi:")
print(df.dtypes)

# Tampilkan informasi dasar dataset
print("\nDimensi data:", df.shape)
print("\nLima baris pertama:")
print(df.head())
print("\nInformasi dataset:")
print(df.info())
print("\nStatistik deskriptif:")
print(df.describe())

# -----------------------------------------------------------------------------
# 6. Eksplorasi Data dan Deteksi Outlier
# -----------------------------------------------------------------------------
# Deteksi outlier (hanya peringatan, tidak dihapus)
Q1 = df['Strength'].quantile(0.25)
Q3 = df['Strength'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Strength'] < (Q1 - 1.5 * IQR)) | (df['Strength'] > (Q3 + 1.5 * IQR))]
if len(outliers) > 0:
    print(f"Terdeteksi {len(outliers)} outlier pada target. (Tetap dipertahankan untuk keutuhan data)")

# Visualisasi distribusi target asli
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(df['Strength'], bins=30, kde=True)
plt.title('Distribusi Kuat Tekan Beton (MPa)')
plt.subplot(1,2,2)
sns.boxplot(x=df['Strength'])
plt.title('Boxplot Kuat Tekan Beton')
plt.tight_layout()
plt.show()

# -----------------------------------------------------------------------------
# 7. Diskritisasi Target Menjadi Tiga Kategori (Seimbang)
# -----------------------------------------------------------------------------
percentiles = [0, 33.3, 66.7, 100]
bin_edges = np.percentile(df['Strength'], percentiles)
# Sesuaikan tepi agar mencakup semua data
bin_edges[0] -= 0.001
bin_edges[-1] += 0.001
labels = ['Rendah', 'Sedang', 'Tinggi']
df['Kategori'] = pd.cut(df['Strength'], bins=bin_edges, labels=labels, include_lowest=True)

print("\nDistribusi kategori setelah diskritisasi:")
print(df['Kategori'].value_counts())

# Visualisasi distribusi kategori
sns.countplot(x='Kategori', data=df)
plt.title('Jumlah Sampel per Kategori')
plt.show()

# -----------------------------------------------------------------------------
# 8. Persiapan Data untuk Pemodelan
# -----------------------------------------------------------------------------
X = df.drop(['Strength', 'Kategori'], axis=1)
y = df['Kategori']

# Encode label kategori menjadi angka
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print("\nMapping label ke angka:", dict(zip(le.classes_, le.transform(le.classes_))))

# Stratified split (menjaga proporsi kelas)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Standarisasi fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nUkuran data training : {X_train_scaled.shape}")
print(f"Ukuran data testing  : {X_test_scaled.shape}")

# -----------------------------------------------------------------------------
# 9. Membangun Arsitektur JST dengan Input Layer Eksplisit (Perbaikan 4)
# -----------------------------------------------------------------------------
model = keras.Sequential([
    Input(shape=(X_train.shape[1],), name='input_layer'),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001),
                 name='hidden_1'),
    layers.Dropout(0.3, name='dropout_1'),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001),
                 name='hidden_2'),
    layers.Dropout(0.3, name='dropout_2'),
    layers.Dense(32, activation='relu', name='hidden_3'),
    layers.Dense(3, activation='softmax', name='output')
])

model.summary()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callback untuk mencegah overfitting dan mengatur learning rate
early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001)

# -----------------------------------------------------------------------------
# 10. Pelatihan Model
# -----------------------------------------------------------------------------
history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=32,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# -----------------------------------------------------------------------------
# 11. Visualisasi Kurva Pelatihan
# -----------------------------------------------------------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss selama pelatihan')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Akurasi selama pelatihan')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------------------------------------------------------
# 12. Evaluasi Model pada Data Testing
# -----------------------------------------------------------------------------
y_pred_prob = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_prob, axis=1)

accuracy = accuracy_score(y_test, y_pred)
precision_macro = precision_score(y_test, y_pred, average='macro')
recall_macro = recall_score(y_test, y_pred, average='macro')
f1_macro = f1_score(y_test, y_pred, average='macro')
precision_weighted = precision_score(y_test, y_pred, average='weighted')
recall_weighted = recall_score(y_test, y_pred, average='weighted')
f1_weighted = f1_score(y_test, y_pred, average='weighted')

print("\n" + "="*50)
print("          HASIL EVALUASI MODEL PADA DATA TESTING")
print("="*50)
print(f"Akurasi              : {accuracy:.4f}")
print(f"Presisi (macro)      : {precision_macro:.4f}")
print(f"Recall (macro)       : {recall_macro:.4f}")
print(f"F1-score (macro)     : {f1_macro:.4f}")
print(f"Presisi (weighted)   : {precision_weighted:.4f}")
print(f"Recall (weighted)    : {recall_weighted:.4f}")
print(f"F1-score (weighted)  : {f1_weighted:.4f}")

print("\nLaporan klasifikasi per kelas:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# -----------------------------------------------------------------------------
# 13. Visualisasi Confusion Matrix
# -----------------------------------------------------------------------------
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Kelas Prediksi')
plt.ylabel('Kelas Aktual')
plt.title('Confusion Matrix')
plt.show()

# -----------------------------------------------------------------------------
# 14. Penyimpanan Model dalam Format .keras (Perbaikan 5)
# -----------------------------------------------------------------------------
model.save('model_kuat_tekan_beton.keras')
joblib.dump(scaler, 'scaler.save')
joblib.dump(le, 'label_encoder.save')
print("\nModel (format .keras), scaler, dan encoder berhasil disimpan.")

# -----------------------------------------------------------------------------
# 15. Simulasi Deployment: Prediksi 5 Sampel Acak dari Data Uji
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("   SIMULASI DEPLOYMENT: PREDIKSI PADA 5 DATA UJI")
print("="*50)

n_samples = 5
sample_indices = np.random.choice(len(X_test), n_samples, replace=False)
X_sample = X_test.iloc[sample_indices]
y_actual = y_test[sample_indices]
y_actual_label = le.inverse_transform(y_actual)

X_sample_scaled = scaler.transform(X_sample)
pred_prob = model.predict(X_sample_scaled)
pred_class = np.argmax(pred_prob, axis=1)
pred_label = le.inverse_transform(pred_class)

print("\n{:<5} {:<70} {:<10} {:<10} {:<20}".format(
    "No", "Fitur (C, BFS, FA, W, SP, CA, FAge, Age)", "Aktual", "Prediksi", "Probabilitas (%)"))
print("-" * 120)

for i in range(n_samples):
    fitur_str = ", ".join([f"{val:.1f}" for val in X_sample.iloc[i]])
    prob_str = ", ".join([f"{p:.2f}" for p in pred_prob[i]])
    print("{:<5} {:<70} {:<10} {:<10} {:<20}".format(
        i+1, fitur_str, y_actual_label[i], pred_label[i], prob_str))

acc_sample = np.mean(pred_class == y_actual)
print("-" * 120)
print(f"Akurasi pada {n_samples} sampel: {acc_sample:.2f}")

# -----------------------------------------------------------------------------
# 16. Simulasi Input Manual (Contoh Data Baru)
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("   SIMULASI INPUT MANUAL (CONTOH DATA BARU)")
print("="*50)

# Data contoh: [Cement, BlastFurnaceSlag, FlyAsh, Water, Superplasticizer, CoarseAggregate, FineAggregate, Age]
contoh_data = [
    [350, 0, 0, 180, 0, 1000, 800, 28],
    [500, 0, 50, 160, 5, 950, 700, 14],
    [200, 150, 0, 190, 0, 1050, 750, 90],
    [400, 0, 100, 170, 10, 900, 720, 56],
    [150, 200, 80, 200, 4, 980, 810, 365]
]

contoh_df = pd.DataFrame(contoh_data, columns=X.columns)
contoh_scaled = scaler.transform(contoh_df)
contoh_prob = model.predict(contoh_scaled)
contoh_pred = np.argmax(contoh_prob, axis=1)
contoh_label = le.inverse_transform(contoh_pred)

print("\n{:<5} {:<70} {:<15} {:<20}".format(
    "No", "Fitur (C, BFS, FA, W, SP, CA, FAge, Age)", "Kategori Prediksi", "Probabilitas (%)"))
print("-" * 110)

for i in range(len(contoh_data)):
    fitur_str = ", ".join([f"{val:.1f}" for val in contoh_data[i]])
    prob_str = ", ".join([f"{p:.2f}" for p in contoh_prob[i]])
    print("{:<5} {:<70} {:<15} {:<20}".format(
        i+1, fitur_str, contoh_label[i], prob_str))

# -----------------------------------------------------------------------------
# 17. Kesimpulan dan Rekomendasi
# -----------------------------------------------------------------------------
print("\n" + "="*50)
print("KESIMPULAN")
print("="*50)
print("1. Model JST berhasil dilatih dengan arsitektur yang mencegah overfitting.")
print("2. Performa model pada data testing menunjukkan akurasi sekitar {:.2f}%.".format(accuracy*100))
print("3. Model, scaler, dan encoder telah disimpan (format .keras dan .save).")
print("4. Simulasi prediksi pada data baru menunjukkan hasil yang konsisten.")
print("5. Tiga rekomendasi perbaikan berikut telah diimplementasikan secara otomatis:")
print("   a. Konversi tipe data numerik dan penanganan missing values.")
print("   b. Penggunaan Input layer eksplisit untuk menghilangkan warning Keras.")
print("   c. Penyimpanan model dalam format .keras (native) untuk kompatibilitas masa depan.")
print("="*50)
```
